## Introduction

This project provides a Python implementation of the analytical framework presented in Köhler, Heckens, and Guhr (2026). The core of this repository is the iPython Notebook `extreme_value_analysis_correlated_systems_draft.ipynb`, which contains a comprehensive suite of functions to replicate the paper's findings.

The pipeline addresses a critical challenge in quantitative finance and complex systems: how to quantify the probability of simultaneous, catastrophic failures in high-dimensional networks where standard multivariate EVT is computationally intractable due to the "Curse of Dimensionality."

The codebase operationalizes the proposed solution:
-   **Synchronizes** asynchronous, millisecond-resolution limit order book updates into a rigid 1-second grid using exact Last-Observation-Carried-Forward (LOCF) imputation.
-   **Decouples** the highly correlated financial network by rotating the system into the eigenbasis of its Pearson correlation matrix (PCA-Whitening).
-   **Estimates** tail risks on the orthogonalized modes using the Peaks-Over-Threshold (POT) approach and Maximum Likelihood Estimation (MLE) of the Generalized Pareto Distribution (GPD).
-   **Neutralizes** non-stationarity by computing intraday volatility profiles and applying rolling-window local thresholds, proving that volatility clustering is largely an artifact of background variance.

## Theoretical Background

The implemented methods combine techniques from Random Matrix Theory (RMT), Microstructure Econometrics, and Extreme Value Theory.

**1. Spectral Decoupling (PCA-Whitening):**
To handle cross-sectional correlations, the normalized return matrix $M$ is rotated into uncorrelated, variance-normalized eigenmodes $R$ using the spectral decomposition of the correlation matrix $C = U \Lambda U^\dagger$:
$$ R = \Lambda^{-1/2} U^\dagger M $$

**2. Peaks-Over-Threshold (POT) EVT:**
The conditional excess distribution over a high threshold $u$ converges to the Generalized Pareto Distribution (GPD). The parameters are estimated via MLE:
$$ G_{\gamma, \sigma}^{\text{GPD}}(x) = \begin{cases} 1 - (1 + \frac{\gamma x}{\sigma})^{-1/\gamma} & \gamma \neq 0 \\ 1 - e^{-x/\sigma} & \gamma = 0 \end{cases} $$

**3. Extremal Clustering:**
The clustering of extreme events is quantified using the Ferro-Segers Extremal Index $\hat{\Theta} \in [0, 1]$, derived from inter-exceedance times $\tau_i$:
$$ \hat{\Theta} = \begin{cases} \min \left( \frac{2 \langle (\tau_i - 1) \rangle^2}{\langle (\tau_i - 1)(\tau_i - 2) \rangle}, 1 \right) & \max \tau_i > 2 \\ \min \left( \frac{2 \langle \tau_i \rangle^2}{\langle \tau_i^2 \rangle}, 1 \right) & \max \tau_i \le 2 \end{cases} $$

**4. Dynamic Residualization:**
To isolate true stochastic risk from deterministic intraday seasonality, the modes are divided by their average volatility profile:
$$ \tilde{R}_k(t) = R_k(t) / V_k(t \pmod{T_{\text{day}}}) $$

## Methodology Implemented

The core analytical steps directly implement the methodology from the paper:

1.  **Data Engineering (Tasks 1-8):** Ingests raw TAQ data, applies microstructure filters (e.g., $ASK > BID$), synchronizes to a 1-second grid via LOCF, computes log-returns, and standardizes to zero mean and unit variance.
2.  **Spectral Analysis (Tasks 9-12):** Computes the dense Pearson correlation matrix, extracts eigenvalues/eigenvectors, performs PCA-whitening, and validates the economic interpretation of the modes (Market, Energy, Utility).
3.  **Static EVT (Tasks 13-16):** Extracts static POT exceedances, fits the GPD via MLE, computes asymptotic standard errors, and estimates the Ferro-Segers Extremal Index.
4.  **Block Maxima Inference (Tasks 17-18):** Computes absolute-return ACF and analytically maps POT parameters to Generalized Extreme Value (GEV) distributions, validating against empirical daily maxima via KS tests.
5.  **Dynamic Analysis (Tasks 19-20):** Computes intraday volatility profiles, deseasonalizes the modes, and applies rolling-window local thresholds to neutralize non-stationary background variance.
6.  **Robustness & Assembly (Tasks 21-24):** Executes threshold and window-size sensitivity analyses, evaluates the manuscript's qualitative claims using strict quantitative rules, and compiles a cryptographic dossier.

# Paper

Title: "*Extreme Value Analysis for Finite, Multivariate and Correlated Systems with Finance as an Example*"

Authors: Benjamin Köhler, Anton J. Heckens, Thomas Guhr

E-Journal Submission Date: 5 March 2026

Link: https://arxiv.org/abs/2603.05260

Abstract:

Extreme values and the tail behavior of probability distributions are essential for quantifying and mitigating risk in complex systems of all kinds. In multivariate settings, accounting for correlations is crucial. Although extreme value analysis for infinite correlated systems remains an open challenge, we propose a practical framework for handling a large but finite number of correlated time series. We develop our approach for finance as a concrete example but emphasize its generality. We study the extremal behavior of high-frequency stock returns after rotating them into the eigenbasis of the correlation matrix. This separates and extracts various collective effects, including information on the correlated market as a whole and on correlated sectoral behavior from idiosyncratic features, while allowing us to use univariate tools of extreme value analysis. This holds even for high-frequency data where discretization effects normally complicate analysis. We employ a peaks-over-threshold approach and thereby fully avoid the analysis of block maxima. We estimate the tail shape of the rotated returns while explicitly accounting for nonstationarity, a key feature in finance and many other complex systems. Our framework facilitates tail risk estimation relative to larger trends and intraday seasonalities at both market and sectoral levels.

# Summary

### **Executive Summary**

This paper addresses the challenge of quantifying tail risk in large, finite, multivariate systems that exhibit strong correlations and non-stationarity—specifically high-frequency financial markets. The authors propose a framework that combines **Principal Component Analysis (PCA)** (via spectral decomposition of the correlation matrix) with **Univariate Extreme Value Theory (EVT)** using a **Peaks-Over-Threshold (POT)** approach.

By rotating high-frequency stock returns (1-second resolution) into their eigenbasis, the authors disentangle collective market dynamics (systemic risk) from idiosyncratic noise. A key contribution is the treatment of non-stationarity: by employing **local, rolling-window thresholds**, they demonstrate that the clustering of extreme events (volatility clustering) is largely an artifact of non-stationary background volatility. Once normalized locally, extreme arrivals approximate a Poisson process while retaining heavy-tailed (Fréchet) characteristics.

<br>

### **Theoretical Foundations & Problem Statement**

The authors identify a gap in current literature: while EVT is well-developed for univariate i.i.d. series, and infinite correlated systems are studied in statistical physics, practical frameworks for *finite* but *large* correlated datasets (like the S&P 500) remain sparse.

#### **1.1 Univariate EVT Refresher**
The paper leverages the **Pickands-Balkema-DeHaan theorem**, which states that for a large class of distributions, the conditional excess distribution over a high threshold $u$ converges to the **Generalized Pareto Distribution (GPD)**:
$$ G_{\gamma, \sigma}^{\text{GPD}}(x) = \begin{cases} 1 - (1 + \frac{\gamma x}{\sigma})^{-1/\gamma} & \gamma \neq 0 \\ 1 - e^{-x/\sigma} & \gamma = 0 \end{cases} $$
*   **$\gamma$ (Shape Parameter):** Determines tail behavior. $\gamma > 0$ implies heavy tails (Fréchet), typical in finance.
*   **Preference for POT:** The authors explicitly reject the Block Maxima (BM) approach in favor of POT to maximize data efficiency and avoid arbitrary block sizing.

#### **1.2 Handling Dependence**
Financial time series violate the i.i.d. assumption due to serial correlation (volatility clustering). The authors utilize the **Extremal Index $\Theta \in [0, 1]$**, which measures the reciprocal of the mean cluster size.
*   $\Theta = 1$: No clustering (Poisson process).
*   $\Theta < 1$: Extremes cluster together.

<br>

### **2. Methodology: The Multivariate Framework**

The core innovation is the reduction of a multivariate problem into independent univariate problems via geometric rotation.

#### **2.1 Data Processing**
*   **Dataset:** NYSE Trade and Quote (TAQ) data for 2014.
*   **Scope:** $K=479$ liquid stocks from the S&P 500.
*   **Resolution:** 1-second intervals ($T \approx 5.4 \times 10^6$ data points per stock).
*   **Metric:** Logarithmic returns normalized to zero mean and unit variance.

#### **2.2 Spectral Decomposition and Rotation**
The authors compute the Pearson correlation matrix $C$ and perform an eigendecomposition:
$$ C = U \Lambda U^\dagger $$
where $\Lambda$ is the diagonal matrix of eigenvalues and $U$ contains the eigenvectors. They then define **Rotated Returns ($R$)** via PCA-whitening:
$$ R = \Lambda^{-1/2} U^\dagger M $$
*   **Interpretation:** This transformation creates uncorrelated, variance-normalized time series (modes).
*   **Mode Identification:**
    *   **$k=1$ (Largest Eigenvalue):** Represents the "Market Mode" (collective behavior).
    *   **$k=2$:** Identified as the Energy sector (specific to 2014 dynamics).
    *   **$k=3$:** Identified as the Utility sector.
    *   **Bulk:** Random noise (Random Matrix Theory regime).

<br>

### **3. Empirical Results: Static Analysis**

The authors first apply POT to the marginal distributions of the rotated modes without correcting for intraday seasonality.

#### **3.1 Tail Behavior**
*   **Heavy Tails:** The estimated shape parameters $\hat{\gamma}$ are positive ($\gamma > 0$), confirming Fréchet-type behavior for the market and sector modes.
*   **Convergence:** The fit stabilizes as the threshold $u$ increases, validated by Q-Q plots and Normalized Root Mean Square Deviation (NRMSD).

#### **3.2 Clustering in Eigenmodes**
*   **Observation:** The Extremal Index $\hat{\Theta}$ is significantly less than 1 for the top modes.
*   **Implication:** Even after rotating into the eigenbasis, the "market mode" and "sector modes" exhibit strong serial dependence (clustering of extremes). This indicates that volatility clustering is a systemic feature, not just an idiosyncratic one.

#### **3.3 Inferring Block Maxima**
A theoretical bridge is built: the authors successfully reconstruct the distribution of *daily maxima* using only the parameters derived from the POT analysis of 1-second data. This validates that POT captures all necessary information regarding the extremes.

<br>

### **4. Empirical Results: Dynamic Analysis (Non-Stationarity)**

The paper argues that "static" EVT is insufficient for high-frequency data due to intraday seasonality (the "U-shape" volatility profile) and regime shifts.

#### **4.1 Residual Risk and Volatility Profiling**
They define a residual time series $\tilde{R}_k(t)$ by normalizing returns by the intraday volatility profile $V_k(t_{\text{intra}})$:
$$ \tilde{R}_k(t) = R_k(t) / V_k(t \pmod{T_{\text{day}}}) $$
This removes deterministic intraday patterns (e.g., market open/close volatility spikes).

#### **4.2 Local Threshold Estimation**
Instead of a fixed threshold $u$, the authors implement a **dynamic threshold $u(t)$** based on a rolling quantile (e.g., 99.9% over a 10,000-second window).

#### **4.3 Key Findings from Dynamic Analysis**
1.  **Clustering Vanishes:** When using local thresholds on residual data, the Extremal Index $\hat{\Theta}$ approaches 1.
    *   *Econometric Insight:* This suggests that the "clustering" of extremes observed in static analysis is largely driven by the non-stationarity of the background volatility. When risk is defined relative to the *current* market state, extreme arrivals become Poissonian (random).
2.  **Tail Shape Stability:** The tail shape parameter $\gamma$ remains positive (Fréchet), indicating that while the *timing* of extremes becomes random under local normalization, the *magnitude* distribution remains heavy-tailed.
3.  **Ordering Changes:** The risk ranking of different modes (Market vs. Energy) shifts when moving from static to local analysis, highlighting that different sectors have different sensitivities to non-stationarity.

<br>

### **5. Critical Synthesis**

#### **Algorithmic Implications**
The rotation method ($R = \Lambda^{-1/2} U^\dagger M$) is computationally efficient and aligns with standard dimensionality reduction techniques in Machine Learning. It effectively compresses the risk of 479 assets into a few interpretable "risk factors" (eigenmodes) that can be monitored in real-time.

#### **Financial Implications**
The distinction between **Marginal Risk** (static) and **Residual Risk** (dynamic) is crucial for portfolio management:
*   **Strategic (Long-term):** Use the static marginal distribution to understand annual crash risks.
*   **Tactical (High-Frequency):** Use the local threshold approach. The finding that $\Theta \to 1$ with local thresholds implies that HFT algorithms cannot easily predict the *exact timing* of the next extreme event based on clustering alone once the local volatility regime is accounted for.

#### **Conclusion**
Köhler, Heckens, and Guhr provide a robust mathematical framework for analyzing multivariate extremes. By combining **spectral decomposition** to handle cross-sectional correlations and **local POT estimation** to handle temporal non-stationarity, they offer a scalable method for systemic risk monitoring. The paper successfully demonstrates that "extremeness" in finance is a relative concept, heavily dependent on the local volatility regime.

*   **Outputs:** `df_session`, `df_pre_session`, and a sorted list of the 248 retained dates.

#### **5. `assign_second_buckets_and_select`**
*   **Inputs:** `df_session`, `df_pre_session`, and the frozen manifest.
*   **Processes:** Truncates the nanosecond timestamps downward to the containing whole second. Uses an $O(N)$ hash-table scan (`drop_duplicates(keep='last')`) to select the final quote per second. Computes the midpoint price.
*   **Outputs:** Sparse midpoint DataFrames for both session and pre-session data.
*   **Data Transformation:** Asynchronous tick data $\to$ Sparse 1-second synchronous grid.
*   **Role in Research Pipeline:** Implements the core synchronization rule: *"use the last available midpoint price in each second $t$,"* and computes Eq. (12):
    $$ m_k(t) = \frac{a_k(t) + b_k(t)}{2} $$

#### **6. `build_complete_midpoint_panel`**
*   **Inputs:** Sparse midpoint DataFrames, date list, and the frozen manifest.
*   **Processes:** Pre-allocates a 20.4 GB memory-mapped array. Maps sparse observations to the dense grid. Initializes the first second of the day using pre-session data, then applies a highly optimized Last-Observation-Carried-Forward (LOCF) algorithm strictly within day boundaries.
*   **Outputs:** The file path to the dense 3D midpoint memmap and a validation report.
*   **Data Transformation:** Sparse relational table $\to$ Dense, fully populated 3D tensor / 2D matrix on disk.
*   **Role in Research Pipeline:** Finalizes the synchronous grid required for multivariate matrix algebra.

#### **7. `compute_log_returns`**
*   **Inputs:** The midpoint panel memmap path, date list, and the frozen manifest.
*   **Processes:** Computes log differences along the intraday axis to strictly prevent cross-day (overnight) returns. Flattens the result into a 2D matrix and constructs a day-boundary index map.
*   **Outputs:** The file path to the return matrix $G$ memmap, the day-boundary map, and a validation report.
*   **Data Transformation:** Price matrix $m_k(t)$ $\to$ Return matrix $G_k(t)$.
*   **Role in Research Pipeline:** Implements Eq. (13):
    $$ G_k(t, \Delta t) = \ln m_k(t + \Delta t) - \ln m_k(t) $$

#### **8. `standardize_returns`**
*   **Inputs:** The return matrix $G$ memmap path and the frozen manifest.
*   **Processes:** Computes the population mean and standard deviation (ddof=0) for each stock. Performs in-place standardization to conserve memory. Validates that every row has exactly zero mean and unit variance.
*   **Outputs:** The file path to the normalized matrix $M$ memmap, mean/std vectors, and a validation report.
*   **Data Transformation:** Raw returns $G_k(t)$ $\to$ Z-scored, dimensionless returns $M_k(t)$.
*   **Role in Research Pipeline:** Implements Eq. (14):
    $$ M_k(t) = \frac{G_k(t) - \langle G_k(t) \rangle_t}{\sigma_k} $$

### **Phase 2: Spectral Analysis and PCA-Whitening**

#### **9. `compute_correlation_matrix`**
*   **Inputs:** The normalized matrix $M$ memmap path and the frozen manifest.
*   **Processes:** Computes the dense matrix product using a chunkwise accumulation strategy to bound RAM usage. Explicitly symmetrizes the result to correct floating-point drift and validates that the diagonal is exactly 1.0.
*   **Outputs:** The dense, symmetric Pearson correlation matrix $C \in \mathbb{R}^{K \times K}$.
*   **Data Transformation:** $K \times T$ time-series matrix $\to$ $K \times K$ spatial topology matrix.
*   **Role in Research Pipeline:** Implements Eq. (19):
    $$ C = \frac{1}{T} M M^\dagger $$

#### **10. `spectral_decomposition`**
*   **Inputs:** The correlation matrix $C$ and the frozen manifest.
*   **Processes:** Executes a symmetric eigendecomposition (`scipy.linalg.eigh`). Sorts eigenvalues descending, forces memory contiguity, validates the reconstruction residual, and floors near-zero negative eigenvalues to prevent division-by-zero.
*   **Outputs:** Eigenvalues $\Lambda$, Eigenvectors $U$, and decomposition diagnostics.
*   **Data Transformation:** Dense covariance matrix $\to$ Orthogonal basis vectors and variance magnitudes.
*   **Role in Research Pipeline:** Implements Eq. (20):
    $$ C = U \Lambda U^\dagger $$

#### **11. `construct_whitened_modes`**
*   **Inputs:** The normalized matrix $M$, $U$, $\Lambda$, and the frozen manifest.
*   **Processes:** Computes the unscaled rotation $Y = U^\top M$. Computes the whitened modes $R$ by scaling rows by $\Lambda^{-1/2}$. Validates that the empirical covariance of $R$ is the Identity matrix. Applies a deterministic sign convention.
*   **Outputs:** Memmap paths for $Y$ and $R$, the sign-adjusted $U$, and mode metadata.
*   **Data Transformation:** Correlated physical space $M$ $\to$ Uncorrelated, variance-normalized spectral space $R$.
*   **Role in Research Pipeline:** Implements the core dimensionality reduction and decoupling step (Eq. 21):
    $$ R = \Lambda^{-1/2} U^\dagger M $$

#### **12. `generate_spectral_diagnostics`**
*   **Inputs:** $\Lambda$, $U$, $Y$, $R$, and the frozen manifest.
*   **Processes:** Computes the eigenvalue density histogram, evaluates market-mode uniformity, calculates sector concentration weights, and computes empirical tail densities for $Y$, $R$, and the aggregated bulk.
*   **Outputs:** A comprehensive spectral diagnostics dictionary.
*   **Data Transformation:** Raw matrices $\to$ Aggregated statistical distributions.
*   **Role in Research Pipeline:** Validates the Random Matrix Theory interpretation of the modes, generating the data required for Figures 5, 6, 7, and 8.

<br>

### **Phase 3: Static Extreme Value Theory (EVT)**

#### **13. `extract_static_pot_exceedances`**
*   **Inputs:** The whitened modes $R$ memmap path and the frozen manifest.
*   **Processes:** Defines positive ($R$) and negative ($-R$) tail series. Computes static thresholds $u_\alpha$ using type-7 linear interpolation. Extracts exceedance magnitudes and global time positions using strict inequality ($X > u$).
*   **Outputs:** A nested dictionary of exceedance objects.
*   **Data Transformation:** Continuous time series $\to$ Discrete sets of extreme events.
*   **Role in Research Pipeline:** Implements the Peaks-Over-Threshold (POT) extraction. As stated in the text: *"We employ a peaks-over-threshold approach and thereby fully avoid the analysis of block maxima."*

#### **14. `fit_gpd_mle`**
*   **Inputs:** The exceedance objects and the frozen manifest.
*   **Processes:** Maximizes the GPD log-likelihood using L-BFGS-B. Employs the reparameterization $\sigma = e^\eta$ and a massive penalty barrier to strictly enforce support constraints. Uses `log1p` for numerical stability as $\gamma \to 0$.
*   **Outputs:** A DataFrame containing the fitted parameters ($\hat{\gamma}, \hat{\sigma}$) and convergence metrics.
*   **Data Transformation:** Raw exceedance magnitudes $\to$ Statistical shape and scale parameters.
*   **Role in Research Pipeline:** Implements the Maximum Likelihood Estimation of Eq. (16):
    $$ l(\gamma, \sigma) = -n \ln \sigma - \left(1 + \frac{1}{\gamma}\right) \sum_{i=1}^n \ln\left(1 + \gamma \frac{x_i}{\sigma}\right) $$

#### **15. `evaluate_gpd_fit_quality`**
*   **Inputs:** The GPD fit table, exceedance objects, and the frozen manifest.
*   **Processes:** Computes asymptotic standard errors via the Fisher Information matrix. Constructs Q-Q diagnostic arrays. Computes the NRMSD ($\rho$). Programmatically identifies credible threshold regions based on $\hat{\gamma}$ stability.
*   **Outputs:** The augmented GPD table, Q-Q data, and credible regions metadata.
*   **Data Transformation:** Point estimates $\to$ Confidence intervals and goodness-of-fit metrics.
*   **Role in Research Pipeline:** Implements Eq. (17) for standard errors and Eq. (18) for the NRMSD to validate the GPD fit quality.

#### **16. `estimate_extremal_index`**
*   **Inputs:** The exceedance objects, augmented GPD table, and the frozen manifest.
*   **Processes:** Computes inter-exceedance times $\tau_i$. Applies the piecewise Ferro-Segers estimator, carefully casting to `float64` to prevent integer overflow during moment calculations.
*   **Outputs:** The final static EVT DataFrame including $\hat{\Theta}$.
*   **Data Transformation:** Exceedance timestamps $\to$ Scalar clustering index.
*   **Role in Research Pipeline:** Implements Eq. (15) to quantify short-range clustering:
    $$ \hat{\Theta} = \min \left( \frac{2 \langle (\tau_i - 1) \rangle_i^2}{\langle (\tau_i - 1)(\tau_i - 2) \rangle_i}, 1 \right) $$

<br>

### **Phase 4: ACF and Block Maxima Inference**

#### **17. `compute_absolute_return_acf`**
*   **Inputs:** The whitened modes $R$ memmap path and the frozen manifest.
*   **Processes:** Extracts $|R_k|$. Computes the empirical autocovariance up to 10,000 lags using an $O(T \log T)$ FFT-based algorithm with zero-padding to prevent circular correlation. Normalizes by the full-sample variance.
*   **Outputs:** ACF vectors and persistence diagnostics.
*   **Data Transformation:** Time series $\to$ Autocorrelation function vectors.
*   **Role in Research Pipeline:** Validates the premise: *"Long range dependency in absolute values of financial returns is well-established"* (generates data for Fig 14).

#### **18. `infer_gev_from_pot`**
*   **Inputs:** The static EVT table, $R$ memmap path, day-boundary map, and the frozen manifest.
*   **Processes:** Maps POT parameters to GEV parameters ($a, b$). Evaluates the analytical GEV density. Extracts empirical daily maxima. Performs a Kolmogorov-Smirnov test to validate the match.
*   **Outputs:** GEV inference artifact containing parameters, density curves, and KS test results.
*   **Data Transformation:** POT parameters ($\gamma, \sigma, u$) $\to$ Block maxima parameters ($a, b$) and density curves.
*   **Role in Research Pipeline:** Implements Eq. (6) ($\sigma = a + \gamma(u - b)$) and Eq. (3) (GEV PDF) to prove that POT implies block maxima statistics without requiring block maxima estimation (Fig 16).

<br>

### **Phase 5: Dynamic Analysis (Non-Stationarity)**

#### **19. `build_residual_modes`**
*   **Inputs:** The whitened modes $R$ memmap path and the frozen manifest.
*   **Processes:** Reshapes the 1D series into a 2D matrix (Days $\times$ Intraday Seconds). Computes the mean absolute return across days to form the volatility profile $V_k$. Deseasonalizes the modes via broadcasted division. Computes residual ACF and $\hat{\Theta}$.

*   **Processes:** Computes rolling backward-looking quantiles $u(t)$ (strictly excluding the current observation to prevent look-ahead bias). Extracts local exceedances. Fits a global GPD to the pooled local exceedances. Computes dynamic $\hat{\Theta}$ and time-varying GEV location $b(t)$.
*   **Outputs:** The dynamic EVT artifact containing the results table and non-stationary density surface.
*   **Data Transformation:** Stationary residuals $\to$ Time-varying risk parameters and non-stationary density surfaces.
*   **Role in Research Pipeline:** Implements the core dynamic innovation of the paper: *"employing local, time dependent thresholds estimated from rolling quantiles to obtain a market state dependent notion of extremes."*

```python
import os
import yaml
import numpy as np
import pandas as pd
from typing import Dict, Any

# Set random seed for deterministic synthetic generation
np.random.seed(42)

def load_study_configuration(filepath: str = "config.yaml") -> Dict[str, Any]:
    """
    Loads the study configuration parameters from a YAML file into a Python dictionary.

    Purpose:
        To ingest the deterministic hyperparameters, data schemas, and EVT
        boundary conditions required by the orchestrator.

    Inputs:
        filepath (str): The path to the YAML configuration file. Default is "config.yaml".

    Outputs:
        Dict[str, Any]: A nested dictionary containing the study configuration.

    Raises:
        FileNotFoundError: If the configuration file is missing.
        yaml.YAMLError: If the file contains invalid YAML syntax.
    """
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Critical Error: Configuration file '{filepath}' not found in the working directory.")
        
    with open(filepath, "r") as file:
        config = yaml.safe_load(file)
        
    print(f"[*] Successfully loaded configuration from {filepath}")
    return config

def generate_synthetic_taq_data(config: Dict[str, Any]) -> pd.DataFrame:
    """
    Generates a high-fidelity, sparse synthetic NYSE TAQ Quote DataFrame.

    Purpose:
        To create a mathematically plausible dataset that strictly satisfies the
        pipeline's schema and initialization invariants without causing Out-Of-Memory
        (OOM) errors. It generates exactly two quotes per stock per day: one
        pre-session quote (to guarantee LOCF initialization) and one mid-session quote.

    Inputs:
        config (Dict[str, Any]): The loaded study configuration dictionary.

    Processes:
        1. Extracts the exact 479-ticker universe and 248-day calendar from the config.
        2. Generates a pre-session quote (09:39:59) and a mid-session quote (12:00:00)
           for every stock-day combination.
        3. Simulates realistic bid/ask spreads ensuring ASK > BID and sizes > 0.
        4. Enforces the strict pandas dtypes required by Task 2.

    Outputs:
        pd.DataFrame: The synthetic `df_raw_taq` DataFrame.
    """
    print("[*] Generating synthetic high-frequency TAQ data...")
    
    # 1. Extract invariants from configuration
    dp_params = config["algorithmic_parameters"]["data_preprocessing"]
    tickers = dp_params["asset_universe_exact_appendix_order"]
    excluded_dates = set(dp_params["excluded_dates"])
    target_year = dp_params["target_year"]
    
    # Generate the 248 valid business days for 2014
    all_b_days = pd.date_range(start=f"{target_year}-01-01", end=f"{target_year}-12-31", freq='B')
    valid_dates = [d for d in all_b_days if d.strftime('%Y-%m-%d') not in excluded_dates]
    
    # 2. Pre-allocate lists for vectorized DataFrame construction
    dates, times, syms, bids, asks = [], [], [], [], []
    
    # Base price for simulation
    base_price = 100.0
    
    for date in valid_dates:
        for ticker in tickers:
            # Quote 1: Pre-session initialization (Guarantees no NaNs at 09:40:00)
            dates.append(date)
            times.append(pd.Timedelta("09:39:59"))
            syms.append(ticker)
            
            # Simulate a tight spread: Bid = 99.95, Ask = 100.05
            noise = np.random.normal(0, 0.1)
            bids.append(round(base_price + noise - 0.05, 2))
            asks.append(round(base_price + noise + 0.05, 2))
            
            # Quote 2: Mid-session update
            dates.append(date)
            times.append(pd.Timedelta("12:00:00"))
            syms.append(ticker)
            
            noise = np.random.normal(0, 0.5)
            bids.append(round(base_price + noise - 0.05, 2))
            asks.append(round(base_price + noise + 0.05, 2))

    # 3. Construct the DataFrame
    df = pd.DataFrame({
        "DATE": dates,
        "TIME_M": times,
        "SYM_ROOT": syms,
        "EX": ['N'] * len(dates),
        "BID": bids,
        "BIDSIZ": [100] * len(dates),
        "ASK": asks,
        "ASKSIZ": [100] * len(dates),
        "QU_COND": ['R'] * len(dates)
    })
    
    # Add deterministic row ordering
    df["RAW_ROW_ORDER"] = np.arange(len(df))
    
    # 4. Enforce strict schema dtypes
    df["DATE"] = df["DATE"].astype("datetime64[ns]")
    df["TIME_M"] = df["TIME_M"].astype("timedelta64[ns]")
    df["SYM_ROOT"] = df["SYM_ROOT"].astype("category")
    df["EX"] = df["EX"].astype("category")
    df["BID"] = df["BID"].astype("float64")
    df["BIDSIZ"] = df["BIDSIZ"].astype("int64")
    df["ASK"] = df["ASK"].astype("float64")
    df["ASKSIZ"] = df["ASKSIZ"].astype("int64")
    df["QU_COND"] = df["QU_COND"].astype("category")
    df["RAW_ROW_ORDER"] = df["RAW_ROW_ORDER"].astype("int64")
    
    print(f"[*] Synthetic data generation complete. Shape: {df.shape}")
    return df

# Execute ingestion and generation
study_config = load_study_configuration("config.yaml")
df_raw_taq = generate_synthetic_taq_data(study_config)
```

````python
# Task 1 — Validate the ingested study configuration object

# ==============================================================================
# Task 1: Validate the ingested study configuration object
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 1: Validate the structural completeness of the configuration dictionary
# -------------------------------------------------------------------------------------------------------------------------------

def validate_structural_completeness(config: Dict[str, Any], report: Dict[str, Any]) -> None:
    """
    Validates the structural completeness of the study configuration dictionary.

    Checks for the presence of required top-level keys, the exact eight required
    sub-dictionaries within algorithmic parameters, and ensures that non-applicable
    components are explicitly set to None.

    Args:
        config (Dict[str, Any]): The raw study configuration dictionary.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Raises:
        TypeError: If the input config is not a dictionary.
        KeyError: If required top-level or sub-level keys are missing.
        ValueError: If non-applicable components are not None.
    """
    # Verify that the input is a dictionary
    if not isinstance(config, dict):
        raise TypeError(f"Configuration must be a dictionary, got {type(config).__name__}")

    # Define the required top-level keys
    required_top_keys: Set[str] = {
        "not_applicable_components",
        "raw_unprocessed_data_schema",
        "algorithmic_parameters"
    }

    # Extract actual top-level keys
    actual_top_keys: Set[str] = set(config.keys())

    # Check for missing top-level keys
    missing_top_keys: Set[str] = required_top_keys - actual_top_keys
    if missing_top_keys:
        raise KeyError(f"Missing required top-level configuration keys: {missing_top_keys}")

    # Log top-level key validation success
    report["top_level_keys_present"] = {"expected": list(required_top_keys), "observed": list(actual_top_keys), "status": "PASS"}

    # Define the exactly eight required sub-dictionaries
    required_sub_keys: Set[str] = {
        "data_preprocessing",
        "quote_cleaning_and_synchronization",
        "return_construction_and_standardization",
        "spectral_decomposition",
        "static_evt",
        "dynamic_analysis",
        "block_maxima_inference",
        "diagnostics_and_plotting"
    }

    # Extract actual sub-keys from algorithmic_parameters
    algo_params: Dict[str, Any] = config["algorithmic_parameters"]
    if not isinstance(algo_params, dict):
        raise TypeError("algorithmic_parameters must be a dictionary.")

    actual_sub_keys: Set[str] = set(algo_params.keys())

    # Check for missing sub-keys
    missing_sub_keys: Set[str] = required_sub_keys - actual_sub_keys
    if missing_sub_keys:
        raise KeyError(f"Missing required sub-dictionaries in algorithmic_parameters: {missing_sub_keys}")

    # Log sub-level key validation success
    report["algorithmic_sub_keys_present"] = {"expected": list(required_sub_keys), "observed": list(actual_sub_keys), "status": "PASS"}

    # Validate that non-applicable components are explicitly None
    not_applicable: Dict[str, Any] = config["not_applicable_components"]
    null_fields: List[str] = ["network_architecture", "prompt_template", "llm_settings"]

    for field in null_fields:
        # Check if field exists and is exactly None
        if field not in not_applicable:
            raise KeyError(f"Missing field '{field}' in not_applicable_components.")
        if not_applicable[field] is not None:
            raise ValueError(f"Field '{field}' in not_applicable_components must be None, got {type(not_applicable[field]).__name__}")

    # Log null field validation success
    report["not_applicable_components_null"] = {"expected": "None for all", "observed": "None for all", "status": "PASS"}


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 2: Validate internal numerical consistency of every manuscript-derived constant
# -------------------------------------------------------------------------------------------------------------------------------

def validate_numerical_consistency(config: Dict[str, Any], report: Dict[str, Any]) -> None:
    """
    Validates the internal numerical consistency of manuscript-derived constants.

    Checks exact integer equalities for dimensions, validates session boundary
    time arithmetic, and confirms the exact set of excluded dates.

    Args:
        config (Dict[str, Any]): The raw study configuration dictionary.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Raises:
        ValueError: If any numerical invariant or date/time logic is violated.
    """
    # Extract data preprocessing parameters
    dp_params: Dict[str, Any] = config["algorithmic_parameters"]["data_preprocessing"]

    # Define expected exact integer values
    expected_integers: Dict[str, int] = {
        "K_assets": 479,
        "N_days_expected_after_filters": 248,
        "delta_t_seconds": 1,
        "return_intervals_per_day_expected": 21899,
        "T_total_return_intervals_expected": 5430952,
        "price_grid_timestamps_per_day_expected": 21900
    }

    # Validate exact integer matches
    for key, expected_val in expected_integers.items():
        observed_val = dp_params.get(key)
        if observed_val != expected_val:
            raise ValueError(f"Numerical inconsistency for {key}: expected {expected_val}, got {observed_val}")
        report[f"exact_match_{key}"] = {"expected": expected_val, "observed": observed_val, "status": "PASS"}

    # Validate derived identity: T = N_days * T_day
    n_days: int = dp_params["N_days_expected_after_filters"]
    t_day: int = dp_params["return_intervals_per_day_expected"]
    t_total: int = dp_params["T_total_return_intervals_expected"]

    # Equation: T = N_days * T_day
    if n_days * t_day != t_total:
        raise ValueError(f"Derived identity failed: {n_days} * {t_day} != {t_total}")
    report["derived_identity_T"] = {"expected": t_total, "observed": n_days * t_day, "status": "PASS"}

    # Validate relationship between timestamps and return intervals
    timestamps: int = dp_params["price_grid_timestamps_per_day_expected"]
    # Equation: timestamps - 1 = return intervals
    if timestamps - 1 != t_day:
        raise ValueError(f"Timestamp/interval mismatch: {timestamps} - 1 != {t_day}")
    report["timestamp_interval_relation"] = {"expected": t_day, "observed": timestamps - 1, "status": "PASS"}

    # Validate session boundaries
    start_str: str = dp_params["session_start_inclusive"]
    end_str: str = dp_params["session_end_exclusive"]

    try:
        # Parse ISO time strings
        start_time: datetime.time = datetime.time.fromisoformat(start_str)
        end_time: datetime.time = datetime.time.fromisoformat(end_str)
    except ValueError as e:
        raise ValueError(f"Failed to parse session boundaries: {e}")

    # Convert to total seconds since midnight
    start_seconds: int = start_time.hour * 3600 + start_time.minute * 60 + start_time.second
    end_seconds: int = end_time.hour * 3600 + end_time.minute * 60 + end_time.second

    # Equation: end_seconds - start_seconds = price_grid_timestamps_per_day_expected
    if end_seconds - start_seconds != timestamps:
        raise ValueError(f"Session duration mismatch: {end_seconds} - {start_seconds} != {timestamps}")
    report["session_duration_seconds"] = {"expected": timestamps, "observed": end_seconds - start_seconds, "status": "PASS"}

    # Validate excluded dates
    excluded_dates_raw: List[str] = dp_params["excluded_dates"]
    try:
        # Parse ISO date strings into a set of datetime.date objects
        observed_dates: Set[datetime.date] = {datetime.date.fromisoformat(d) for d in excluded_dates_raw}
    except ValueError as e:
        raise ValueError(f"Failed to parse excluded dates: {e}")

    # Define the exact expected set of excluded dates
    expected_dates: Set[datetime.date] = {
        datetime.date(2014, 7, 3),
        datetime.date(2014, 11, 28),
        datetime.date(2014, 12, 24)
    }

    if observed_dates != expected_dates:
        raise ValueError(f"Excluded dates mismatch. Expected {expected_dates}, got {observed_dates}")
    report["excluded_dates_match"] = {"expected": [str(d) for d in expected_dates], "observed": [str(d) for d in observed_dates], "status": "PASS"}

    # Validate overnight returns flag
    overnight_flag: bool = dp_params["overnight_returns_included"]
    if overnight_flag is not False:
        raise ValueError(f"overnight_returns_included must be False, got {overnight_flag}")
    report["overnight_returns_flag"] = {"expected": False, "observed": overnight_flag, "status": "PASS"}


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 3: Validate the asset universe and freeze the entire configuration into an immutable run manifest
# -------------------------------------------------------------------------------------------------------------------------------

def _recursive_freeze(obj: Any) -> Any:
    """
    Recursively freezes a Python object. Converts dicts to MappingProxyType
    and lists to tuples. Leaves scalars unchanged.

    Args:
        obj (Any): The object to freeze.

    Returns:
        Any: The deeply frozen object.
    """
    if isinstance(obj, dict):
        # Recursively freeze dictionary values and wrap in MappingProxyType
        return MappingProxyType({k: _recursive_freeze(v) for k, v in obj.items()})
    elif isinstance(obj, list):
        # Recursively freeze list elements and wrap in tuple
        return tuple(_recursive_freeze(v) for v in obj)
    else:
        # Return scalars (str, int, float, bool, None) as is
        return obj

def validate_universe_and_freeze(config: Dict[str, Any], report: Dict[str, Any]) -> MappingProxyType:
    """
    Validates the asset universe constraints and freezes the configuration
    into an immutable run manifest with a cryptographic hash.

    Args:
        config (Dict[str, Any]): The raw study configuration dictionary.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Returns:
        MappingProxyType: The deeply frozen, immutable run manifest.

    Raises:
        ValueError: If the asset universe violates length, uniqueness, boundary, or sorting constraints.
    """
    # Extract the ticker list
    tickers: List[str] = config["algorithmic_parameters"]["data_preprocessing"]["asset_universe_exact_appendix_order"]

    # Validate length
    if len(tickers) != 479:
        raise ValueError(f"Asset universe must contain exactly 479 tickers, got {len(tickers)}")
    report["universe_length"] = {"expected": 479, "observed": len(tickers), "status": "PASS"}

    # Validate uniqueness
    if len(set(tickers)) != 479:
        raise ValueError("Asset universe contains duplicate tickers.")
    report["universe_uniqueness"] = {"expected": 479, "observed": len(set(tickers)), "status": "PASS"}

    # Validate boundaries
    if tickers[0] != "A" or tickers[-1] != "YUM":
        raise ValueError(f"Asset universe boundaries invalid. Expected 'A' and 'YUM', got '{tickers[0]}' and '{tickers[-1]}'")
    report["universe_boundaries"] = {"expected": ["A", "YUM"], "observed": [tickers[0], tickers[-1]], "status": "PASS"}

    # Validate alphabetical sorting
    if tickers != sorted(tickers):
        # Find the first out-of-order element for the error message
        for i in range(len(tickers) - 1):
            if tickers[i] > tickers[i+1]:
                raise ValueError(f"Asset universe is not alphabetically sorted. Out of order at index {i}: '{tickers[i]}' > '{tickers[i+1]}'")
    report["universe_sorted"] = {"expected": True, "observed": True, "status": "PASS"}

    # Create a deep copy to prevent mutating the original input during hashing preparation
    config_copy: Dict[str, Any] = copy.deepcopy(config)

    # Generate canonical JSON string for hashing
    # default=str handles non-serializable objects like datetime if any slipped in
    canonical_json: str = json.dumps(config_copy, sort_keys=True, default=str)

    # Compute SHA-256 hash
    manifest_hash: str = hashlib.sha256(canonical_json.encode('utf-8')).hexdigest()

    # Inject the hash into the configuration copy
    config_copy["manifest_hash"] = manifest_hash
    report["manifest_hash_generated"] = {"expected": "SHA-256 hash", "observed": manifest_hash, "status": "PASS"}

    # Recursively freeze the configuration
    frozen_manifest: MappingProxyType = _recursive_freeze(config_copy)

    return frozen_manifest


# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def validate_study_configuration(study_configuration: Dict[str, Any]) -> Tuple[MappingProxyType, Dict[str, Any]]:
    """
    Orchestrates the validation of the ingested study configuration object.

    Executes structural completeness checks, numerical consistency checks,
    and asset universe validation. If all checks pass, it freezes the
    configuration into an immutable manifest.

    Args:
        study_configuration (Dict[str, Any]): The raw study configuration dictionary.

    Returns:
        Tuple[MappingProxyType, Dict[str, Any]]: A tuple containing:
            - The frozen, immutable run manifest.
            - The comprehensive validation report dictionary.

    Raises:
        Exception: Bubbles up any KeyError, ValueError, or TypeError encountered
                   during the strict validation steps, halting the pipeline.
    """
    # Initialize the validation report
    validation_report: Dict[str, Any] = {}

    # Step 1: Validate structural completeness
    validate_structural_completeness(study_configuration, validation_report)

    # Step 2: Validate numerical consistency
    validate_numerical_consistency(study_configuration, validation_report)

    # Step 3: Validate universe and freeze manifest
    frozen_manifest = validate_universe_and_freeze(study_configuration, validation_report)

    return frozen_manifest, validation_report
````

````python
# Task 2: Validate the ingested raw data schema and DataFrame structure

# ==============================================================================
# Task 2: Validate the ingested raw data schema and DataFrame structure
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 1: Confirm column presence and dtype compatibility
# -------------------------------------------------------------------------------------------------------------------------------

def validate_columns_and_dtypes(df: pd.DataFrame, manifest: MappingProxyType, report: Dict[str, Any]) -> None:
    """
    Validates that the raw TAQ DataFrame contains all required columns and that
    their data types are strictly compatible with the expected schema.

    Args:
        df (pd.DataFrame): The raw TAQ DataFrame.
        manifest (MappingProxyType): The frozen run manifest containing the schema.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Raises:
        ValueError: If required columns are missing or if critical dtype
                    incompatibilities (e.g., float32 instead of float64, or
                    float64 for integer columns implying NaNs) are detected.
    """
    # Extract expected schema from manifest
    expected_schema: MappingProxyType = manifest["raw_unprocessed_data_schema"]["expected_dtypes"]
    required_columns: Set[str] = set(expected_schema.keys())
    actual_columns: Set[str] = set(df.columns)

    # Check for missing columns
    missing_columns: Set[str] = required_columns - actual_columns
    if missing_columns:
        raise ValueError(f"Raw DataFrame is missing required columns: {missing_columns}")

    # Log extra columns (informational)
    extra_columns: Set[str] = actual_columns - required_columns
    report["extra_columns_present"] = list(extra_columns)

    # Validate dtypes
    for col, expected_dtype_str in expected_schema.items():
        actual_dtype = df[col].dtype
        status = "PASS"
        notes = ""

        if expected_dtype_str == "datetime64[ns]":
            if not pd.api.types.is_datetime64_any_dtype(actual_dtype):
                raise ValueError(f"Column '{col}' must be datetime64[ns], got {actual_dtype}")

        elif expected_dtype_str == "timedelta64[ns]":
            if not pd.api.types.is_timedelta64_dtype(actual_dtype):
                raise ValueError(f"Column '{col}' must be timedelta64[ns], got {actual_dtype}")

        elif expected_dtype_str == "category":
            if not pd.api.types.is_categorical_dtype(actual_dtype):
                if pd.api.types.is_object_dtype(actual_dtype) or pd.api.types.is_string_dtype(actual_dtype):
                    status = "WARNING"
                    notes = "Object/string dtype detected. Conversion to category recommended for memory efficiency."
                else:
                    raise ValueError(f"Column '{col}' must be category or object/string, got {actual_dtype}")

        elif expected_dtype_str == "float64":
            if not pd.api.types.is_float_dtype(actual_dtype):
                raise ValueError(f"Column '{col}' must be float64, got {actual_dtype}")
            if actual_dtype == np.float32:
                raise ValueError(f"Column '{col}' is float32. Strict float64 precision is required to prevent downstream numerical drift.")

        elif expected_dtype_str == "int64":
            if pd.api.types.is_float_dtype(actual_dtype):
                raise ValueError(f"Column '{col}' is expected to be int64 but is {actual_dtype}. This usually indicates the presence of NaN values in an integer column, which is invalid for this field.")
            elif not pd.api.types.is_integer_dtype(actual_dtype):
                raise ValueError(f"Column '{col}' must be int64, got {actual_dtype}")
            elif str(actual_dtype) == "Int64":
                status = "WARNING"
                notes = "Nullable Int64 detected. Standard int64 is preferred if no missing values exist."

        report[f"dtype_check_{col}"] = {
            "expected": expected_dtype_str,
            "observed": str(actual_dtype),
            "status": status,
            "notes": notes
        }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 2: Confirm RAW_ROW_ORDER properties, non-emptiness, and date coverage
# -------------------------------------------------------------------------------------------------------------------------------

def validate_row_order_and_dates(df: pd.DataFrame, report: Dict[str, Any]) -> None:
    """
    Validates that the DataFrame is not empty, that RAW_ROW_ORDER is globally
    unique (and checks monotonicity), and that the DATE column spans the target year.

    Args:
        df (pd.DataFrame): The raw TAQ DataFrame.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Raises:
        ValueError: If the DataFrame is empty, if RAW_ROW_ORDER is not unique,
                    or if no dates fall within 2014.
    """
    # Check non-emptiness
    row_count: int = len(df)
    if row_count == 0:
        raise ValueError("The raw DataFrame is empty.")
    report["total_row_count"] = row_count

    # Validate RAW_ROW_ORDER uniqueness
    # This is critical for deterministic tie-breaking within the same second
    is_unique: bool = df["RAW_ROW_ORDER"].is_unique
    if not is_unique:
        duplicate_count = row_count - df["RAW_ROW_ORDER"].nunique()
        raise ValueError(f"RAW_ROW_ORDER is not globally unique. Found {duplicate_count} duplicate values. Deterministic tie-breaking is impossible.")
    report["raw_row_order_unique"] = {"status": "PASS"}

    # Check monotonicity (informational, as Task 3 will sort anyway)
    is_monotonic: bool = df["RAW_ROW_ORDER"].is_monotonic_increasing
    report["raw_row_order_monotonic"] = {
        "status": "PASS" if is_monotonic else "WARNING",
        "notes": "Data is not sorted by RAW_ROW_ORDER. Task 3 sort will correct this." if not is_monotonic else ""
    }

    # Validate date coverage
    min_date: pd.Timestamp = df["DATE"].min()
    max_date: pd.Timestamp = df["DATE"].max()

    target_start = pd.Timestamp("2014-01-01")
    target_end = pd.Timestamp("2014-12-31")

    # Check if ANY dates fall in 2014
    if max_date < target_start or min_date > target_end:
        raise ValueError(f"No dates found in the target year 2014. Date range in data: {min_date.date()} to {max_date.date()}")

    status = "PASS"
    notes = ""
    if min_date < target_start or max_date > target_end:
        status = "WARNING"
        notes = "Data contains dates outside 2014. These will be filtered out in Task 4."

    report["date_coverage"] = {
        "min_date": str(min_date.date()),
        "max_date": str(max_date.date()),
        "status": status,
        "notes": notes
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 3: Confirm ticker universe coverage
# -------------------------------------------------------------------------------------------------------------------------------

def validate_ticker_universe(df: pd.DataFrame, manifest: MappingProxyType, report: Dict[str, Any]) -> None:
    """
    Validates that all 479 required tickers specified in the manifest are
    present in the raw DataFrame.

    Args:
        df (pd.DataFrame): The raw TAQ DataFrame.
        manifest (MappingProxyType): The frozen run manifest.
        report (Dict[str, Any]): The validation report dictionary to accumulate results.

    Raises:
        ValueError: If any of the required 479 tickers are entirely missing from the data.
    """
    # Extract required tickers from manifest
    required_tickers_list: tuple = manifest["algorithmic_parameters"]["data_preprocessing"]["asset_universe_exact_appendix_order"]
    required_tickers: Set[str] = set(required_tickers_list)

    # Extract unique tickers from data, stripping whitespace and converting to uppercase for robust comparison
    # We use drop_duplicates on the series first to minimize string operations
    raw_unique_series = pd.Series(df["SYM_ROOT"].unique())
    if pd.api.types.is_object_dtype(raw_unique_series) or pd.api.types.is_string_dtype(raw_unique_series):
        normalized_raw_tickers: Set[str] = set(raw_unique_series.str.strip().str.upper())
    else:
        # If it's already a clean category, just take the values
        normalized_raw_tickers: Set[str] = set(raw_unique_series)

    # Compute intersection and differences
    present_tickers: Set[str] = required_tickers & normalized_raw_tickers
    missing_tickers: Set[str] = required_tickers - normalized_raw_tickers
    extra_tickers: Set[str] = normalized_raw_tickers - required_tickers

    # Hard halt if any required ticker is missing
    if missing_tickers:
        raise ValueError(f"Critical Data Failure: {len(missing_tickers)} required tickers are missing from the raw data: {sorted(list(missing_tickers))}")

    report["ticker_universe_coverage"] = {
        "required_count": len(required_tickers),
        "present_count": len(present_tickers),
        "extra_count": len(extra_tickers),
        "status": "PASS"
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def validate_raw_dataframe(df_raw_taq: pd.DataFrame, manifest: MappingProxyType) -> Dict[str, Any]:
    """
    Orchestrates the validation of the raw TAQ DataFrame against the study schema.

    Executes column/dtype checks, row order and date range validation, and
    ensures the complete 479-ticker universe is present.

    Args:
        df_raw_taq (pd.DataFrame): The raw, unprocessed TAQ DataFrame.
        manifest (MappingProxyType): The frozen, immutable run manifest.

    Returns:
        Dict[str, Any]: A comprehensive schema validation report.

    Raises:
        Exception: Bubbles up any ValueError or TypeError encountered during
                   validation, halting the pipeline immediately.
    """
    schema_report: Dict[str, Any] = {}

    # Step 1: Validate columns and dtypes
    validate_columns_and_dtypes(df_raw_taq, manifest, schema_report)

    # Step 2: Validate row order, non-emptiness, and dates
    validate_row_order_and_dates(df_raw_taq, schema_report)

    # Step 3: Validate ticker universe coverage
    validate_ticker_universe(df_raw_taq, manifest, schema_report)

    return schema_report
````

````python
# Task 3: Cleanse the raw quote data using microstructure filters

# ==============================================================================
# Task 3: Cleanse the raw quote data using microstructure filters
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 1: Sort the raw DataFrame deterministically
# -------------------------------------------------------------------------------------------------------------------------------

def sort_raw_dataframe_deterministically(df: pd.DataFrame) -> pd.DataFrame:
    """
    Sorts the raw TAQ DataFrame deterministically to ensure reproducible
    downstream selection of the last valid quote per second.

    The sort keys are strictly: DATE, SYM_ROOT, TIME_M, RAW_ROW_ORDER.
    A stable sort algorithm (mergesort) is used.

    Args:
        df (pd.DataFrame): The validated raw TAQ DataFrame.

    Returns:
        pd.DataFrame: A new DataFrame sorted by the required keys with a reset index.
    """
    sort_keys: List[str] = ["DATE", "SYM_ROOT", "TIME_M", "RAW_ROW_ORDER"]

    # Perform a stable sort and reset the index to a clean RangeIndex
    # We use kind="mergesort" to guarantee stability.
    df_sorted: pd.DataFrame = df.sort_values(
        by=sort_keys,
        ascending=True,
        kind="mergesort",
        ignore_index=True
    )

    # Defensive check: Verify monotonicity of RAW_ROW_ORDER within a sample group
    # We check the first available group to ensure the sort was effective.
    if not df_sorted.empty:
        first_date = df_sorted["DATE"].iloc[0]
        first_sym = df_sorted["SYM_ROOT"].iloc[0]
        sample_group = df_sorted[
            (df_sorted["DATE"] == first_date) & (df_sorted["SYM_ROOT"] == first_sym)
        ]
        if not sample_group["RAW_ROW_ORDER"].is_monotonic_increasing:
            raise RuntimeError("Deterministic sort failed: RAW_ROW_ORDER is not monotonically increasing within groups.")

    return df_sorted

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 2: Apply quote-admissibility filters and record removal counts
# -------------------------------------------------------------------------------------------------------------------------------

def apply_microstructure_filters(df_sorted: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, Dict[str, Any]]:
    """
    Applies a sequence of microstructure filters to retain only firm, actionable quotes.
    Tracks the sequential removal counts for auditing purposes.

    Args:
        df_sorted (pd.DataFrame): The deterministically sorted TAQ DataFrame.
        manifest (MappingProxyType): The frozen run manifest containing filter configs.

    Returns:
        Tuple[pd.DataFrame, Dict[str, Any]]:
            - The cleaned DataFrame (a copy, to prevent SettingWithCopy warnings).
            - A dictionary containing the sequential filtering audit.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["quote_cleaning_and_synchronization"]

    keep_codes: tuple = config.get("quote_condition_keep_codes", ("R",))
    req_bid_gt_zero: bool = config.get("require_bid_gt_zero", True)
    req_ask_gt_zero: bool = config.get("require_ask_gt_zero", True)
    req_bidsiz_gt_zero: bool = config.get("require_bidsiz_gt_zero", True)
    req_asksiz_gt_zero: bool = config.get("require_asksiz_gt_zero", True)
    req_ask_gt_bid: bool = config.get("require_ask_gt_bid", True)

    total_initial_rows: int = len(df_sorted)

    # Initialize a running boolean mask (all True)
    running_mask: np.ndarray = np.ones(total_initial_rows, dtype=bool)
    current_surviving_count: int = total_initial_rows

    audit_log: List[Dict[str, Any]] = []

    def _apply_and_log(condition_mask: pd.Series, filter_name: str, provenance: str) -> None:
        nonlocal running_mask, current_surviving_count
        # Calculate the new mask: must have survived previous filters AND pass current filter
        new_mask: np.ndarray = running_mask & condition_mask.to_numpy()
        new_surviving_count: int = int(new_mask.sum())
        removed_in_this_step: int = current_surviving_count - new_surviving_count

        audit_log.append({
            "filter_name": filter_name,
            "provenance": provenance,
            "rows_before": current_surviving_count,
            "rows_after": new_surviving_count,
            "rows_removed": removed_in_this_step,
            "removal_percentage": (removed_in_this_step / total_initial_rows) * 100 if total_initial_rows > 0 else 0.0
        })

        running_mask = new_mask
        current_surviving_count = new_surviving_count

    # Filter 1: Quote condition
    cond_mask = df_sorted["QU_COND"].isin(keep_codes)
    _apply_and_log(cond_mask, "QU_COND in keep_codes", "replication_convention")

    # Filter 2: Positive bid
    if req_bid_gt_zero:
        cond_mask = df_sorted["BID"] > 0
        _apply_and_log(cond_mask, "BID > 0", "replication_convention")

    # Filter 3: Positive ask
    if req_ask_gt_zero:
        cond_mask = df_sorted["ASK"] > 0
        _apply_and_log(cond_mask, "ASK > 0", "replication_convention")

    # Filter 4: Positive bid size
    if req_bidsiz_gt_zero:
        cond_mask = df_sorted["BIDSIZ"] > 0
        _apply_and_log(cond_mask, "BIDSIZ > 0", "replication_convention")

    # Filter 5: Positive ask size
    if req_asksiz_gt_zero:
        cond_mask = df_sorted["ASKSIZ"] > 0
        _apply_and_log(cond_mask, "ASKSIZ > 0", "replication_convention")

    # Filter 6: Ask strictly greater than Bid (no locked/crossed markets)
    if req_ask_gt_bid:
        cond_mask = df_sorted["ASK"] > df_sorted["BID"]
        _apply_and_log(cond_mask, "ASK > BID", "replication_convention")

    # Extract unique QU_COND values and their frequencies for the audit
    qu_cond_counts = df_sorted["QU_COND"].value_counts().to_dict()

    # Create the final cleaned DataFrame
    df_cleaned: pd.DataFrame = df_sorted.loc[running_mask].copy()

    audit_summary: Dict[str, Any] = {
        "sequential_filters": audit_log,
        "total_initial_rows": total_initial_rows,
        "total_surviving_rows": current_surviving_count,
        "overall_survival_fraction": current_surviving_count / total_initial_rows if total_initial_rows > 0 else 0.0,
        "raw_qu_cond_frequencies": qu_cond_counts
    }

    return df_cleaned, audit_summary

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 3: Produce the cleaning audit and validate minimum data coverage per stock
# -------------------------------------------------------------------------------------------------------------------------------

def validate_data_coverage(df_cleaned: pd.DataFrame, manifest: MappingProxyType, audit_summary: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validates that every required stock has sufficient trading days with valid quotes.
    Appends the coverage statistics to the audit summary.

    Args:
        df_cleaned (pd.DataFrame): The cleaned TAQ DataFrame.
        manifest (MappingProxyType): The frozen run manifest.
        audit_summary (Dict[str, Any]): The audit dictionary from Step 2.

    Returns:
        Dict[str, Any]: The finalized cleaning audit dictionary.

    Raises:
        ValueError: If any required stock fails the minimum day coverage threshold.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["quote_cleaning_and_synchronization"]
    required_tickers: tuple = manifest["algorithmic_parameters"]["data_preprocessing"]["asset_universe_exact_appendix_order"]
    expected_days: int = manifest["algorithmic_parameters"]["data_preprocessing"]["N_days_expected_after_filters"]

    # Extract max allowable missing fraction, default to 0.05 if absent
    max_missing_frac: float = config.get("max_allowable_zero_quote_day_fraction", 0.05)

    # Efficiently count unique dates per ticker
    # Drop duplicates on DATE and SYM_ROOT to get unique days per stock
    unique_days_df = df_cleaned[["DATE", "SYM_ROOT"]].drop_duplicates()
    days_per_ticker_series = unique_days_df.groupby("SYM_ROOT", observed=False).size()

    coverage_stats: Dict[str, Dict[str, Any]] = {}
    failed_tickers: List[str] = []

    for ticker in required_tickers:
        # Get count, defaulting to 0 if ticker was completely filtered out
        days_with_quotes: int = int(days_per_ticker_series.get(ticker, 0))
        missing_fraction: float = 1.0 - (days_with_quotes / expected_days)

        coverage_stats[ticker] = {
            "days_with_quotes": days_with_quotes,
            "missing_fraction": missing_fraction,
            "passed": missing_fraction <= max_missing_frac
        }

        if missing_fraction > max_missing_frac:
            failed_tickers.append(f"{ticker} (Missing: {missing_fraction:.2%})")

    audit_summary["stock_coverage_stats"] = coverage_stats
    audit_summary["max_allowable_zero_quote_day_fraction"] = max_missing_frac

    if failed_tickers:
        raise ValueError(f"Structural Data Failure: The following stocks exceeded the maximum allowable fraction ({max_missing_frac:.2%}) of days without valid quotes: {failed_tickers}")

    return audit_summary

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def cleanse_raw_quotes(df_raw_taq: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, Dict[str, Any]]:
    """
    Orchestrates the microstructure cleansing of the raw TAQ DataFrame.

    Executes deterministic sorting, applies sequential quote-admissibility filters,
    and validates that all required stocks maintain sufficient data coverage.

    Args:
        df_raw_taq (pd.DataFrame): The validated raw TAQ DataFrame.
        manifest (MappingProxyType): The frozen, immutable run manifest.

    Returns:
        Tuple[pd.DataFrame, Dict[str, Any]]: A tuple containing:
            - The cleaned, sorted DataFrame.
            - The comprehensive cleaning audit dictionary.

    Raises:
        Exception: Bubbles up any ValueError or RuntimeError encountered during
                   sorting, filtering, or coverage validation.
    """
    # Step 1: Sort deterministically
    df_sorted: pd.DataFrame = sort_raw_dataframe_deterministically(df_raw_taq)

    # Step 2: Apply filters and get sequential audit
    df_cleaned, partial_audit = apply_microstructure_filters(df_sorted, manifest)

    # Explicitly delete the intermediate sorted DataFrame to free memory
    del df_sorted
    gc.collect()

    # Step 3: Validate coverage and finalize audit
    final_audit: Dict[str, Any] = validate_data_coverage(df_cleaned, manifest, partial_audit)

    return df_cleaned, final_audit
````

````python
# Task 4: Filter to the retained sample and remove excluded dates

# ==============================================================================
# Task 4: Filter to the retained sample and remove excluded dates
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 1: Filter to the exact 479-ticker appendix universe
# -------------------------------------------------------------------------------------------------------------------------------

def filter_to_exact_universe(df: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, Dict[str, Any]]:
    """
    Filters the cleaned DataFrame to retain only the exact 479 tickers specified
    in the study universe. Converts the SYM_ROOT column to an ordered Categorical
    type to enforce canonical matrix row ordering downstream.

    Args:
        df (pd.DataFrame): The cleaned TAQ DataFrame.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, Dict[str, Any]]:
            - The universe-filtered DataFrame.
            - A dictionary containing filtering statistics.

    Raises:
        ValueError: If any of the 479 required tickers are completely missing after filtering.
    """
    # Extract the exact ordered ticker list from the manifest
    ticker_list: tuple = manifest["algorithmic_parameters"]["data_preprocessing"]["asset_universe_exact_appendix_order"]
    ticker_set: Set[str] = set(ticker_list)

    initial_rows: int = len(df)

    # Create boolean mask for required tickers
    universe_mask: pd.Series = df["SYM_ROOT"].isin(ticker_set)

    # Filter the DataFrame (creates a copy)
    df_universe: pd.DataFrame = df.loc[universe_mask].copy()

    final_rows: int = len(df_universe)
    dropped_rows: int = initial_rows - final_rows

    # Convert SYM_ROOT to an ordered Categorical type
    # This is critical: it ensures that groupby operations and pivot tables
    # will always output rows in the exact appendix order (A to YUM).
    cat_dtype = pd.CategoricalDtype(categories=list(ticker_list), ordered=True)
    df_universe["SYM_ROOT"] = df_universe["SYM_ROOT"].astype(cat_dtype)

    # Defensive check: ensure all 479 categories have at least one row
    # observed=False forces pandas to show counts for all categories, even empty ones
    category_counts = df_universe.groupby("SYM_ROOT", observed=False).size()
    empty_categories = category_counts[category_counts == 0].index.tolist()

    if empty_categories:
        raise ValueError(f"Critical Error: The following required tickers have 0 rows after universe filtering: {empty_categories}")

    stats: Dict[str, Any] = {
        "initial_rows": initial_rows,
        "final_rows": final_rows,
        "dropped_rows": dropped_rows,
        "retained_tickers": 479
    }

    return df_universe, stats

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 2: Remove excluded dates and validate N_days = 248
# -------------------------------------------------------------------------------------------------------------------------------

def filter_excluded_dates(df: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, List[pd.Timestamp], Dict[str, Any]]:
    """
    Removes explicitly excluded early-close dates from the DataFrame and validates
    that exactly 248 trading days remain. Extracts the sorted list of retained dates.

    Args:
        df (pd.DataFrame): The universe-filtered DataFrame.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, List[pd.Timestamp], Dict[str, Any]]:
            - The date-filtered DataFrame.
            - A sorted list of the 248 retained pd.Timestamp dates.
            - A dictionary containing filtering statistics.

    Raises:
        ValueError: If the number of unique retained dates does not equal exactly 248.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    excluded_dates_str: tuple = config["excluded_dates"]
    expected_n_days: int = config["N_days_expected_after_filters"]

    # Parse excluded dates into pd.Timestamp to match the DATE column dtype
    excluded_timestamps: List[pd.Timestamp] = [pd.Timestamp(d) for d in excluded_dates_str]

    initial_rows: int = len(df)

    # Create boolean mask to drop excluded dates
    valid_date_mask: pd.Series = ~df["DATE"].isin(excluded_timestamps)

    # Filter the DataFrame
    df_dates: pd.DataFrame = df.loc[valid_date_mask].copy()

    final_rows: int = len(df_dates)
    dropped_rows: int = initial_rows - final_rows

    # Extract unique dates and sort them
    unique_dates_array = df_dates["DATE"].unique()
    sorted_dates: List[pd.Timestamp] = sorted([pd.Timestamp(d) for d in unique_dates_array])
    n_days_observed: int = len(sorted_dates)

    # Strict invariant validation
    if n_days_observed != expected_n_days:
        raise ValueError(f"Invariant Violation: Expected exactly {expected_n_days} trading days, but found {n_days_observed} after filtering. Observed dates: {[d.date() for d in sorted_dates]}")

    stats: Dict[str, Any] = {
        "initial_rows": initial_rows,
        "final_rows": final_rows,
        "dropped_rows": dropped_rows,
        "n_days_retained": n_days_observed,
        "excluded_dates_applied": [str(d.date()) for d in excluded_timestamps]
    }

    return df_dates, sorted_dates, stats

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 3: Filter to the retained intraday session window and separate pre-session quotes
# -------------------------------------------------------------------------------------------------------------------------------

def partition_intraday_session(df: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, pd.DataFrame, Dict[str, Any]]:
    """
    Partitions the DataFrame into pre-session, in-session, and post-session subsets
    based on TIME_M. Discards the post-session data.

    Args:
        df (pd.DataFrame): The date-filtered DataFrame.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame, Dict[str, Any]]:
            - df_session: The analytical sample (09:40:00 <= TIME_M < 15:45:00).
            - df_pre_session: Quotes before 09:40:00, used only for initialization.
            - A dictionary containing partitioning statistics.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]

    # Parse session boundaries into pd.Timedelta
    start_td = pd.Timedelta(config["session_start_inclusive"])
    end_td = pd.Timedelta(config["session_end_exclusive"])

    total_rows: int = len(df)

    # Create boolean masks for the three disjoint subsets
    pre_mask: pd.Series = df["TIME_M"] < start_td
    session_mask: pd.Series = (df["TIME_M"] >= start_td) & (df["TIME_M"] < end_td)
    post_mask: pd.Series = df["TIME_M"] >= end_td

    # Extract the subsets as independent copies
    df_pre_session: pd.DataFrame = df.loc[pre_mask].copy()
    df_session: pd.DataFrame = df.loc[session_mask].copy()

    # Calculate counts
    pre_count: int = len(df_pre_session)
    session_count: int = len(df_session)
    post_count: int = int(post_mask.sum())

    # Validate partition sum
    if pre_count + session_count + post_count != total_rows:
        raise RuntimeError(f"Partitioning error: {pre_count} + {session_count} + {post_count} != {total_rows}")

    stats: Dict[str, Any] = {
        "total_input_rows": total_rows,
        "pre_session_rows": pre_count,
        "in_session_rows": session_count,
        "post_session_discarded_rows": post_count,
        "session_start": str(start_td),
        "session_end": str(end_td)
    }

    return df_session, df_pre_session, stats

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def filter_sample_dates_and_universe(df_cleaned: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, pd.DataFrame, List[pd.Timestamp]]:
    """
    Orchestrates the filtering of the cleaned TAQ data to the exact study sample.

    Executes universe filtering (479 tickers), date filtering (removing early closes,
    validating 248 days), and session partitioning (separating pre-session and
    in-session data, discarding post-session).

    Args:
        df_cleaned (pd.DataFrame): The cleaned TAQ DataFrame from Task 3.
        manifest (MappingProxyType): The frozen, immutable run manifest.

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame, List[pd.Timestamp]]: A tuple containing:
            - df_session: The analytical in-session DataFrame.
            - df_pre_session: The pre-session DataFrame for initialization.
            - date_list: A sorted list of the 248 retained pd.Timestamp dates.

    Raises:
        Exception: Bubbles up any ValueError or RuntimeError encountered during
                   filtering or invariant validation.
    """
    # Step 1: Filter to exact 479-ticker universe
    df_universe, uni_stats = filter_to_exact_universe(df_cleaned, manifest)

    # Step 2: Remove excluded dates and validate N_days = 248
    df_dates, date_list, date_stats = filter_excluded_dates(df_universe, manifest)

    # Explicitly delete intermediate DataFrame to free memory
    del df_universe
    gc.collect()

    # Step 3: Partition into pre-session and in-session
    df_session, df_pre_session, part_stats = partition_intraday_session(df_dates, manifest)

    # Explicitly delete intermediate DataFrame to free memory
    del df_dates
    gc.collect()

    # Note: The statistics dictionaries (uni_stats, date_stats, part_stats) are
    # computed and available for logging, but the orchestrator signature requires
    # returning only the three core artifacts. In a full production system, these
    # stats would be written to a central logger.

    return df_session, df_pre_session, date_list
````

````python
# Task 5: Assign second buckets and select the last valid quote per second

# ==============================================================================
# Task 5: Assign second buckets and select the last valid quote per second
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 1: Compute the second-bucket label by downward truncation
# -------------------------------------------------------------------------------------------------------------------------------

def assign_second_buckets(df: pd.DataFrame) -> pd.DataFrame:
    """
    Computes the second-bucket label for each quote by strictly truncating
    the TIME_M column downward to the containing whole second.

    Uses exact integer arithmetic on the underlying nanosecond representation
    to avoid floating-point or version-dependent rounding issues.

    Args:
        df (pd.DataFrame): The input TAQ DataFrame (session or pre-session).

    Returns:
        pd.DataFrame: The DataFrame augmented with a 'SECOND_BUCKET' column.

    Raises:
        ValueError: If TIME_M contains NaT values.
    """
    if df.empty:
        # Return empty DataFrame with the new column if input is empty
        df_out = df.copy()
        df_out["SECOND_BUCKET"] = pd.Series(dtype='timedelta64[ns]')
        return df_out

    # Defensive check: ensure no NaT values which would corrupt integer conversion
    if df["TIME_M"].isna().any():
        raise ValueError("TIME_M column contains NaT values. Cannot compute second buckets.")

    df_out = df.copy()

    # Extract underlying int64 nanoseconds
    ns_array: np.ndarray = df_out["TIME_M"].values.astype("int64")

    # Perform downward truncation to the nearest second (1,000,000,000 ns)
    floor_ns_array: np.ndarray = (ns_array // 1_000_000_000) * 1_000_000_000

    # Convert back to timedelta64[ns]
    df_out["SECOND_BUCKET"] = pd.to_timedelta(floor_ns_array, unit="ns")

    return df_out

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 2: Select the last valid quote within each group
# -------------------------------------------------------------------------------------------------------------------------------

def select_last_quote_per_second(df: pd.DataFrame) -> pd.DataFrame:
    """
    Selects the last valid quote within each (DATE, SYM_ROOT, SECOND_BUCKET) group.

    Relies on the deterministic sort order established in Task 3
    (DATE, SYM_ROOT, TIME_M, RAW_ROW_ORDER ascending). The 'last' row in each
    group is guaranteed to be the one with the latest TIME_M and highest RAW_ROW_ORDER.

    Args:
        df (pd.DataFrame): The DataFrame with SECOND_BUCKET assigned.

    Returns:
        pd.DataFrame: A deduplicated DataFrame with at most one row per second per stock.
    """
    if df.empty:
        return df.copy()

    initial_rows: int = len(df)

    # Use drop_duplicates with keep="last" for highly optimized group-tail selection
    df_selected: pd.DataFrame = df.drop_duplicates(
        subset=["DATE", "SYM_ROOT", "SECOND_BUCKET"],
        keep="last"
    ).copy()

    final_rows: int = len(df_selected)

    # Optional: Log the compression ratio (quotes per second)
    # compression_ratio = initial_rows / final_rows if final_rows > 0 else 0

    return df_selected

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 3: Compute midpoint prices for each selected quote
# -------------------------------------------------------------------------------------------------------------------------------

def compute_sparse_midpoints(df_selected: pd.DataFrame) -> pd.DataFrame:
    """
    Computes the midpoint price for each selected quote and projects the DataFrame
    down to only the required columns to conserve memory.

    Equation: m_k(t) = (a_k(t) + b_k(t)) / 2

    Args:
        df_selected (pd.DataFrame): The deduplicated DataFrame.

    Returns:
        pd.DataFrame: A sparse table containing DATE, SYM_ROOT, SECOND_BUCKET, and MIDPOINT.

    Raises:
        ValueError: If any computed midpoint is zero or negative.
    """
    if df_selected.empty:
        return pd.DataFrame(columns=["DATE", "SYM_ROOT", "SECOND_BUCKET", "MIDPOINT"])

    # Vectorized midpoint computation
    midpoints: np.ndarray = (df_selected["ASK"].values + df_selected["BID"].values) / 2.0

    # Strict validation: midpoints must be strictly positive
    if not (midpoints > 0).all():
        raise ValueError("Critical Error: Computed midpoints contain zero or negative values.")

    # Assign to DataFrame
    df_selected["MIDPOINT"] = midpoints

    # Project down to required columns to free memory
    required_columns: List[str] = ["DATE", "SYM_ROOT", "SECOND_BUCKET", "MIDPOINT"]
    df_sparse: pd.DataFrame = df_selected[required_columns].copy()

    # Reset index for clean downstream processing
    df_sparse.reset_index(drop=True, inplace=True)

    return df_sparse

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def assign_second_buckets_and_select(df_session: pd.DataFrame, df_pre_session: pd.DataFrame, manifest: MappingProxyType) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Orchestrates the assignment of second buckets, selection of the last valid quote,
    and computation of midpoints for both the analytical session and pre-session data.

    Args:
        df_session (pd.DataFrame): The analytical in-session DataFrame.
        df_pre_session (pd.DataFrame): The pre-session DataFrame for initialization.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame]: A tuple containing:
            - df_sparse_midpoints: The sparse in-session midpoint table.
            - df_pre_session_midpoints: The sparse pre-session midpoint table.
    """
    # Process the analytical session data
    df_session_bucketed = assign_second_buckets(df_session)
    df_session_selected = select_last_quote_per_second(df_session_bucketed)
    df_sparse_midpoints = compute_sparse_midpoints(df_session_selected)

    # Explicitly free intermediate session DataFrames
    del df_session_bucketed
    del df_session_selected
    gc.collect()

    # Process the pre-session data
    df_pre_bucketed = assign_second_buckets(df_pre_session)
    df_pre_selected = select_last_quote_per_second(df_pre_bucketed)
    df_pre_session_midpoints = compute_sparse_midpoints(df_pre_selected)

    # Explicitly free intermediate pre-session DataFrames
    del df_pre_bucketed
    del df_pre_selected
    gc.collect()

    # Validate session boundaries for the analytical sample
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    start_td = pd.Timedelta(config["session_start_inclusive"])
    end_td = pd.Timedelta(config["session_end_exclusive"])

    if not df_sparse_midpoints.empty:
        min_bucket = df_sparse_midpoints["SECOND_BUCKET"].min()
        max_bucket = df_sparse_midpoints["SECOND_BUCKET"].max()
        if min_bucket < start_td or max_bucket >= end_td:
            raise ValueError(f"Session boundary violation in sparse midpoints. Min: {min_bucket}, Max: {max_bucket}")

    return df_sparse_midpoints, df_pre_session_midpoints
````

````python
# Task 6: Build the complete synchronized midpoint panel with forward fill

# ==============================================================================
# Task 6: Build the complete synchronized midpoint panel with forward fill
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 1: Construct the target full second grid and pre-allocate memmap
# -------------------------------------------------------------------------------------------------------------------------------

def pre_allocate_midpoint_panel(manifest: MappingProxyType) -> Tuple[str, np.memmap, Tuple[int, int, int]]:
    """
    Calculates the exact dimensions of the synchronized midpoint panel and
    pre-allocates a memory-mapped numpy array on disk initialized to NaN.

    Args:
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, np.memmap, Tuple[int, int, int]]:
            - The absolute file path to the memory-mapped file.
            - The numpy memmap object (writeable).
            - The 3D shape tuple: (K_assets, N_days, T_day).
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]

    k_assets: int = config["K_assets"]
    n_days: int = config["N_days_expected_after_filters"]
    t_day: int = config["price_grid_timestamps_per_day_expected"]

    # Verify the exact dimensions
    if k_assets != 479 or n_days != 248 or t_day != 21900:
        raise ValueError(f"Dimension mismatch in panel allocation. Expected (479, 248, 21900), got ({k_assets}, {n_days}, {t_day})")

    shape_3d: Tuple[int, int, int] = (k_assets, n_days, t_day)

    # Generate a deterministic filename based on the manifest hash
    manifest_hash: str = manifest.get("manifest_hash", "unknown_hash")
    temp_dir: str = tempfile.gettempdir()
    memmap_path: str = os.path.join(temp_dir, f"midpoint_panel_{manifest_hash}.npy")

    # Pre-allocate the memmap file
    # mode='w+' creates or overwrites the file
    panel_memmap = np.memmap(memmap_path, dtype=np.float64, mode='w+', shape=shape_3d)

    # Initialize entirely to NaN
    panel_memmap[:] = np.nan
    panel_memmap.flush()

    return memmap_path, panel_memmap, shape_3d

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 2: Left-join sparse midpoints, apply LOCF, and handle initialization
# -------------------------------------------------------------------------------------------------------------------------------

def _numpy_ffill(arr: np.ndarray) -> np.ndarray:
    """
    Highly optimized 1D forward-fill using numpy.
    Replaces NaNs with the most recent non-NaN value.

    Args:
        arr (np.ndarray): 1D float64 array containing NaNs.

    Returns:
        np.ndarray: Forward-filled array.
    """
    mask = np.isnan(arr)
    idx = np.where(~mask, np.arange(mask.shape[0]), 0)
    np.maximum.accumulate(idx, out=idx)
    return arr[idx]

def populate_midpoint_panel(
    df_sparse: pd.DataFrame,
    df_pre_session: pd.DataFrame,
    date_list: List[pd.Timestamp],
    panel_memmap: np.memmap,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Populates the pre-allocated memmap with observed midpoints, initializes the
    first second using pre-session data if necessary, and applies within-day LOCF.

    Args:
        df_sparse (pd.DataFrame): Sparse in-session midpoints.
        df_pre_session (pd.DataFrame): Sparse pre-session midpoints.
        date_list (List[pd.Timestamp]): Sorted list of the 248 retained dates.
        panel_memmap (np.memmap): The pre-allocated 3D memmap array.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Statistics regarding initialization and fill rates.

    Raises:
        ValueError: If a stock-day cannot be initialized (no pre-session or 09:40:00 quote).
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    tickers: tuple = config["asset_universe_exact_appendix_order"]
    start_td = pd.Timedelta(config["session_start_inclusive"])
    start_seconds: int = int(start_td.total_seconds())
    t_day: int = config["price_grid_timestamps_per_day_expected"]

    # Create fast lookup dictionaries mapping (ticker, date) to their respective DataFrames
    # This avoids O(N) filtering inside the nested loop
    sparse_grouped = dict(tuple(df_sparse.groupby(["SYM_ROOT", "DATE"], observed=True)))
    pre_grouped = dict(tuple(df_pre_session.groupby(["SYM_ROOT", "DATE"], observed=True)))

    initialization_failures: List[str] = []
    pre_session_inits: int = 0
    total_observed_points: int = 0

    for k, ticker in enumerate(tickers):
        for d, current_date in enumerate(date_list):
            group_key = (ticker, current_date)

            # Initialize a 1D array with NaNs
            day_vector = np.full(t_day, np.nan, dtype=np.float64)

            # 1. Map observed in-session midpoints to the grid
            if group_key in sparse_grouped:
                df_day = sparse_grouped[group_key]
                # Convert SECOND_BUCKET to integer seconds since midnight, then subtract start_seconds
                seconds_since_midnight = df_day["SECOND_BUCKET"].dt.total_seconds().astype(int).values
                indices = seconds_since_midnight - start_seconds

                # Assign observed values to the correct indices
                day_vector[indices] = df_day["MIDPOINT"].values
                total_observed_points += len(indices)

            # 2. Handle first-second initialization if index 0 is NaN
            if np.isnan(day_vector[0]):
                if group_key in pre_grouped:
                    df_pre_day = pre_grouped[group_key]
                    # The pre-session data is already deduplicated to the last quote per second.
                    # We need the absolute last quote before 09:40:00.
                    # Since it's sorted by time, the last row is the most recent.
                    last_pre_quote = df_pre_day.iloc[-1]
                    day_vector[0] = last_pre_quote["MIDPOINT"]
                    pre_session_inits += 1
                else:
                    # Unresolvable failure: no quote at 09:40:00 and no pre-session quote
                    initialization_failures.append(f"{ticker} on {current_date.date()}")
                    continue

            # 3. Apply Last-Observation-Carried-Forward (LOCF)
            # At this point, day_vector[0] is guaranteed to be non-NaN
            day_vector = _numpy_ffill(day_vector)

            # 4. Write to memmap
            panel_memmap[k, d, :] = day_vector

    # Flush changes to disk
    panel_memmap.flush()

    if initialization_failures:
        raise ValueError(f"Critical Error: {len(initialization_failures)} stock-days could not be initialized due to missing pre-session and opening quotes. Examples: {initialization_failures[:5]}")

    stats = {
        "pre_session_initializations": pre_session_inits,
        "total_observed_points": total_observed_points,
        "total_grid_points": 479 * 248 * 21900,
        "forward_fill_fraction": 1.0 - (total_observed_points / (479 * 248 * 21900))
    }

    return stats

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 3: Validate the fully populated panel
# -------------------------------------------------------------------------------------------------------------------------------

def validate_midpoint_panel(panel_memmap: np.memmap, shape_3d: Tuple[int, int, int]) -> Dict[str, Any]:
    """
    Validates that the memory-mapped panel contains no NaNs, no non-positive
    prices, and computes global summary statistics chunk-by-chunk to conserve memory.

    Args:
        panel_memmap (np.memmap): The populated 3D memmap array.
        shape_3d (Tuple[int, int, int]): The expected shape (479, 248, 21900).

    Returns:
        Dict[str, Any]: Validation report and summary statistics.

    Raises:
        ValueError: If NaNs or non-positive values are detected.
    """
    if panel_memmap.shape != shape_3d:
        raise ValueError(f"Panel shape mismatch. Expected {shape_3d}, got {panel_memmap.shape}")

    k_assets, n_days, t_day = shape_3d

    global_min = float('inf')
    global_max = float('-inf')

    # Process chunk-by-chunk (one stock at a time) to bound memory usage
    for k in range(k_assets):
        stock_slice = panel_memmap[k, :, :]  # Shape: (248, 21900)

        # Check for NaNs
        if np.isnan(stock_slice).any():
            # Find exact location for error message
            nan_indices = np.argwhere(np.isnan(stock_slice))
            first_nan = nan_indices[0]
            raise ValueError(f"NaN detected in panel for stock index {k}, day index {first_nan[0]}, intraday index {first_nan[1]}")

        # Check for non-positive prices
        if not (stock_slice > 0).all():
            raise ValueError(f"Zero or negative price detected in panel for stock index {k}")

        # Update global stats
        chunk_min = float(np.min(stock_slice))
        chunk_max = float(np.max(stock_slice))

        if chunk_min < global_min:
            global_min = chunk_min
        if chunk_max > global_max:
            global_max = chunk_max

    report = {
        "shape_validated": shape_3d,
        "contains_nans": False,
        "all_strictly_positive": True,
        "global_min_price": global_min,
        "global_max_price": global_max
    }

    return report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def build_complete_midpoint_panel(
    df_sparse_midpoints: pd.DataFrame,
    df_pre_session_midpoints: pd.DataFrame,
    date_list: List[pd.Timestamp],
    manifest: MappingProxyType
) -> Tuple[str, Dict[str, Any]]:
    """
    Orchestrates the construction of the dense, synchronized midpoint panel.

    Pre-allocates a memory-mapped array, populates it using exact forward-fill
    and pre-session initialization, and rigorously validates the final dense matrix.

    Args:
        df_sparse_midpoints (pd.DataFrame): Sparse in-session midpoints.
        df_pre_session_midpoints (pd.DataFrame): Sparse pre-session midpoints.
        date_list (List[pd.Timestamp]): Sorted list of retained dates.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, Dict[str, Any]]: A tuple containing:
            - The absolute file path to the validated memory-mapped panel.
            - A dictionary containing build statistics and validation results.

    Raises:
        Exception: Bubbles up any ValueError encountered during allocation,
                   filling, or validation.
    """
    # Step 1: Pre-allocate memmap
    memmap_path, panel_memmap, shape_3d = pre_allocate_midpoint_panel(manifest)

    # Step 2: Populate panel with LOCF
    fill_stats = populate_midpoint_panel(
        df_sparse_midpoints,
        df_pre_session_midpoints,
        date_list,
        panel_memmap,
        manifest
    )

    # Step 3: Validate the populated panel
    val_report = validate_midpoint_panel(panel_memmap, shape_3d)

    # Combine stats
    final_report = {**fill_stats, **val_report}

    # Ensure memmap is flushed and closed (reference deleted)
    panel_memmap.flush()
    del panel_memmap
    gc.collect()

    return memmap_path, final_report
````

````python
# Task 7: Compute intraday 1-second log returns

# ==============================================================================
# Task 7: Compute intraday 1-second log returns
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 1: Compute within-day 1-second log returns for each stock
# -------------------------------------------------------------------------------------------------------------------------------

def compute_chunked_log_returns(
    midpoint_panel_path: str,
    manifest: MappingProxyType
) -> Tuple[str, np.memmap]:
    """
    Computes 1-second log returns from the midpoint panel.
    Processes data chunk-by-chunk (stock-by-stock) to maintain a low memory footprint.
    Strictly avoids computing returns across overnight boundaries.

    Equation: G_k(t, \Delta t) = \ln m_k(t + \Delta t) - \ln m_k(t)

    Args:
        midpoint_panel_path (str): Path to the 3D midpoint panel memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, np.memmap]:
            - The absolute file path to the return matrix memmap.
            - The writeable memmap object for the return matrix.

    Raises:
        ValueError: If any midpoint is zero or negative, preventing log computation.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    n_days: int = config["N_days_expected_after_filters"]
    t_day_midpoints: int = config["price_grid_timestamps_per_day_expected"]
    t_day_returns: int = config["return_intervals_per_day_expected"]
    t_total_returns: int = config["T_total_return_intervals_expected"]

    # Open input panel in read-only mode
    shape_in = (k_assets, n_days, t_day_midpoints)
    panel_in = np.memmap(midpoint_panel_path, dtype=np.float64, mode='r', shape=shape_in)

    # Pre-allocate output return matrix
    manifest_hash: str = manifest.get("manifest_hash", "unknown_hash")
    temp_dir: str = tempfile.gettempdir()
    returns_path: str = os.path.join(temp_dir, f"return_matrix_{manifest_hash}.npy")

    shape_out = (k_assets, t_total_returns)
    returns_out = np.memmap(returns_path, dtype=np.float64, mode='w+', shape=shape_out)

    # Process stock by stock to bound memory usage
    for k in range(k_assets):
        # Extract the 2D slice for stock k: shape (248, 21900)
        stock_midpoints = panel_in[k, :, :]

        # Strict validation before log
        if not (stock_midpoints > 0).all():
            raise ValueError(f"Cannot compute log returns: stock index {k} contains zero or negative midpoints.")

        # Compute log midpoints
        log_midpoints = np.log(stock_midpoints)

        # Compute differences along the intraday axis (axis=1 for the 2D slice)
        # This strictly prevents cross-day returns. Shape becomes (248, 21899)
        stock_returns_2d = np.diff(log_midpoints, axis=1)

        # Flatten to 1D: shape (5430952,)
        stock_returns_1d = stock_returns_2d.reshape(-1)

        # Write to output memmap
        returns_out[k, :] = stock_returns_1d

    returns_out.flush()

    # Explicitly delete the read-only memmap reference
    del panel_in
    gc.collect()

    return returns_path, returns_out

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 2: Construct the day-boundary index map
# -------------------------------------------------------------------------------------------------------------------------------

def construct_day_boundary_map(
    date_list: List[pd.Timestamp],
    manifest: MappingProxyType
) -> Tuple[Tuple[pd.Timestamp, int, int], ...]:
    """
    Constructs an immutable map linking calendar dates to their corresponding
    column index ranges in the flattened return matrix.

    Args:
        date_list (List[pd.Timestamp]): Sorted list of the 248 retained dates.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[Tuple[pd.Timestamp, int, int], ...]: A tuple of tuples, where each
        inner tuple is (date, start_col_inclusive, end_col_exclusive).

    Raises:
        ValueError: If the final end_col does not match the expected total T.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    t_day_returns: int = config["return_intervals_per_day_expected"]
    t_total_returns: int = config["T_total_return_intervals_expected"]

    boundaries: List[Tuple[pd.Timestamp, int, int]] = []

    for d, current_date in enumerate(date_list):
        start_col = d * t_day_returns
        end_col = (d + 1) * t_day_returns
        boundaries.append((current_date, start_col, end_col))

    # Validate the final boundary
    if boundaries[-1][2] != t_total_returns:
        raise ValueError(f"Boundary map generation failed. Final end_col {boundaries[-1][2]} != expected T {t_total_returns}")

    return tuple(boundaries)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 3: Validate exact dimensions and properties of the return matrix
# -------------------------------------------------------------------------------------------------------------------------------

def validate_return_matrix(
    returns_memmap: np.memmap,
    day_boundary_map: Tuple[Tuple[pd.Timestamp, int, int], ...],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Validates the dimensions, day counts, and numerical health (no NaNs/Infs)
    of the return matrix. Computes global summary statistics.

    Args:
        returns_memmap (np.memmap): The populated return matrix memmap.
        day_boundary_map (Tuple[Tuple[pd.Timestamp, int, int], ...]): The day boundary map.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Validation report and summary statistics.

    Raises:
        ValueError: If shape mismatches, day counts are wrong, or NaNs/Infs are found.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total_returns: int = config["T_total_return_intervals_expected"]
    t_day_returns: int = config["return_intervals_per_day_expected"]

    expected_shape = (k_assets, t_total_returns)

    if returns_memmap.shape != expected_shape:
        raise ValueError(f"Return matrix shape mismatch. Expected {expected_shape}, got {returns_memmap.shape}")

    # Validate per-day return count from the map
    for date_val, start_col, end_col in day_boundary_map:
        if end_col - start_col != t_day_returns:
            raise ValueError(f"Day boundary map error for {date_val.date()}: count is {end_col - start_col}, expected {t_day_returns}")

    global_min = float('inf')
    global_max = float('-inf')

    # We need to compute global mean and std.
    # To do this chunk-by-chunk without loading the whole array:
    sum_x = 0.0
    sum_x2 = 0.0
    total_elements = k_assets * t_total_returns

    for k in range(k_assets):
        stock_returns = returns_memmap[k, :]

        if np.isnan(stock_returns).any():
            raise ValueError(f"NaN detected in return matrix for stock index {k}")

        if np.isinf(stock_returns).any():
            raise ValueError(f"Inf detected in return matrix for stock index {k}")

        chunk_min = float(np.min(stock_returns))
        chunk_max = float(np.max(stock_returns))

        if chunk_min < global_min:
            global_min = chunk_min
        if chunk_max > global_max:
            global_max = chunk_max

        # Accumulate for global moments
        # Using float64 for accumulation to minimize precision loss
        sum_x += np.sum(stock_returns, dtype=np.float64)
        sum_x2 += np.sum(stock_returns**2, dtype=np.float64)

    global_mean = sum_x / total_elements
    global_var = (sum_x2 / total_elements) - (global_mean**2)
    global_std = np.sqrt(global_var) if global_var > 0 else 0.0

    report = {
        "shape_validated": expected_shape,
        "contains_nans": False,
        "contains_infs": False,
        "global_min_return": global_min,
        "global_max_return": global_max,
        "global_mean_return": float(global_mean),
        "global_std_return": float(global_std)
    }

    return report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def compute_log_returns(
    midpoint_panel_path: str,
    date_list: List[pd.Timestamp],
    manifest: MappingProxyType
) -> Tuple[str, Tuple[Tuple[pd.Timestamp, int, int], ...], Dict[str, Any]]:
    """
    Orchestrates the computation of intraday 1-second log returns.

    Reads the midpoint panel, computes returns strictly within day boundaries,
    constructs the day-boundary index map, and validates the resulting matrix.

    Args:
        midpoint_panel_path (str): Path to the 3D midpoint panel memmap.
        date_list (List[pd.Timestamp]): Sorted list of retained dates.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, Tuple[Tuple[pd.Timestamp, int, int], ...], Dict[str, Any]]:
            - The absolute file path to the return matrix memmap.
            - The immutable day-boundary index map.
            - A dictionary containing validation statistics.

    Raises:
        Exception: Bubbles up any ValueError encountered during computation or validation.
    """
    # Step 1: Compute chunked log returns
    returns_path, returns_memmap = compute_chunked_log_returns(midpoint_panel_path, manifest)

    # Step 2: Construct day boundary map
    day_boundary_map = construct_day_boundary_map(date_list, manifest)

    # Step 3: Validate the return matrix
    val_report = validate_return_matrix(returns_memmap, day_boundary_map, manifest)

    # Ensure memmap is flushed and closed
    returns_memmap.flush()
    del returns_memmap
    gc.collect()

    return returns_path, day_boundary_map, val_report
````

````python
# Task 8: Standardize returns to the normalized data matrix M

# ==============================================================================
# Task 8: Standardize returns to the normalized data matrix M
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 1: Compute per-stock full-sample mean and standard deviation
# -------------------------------------------------------------------------------------------------------------------------------

def compute_population_moments(
    returns_memmap: np.memmap,
    manifest: MappingProxyType
) -> Tuple[np.ndarray, np.ndarray]:
    """
    Computes the full-sample mean and population standard deviation (ddof=0)
    for each stock's return series. Processes chunk-by-chunk to bound memory.

    Equations:
        <G_k(t)>_t = (1/T) * sum(G_k(t))
        sigma_k = sqrt((1/T) * sum((G_k(t) - <G_k(t)>_t)^2))

    Args:
        returns_memmap (np.memmap): The 2D return matrix memmap (shape K x T).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[np.ndarray, np.ndarray]:
            - 1D array of means (length K).
            - 1D array of standard deviations (length K).

    Raises:
        ValueError: If any standard deviation is extremely small (< 1e-15),
                    indicating a flat price series.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]

    means = np.zeros(k_assets, dtype=np.float64)
    stds = np.zeros(k_assets, dtype=np.float64)

    for k in range(k_assets):
        # Extract 1D slice for stock k
        stock_returns: np.ndarray = returns_memmap[k, :]

        # Compute mean using float64 accumulation for stability
        mean_k = np.mean(stock_returns, dtype=np.float64)

        # Compute population standard deviation (ddof=0)
        # This is strictly required so that the downstream correlation matrix
        # has exactly 1.0 on its diagonal.
        std_k = np.std(stock_returns, dtype=np.float64, ddof=0)

        if std_k < 1e-15:
            raise ValueError(f"Standard deviation for stock index {k} is {std_k}, which is too close to zero. This indicates a flat price series.")

        means[k] = mean_k
        stds[k] = std_k

    return means, stds

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 2: Standardize each return series in-place
# -------------------------------------------------------------------------------------------------------------------------------

def standardize_returns_inplace(
    returns_memmap: np.memmap,
    means: np.ndarray,
    stds: np.ndarray,
    manifest: MappingProxyType
) -> None:
    """
    Standardizes the return matrix in-place to produce the normalized data matrix M.
    In-place modification is used to avoid doubling the 20.4 GB memory/disk footprint.

    Equation: M_k(t) = (G_k(t) - <G_k(t)>_t) / sigma_k

    Args:
        returns_memmap (np.memmap): The 2D return matrix memmap (opened in r+ mode).
        means (np.ndarray): 1D array of means.
        stds (np.ndarray): 1D array of standard deviations.
        manifest (MappingProxyType): The frozen run manifest.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]

    for k in range(k_assets):
        # Extract 1D slice (this is a view into the memmap)
        stock_returns: np.ndarray = returns_memmap[k, :]

        # Perform in-place standardization
        # Subtract mean
        stock_returns -= means[k]
        # Divide by standard deviation
        stock_returns /= stds[k]

    # Flush all modifications to disk
    returns_memmap.flush()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 3: Validate zero-mean and unit-variance properties
# -------------------------------------------------------------------------------------------------------------------------------

def validate_normalized_matrix(
    normalized_memmap: np.memmap,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Validates that every row of the normalized matrix M has a mean of exactly 0
    and a population variance of exactly 1, within strict floating-point tolerances.

    Args:
        normalized_memmap (np.memmap): The standardized 2D memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Validation report containing maximum deviations.

    Raises:
        ValueError: If any row violates the mean or variance tolerances.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]

    eps_mean: float = 1e-10
    eps_var: float = 1e-8

    max_mean_dev: float = 0.0
    max_var_dev: float = 0.0

    for k in range(k_assets):
        stock_m: np.ndarray = normalized_memmap[k, :]

        # Compute empirical mean
        emp_mean = np.mean(stock_m, dtype=np.float64)
        mean_dev = abs(emp_mean)

        if mean_dev > max_mean_dev:
            max_mean_dev = mean_dev

        if mean_dev >= eps_mean:
            raise ValueError(f"Normalization failure: Stock index {k} has mean {emp_mean}, exceeding tolerance {eps_mean}")

        # Compute empirical population variance: (1/T) * sum(M_k(t)^2)
        # Since mean is ~0, sum of squares / T is the variance
        emp_var = np.mean(stock_m**2, dtype=np.float64)
        var_dev = abs(emp_var - 1.0)

        if var_dev > max_var_dev:
            max_var_dev = var_dev

        if var_dev >= eps_var:
            raise ValueError(f"Normalization failure: Stock index {k} has variance {emp_var}, exceeding tolerance {eps_var}")

    report = {
        "normalization_validated": True,
        "max_mean_deviation": float(max_mean_dev),
        "max_variance_deviation": float(max_var_dev),
        "tolerance_mean": eps_mean,
        "tolerance_variance": eps_var
    }

    return report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def standardize_returns(
    returns_memmap_path: str,
    manifest: MappingProxyType
) -> Tuple[str, np.ndarray, np.ndarray, Dict[str, Any]]:
    """
    Orchestrates the standardization of the return matrix to produce the
    normalized data matrix M.

    Computes population moments, applies in-place standardization to conserve
    memory, and rigorously validates the zero-mean, unit-variance properties.

    Args:
        returns_memmap_path (str): Path to the 2D return matrix memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, np.ndarray, np.ndarray, Dict[str, Any]]:
            - The absolute file path to the normalized matrix memmap (same as input path).
            - 1D array of per-stock means.
            - 1D array of per-stock standard deviations.
            - Validation report dictionary.

    Raises:
        Exception: Bubbles up any ValueError encountered during moment computation
                   or strict validation.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    # Open the memmap in read-write mode ('r+') to allow in-place modification
    returns_memmap = np.memmap(returns_memmap_path, dtype=np.float64, mode='r+', shape=shape_2d)

    # Step 1: Compute moments
    means, stds = compute_population_moments(returns_memmap, manifest)

    # Step 2: Standardize in-place
    standardize_returns_inplace(returns_memmap, means, stds, manifest)

    # Step 3: Validate properties
    val_report = validate_normalized_matrix(returns_memmap, manifest)

    # Ensure memmap is flushed and closed
    returns_memmap.flush()
    del returns_memmap
    gc.collect()

    # The file at returns_memmap_path now contains the normalized matrix M
    normalized_memmap_path = returns_memmap_path

    return normalized_memmap_path, means, stds, val_report
````

````python
# Task 9: Compute the Pearson correlation matrix

# ==============================================================================
# Task 9: Compute the Pearson correlation matrix
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 1: Compute the Pearson correlation matrix via chunkwise accumulation
# -------------------------------------------------------------------------------------------------------------------------------

def compute_dense_correlation_matrix(
    normalized_memmap: np.memmap,
    manifest: MappingProxyType
) -> np.ndarray:
    """
    Computes the Pearson correlation matrix C = (1/T) * M * M^T using a
    memory-efficient chunkwise accumulation strategy.

    Strictly adheres to the manuscript: no shrinkage, no asynchronous corrections,
    and no Epps-effect adjustments are applied.

    Args:
        normalized_memmap (np.memmap): The 2D normalized data matrix M (shape K x T).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        np.ndarray: The dense, un-symmetrized correlation matrix (shape K x K).
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]

    # Initialize the accumulator in float64 to prevent precision loss
    c_accum = np.zeros((k_assets, k_assets), dtype=np.float64)

    # Define chunk size (number of columns to process at once)
    # 100,000 columns * 479 rows * 8 bytes = ~383 MB per chunk
    chunk_size: int = 100_000
    num_chunks: int = int(np.ceil(t_total / chunk_size))

    for b in range(num_chunks):
        start_col = b * chunk_size
        end_col = min((b + 1) * chunk_size, t_total)

        # Extract chunk (view into memmap)
        m_chunk: np.ndarray = normalized_memmap[:, start_col:end_col]

        # Compute M_chunk * M_chunk^T and accumulate
        # This delegates to highly optimized BLAS DGEMM
        c_accum += m_chunk @ m_chunk.T

    # Divide by T to finalize the Pearson correlation formula
    c_matrix: np.ndarray = c_accum / float(t_total)

    return c_matrix

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 2: Validate symmetry and symmetrize
# -------------------------------------------------------------------------------------------------------------------------------

def validate_and_symmetrize_matrix(c_matrix: np.ndarray) -> Tuple[np.ndarray, float]:
    """
    Validates the symmetry of the computed correlation matrix and explicitly
    symmetrizes it to guarantee exact bitwise symmetry for the eigensolver.

    Args:
        c_matrix (np.ndarray): The raw computed correlation matrix.

    Returns:
        Tuple[np.ndarray, float]:
            - The explicitly symmetrized correlation matrix.
            - The maximum absolute asymmetry before correction.

    Raises:
        ValueError: If the initial asymmetry exceeds the strict tolerance (1e-12).
    """
    # Compute maximum absolute asymmetry: || C - C^T ||_max
    delta_sym: float = float(np.max(np.abs(c_matrix - c_matrix.T)))

    tolerance: float = 1e-12
    if delta_sym >= tolerance:
        raise ValueError(f"Correlation matrix asymmetry ({delta_sym}) exceeds tolerance ({tolerance}). This indicates a severe numerical accumulation error.")

    # Explicitly symmetrize: C <- (C + C^T) / 2
    c_sym: np.ndarray = (c_matrix + c_matrix.T) / 2.0

    # Defensive check: ensure exact bitwise symmetry
    if not np.array_equal(c_sym, c_sym.T):
        raise RuntimeError("Explicit symmetrization failed to produce a bitwise symmetric matrix.")

    return c_sym, delta_sym

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 3: Validate diagonal and off-diagonal bounds
# -------------------------------------------------------------------------------------------------------------------------------

def validate_correlation_bounds(c_sym: np.ndarray) -> Dict[str, Any]:
    """
    Validates that all diagonal elements of the correlation matrix are exactly 1
    (within tolerance) and that all off-diagonal elements are bounded by [-1, 1].

    Args:
        c_sym (np.ndarray): The symmetrized correlation matrix.

    Returns:
        Dict[str, Any]: Validation report containing maximum deviations.

    Raises:
        ValueError: If the diagonal deviates from 1.0 or off-diagonals exceed bounds.
    """
    diag_elements: np.ndarray = np.diag(c_sym)

    # Compute maximum deviation from unity on the diagonal
    delta_diag: float = float(np.max(np.abs(diag_elements - 1.0)))

    tolerance_diag: float = 1e-10
    if delta_diag >= tolerance_diag:
        # Find the specific index for the error message
        bad_idx = int(np.argmax(np.abs(diag_elements - 1.0)))
        bad_val = diag_elements[bad_idx]
        raise ValueError(f"Correlation matrix diagonal validation failed. Index {bad_idx} has variance {bad_val}, deviation {delta_diag} exceeds tolerance {tolerance_diag}. Check Task 8 normalization.")

    # Check off-diagonal bounds
    # We subtract the diagonal to isolate off-diagonals, though technically
    # the diagonal is 1.0 so max(abs(C)) should be <= 1.0 overall.
    max_abs_val: float = float(np.max(np.abs(c_sym)))

    tolerance_bounds: float = 1e-10
    if max_abs_val > 1.0 + tolerance_bounds:
        raise ValueError(f"Correlation matrix contains values exceeding 1.0 (max absolute value: {max_abs_val}).")

    report = {
        "correlation_matrix_validated": True,
        "max_asymmetry_before_correction": None, # Will be injected by orchestrator
        "max_diagonal_deviation": delta_diag,
        "max_absolute_correlation": max_abs_val,
        "tolerance_diagonal": tolerance_diag
    }

    return report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def compute_correlation_matrix(
    normalized_memmap_path: str,
    manifest: MappingProxyType
) -> Tuple[np.ndarray, Dict[str, Any]]:
    """
    Orchestrates the computation of the Pearson correlation matrix.

    Reads the normalized data matrix chunk-by-chunk, computes the dense
    correlation matrix, enforces exact symmetry, and rigorously validates
    the diagonal and bounds.

    Args:
        normalized_memmap_path (str): Path to the 2D normalized matrix memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[np.ndarray, Dict[str, Any]]:
            - The validated, symmetric correlation matrix C (shape K x K) in RAM.
            - A dictionary containing validation statistics.

    Raises:
        Exception: Bubbles up any ValueError encountered during computation or validation.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    # Open the memmap in read-only mode
    normalized_memmap = np.memmap(normalized_memmap_path, dtype=np.float64, mode='r', shape=shape_2d)

    # Step 1: Compute raw correlation matrix
    c_raw = compute_dense_correlation_matrix(normalized_memmap, manifest)

    # Explicitly delete the memmap reference to free file handles
    del normalized_memmap
    gc.collect()

    # Step 2: Validate symmetry and symmetrize
    c_sym, delta_sym = validate_and_symmetrize_matrix(c_raw)

    # Step 3: Validate bounds
    val_report = validate_correlation_bounds(c_sym)
    val_report["max_asymmetry_before_correction"] = delta_sym

    return c_sym, val_report
````

````python
# Task 10: Perform spectral decomposition of the correlation matrix

# ==============================================================================
# Task 10: Perform spectral decomposition of the correlation matrix
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 1: Compute eigendecomposition and sort descending
# -------------------------------------------------------------------------------------------------------------------------------

def compute_sorted_eigendecomposition(c_sym: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    """
    Computes the spectral decomposition of the symmetric correlation matrix.
    Sorts the eigenvalues and corresponding eigenvectors in strictly descending order.

    Equation: C = U \Lambda U^T

    Args:
        c_sym (np.ndarray): The validated, symmetric correlation matrix (K x K).

    Returns:
        Tuple[np.ndarray, np.ndarray]:
            - 1D array of eigenvalues sorted descending (length K).
            - 2D array of eigenvectors (K x K), where column j corresponds to eigenvalue j.
    """
    # Use scipy.linalg.eigh for symmetric matrices. It is faster and more
    # numerically stable than numpy.linalg.eig, and guarantees real eigenvalues.
    # By default, eigh returns eigenvalues in ascending order.
    eigenvalues_asc, eigenvectors_asc = scipy.linalg.eigh(c_sym)

    # Reverse to achieve strictly descending order: \Lambda_1 >= \Lambda_2 >= ... >= \Lambda_K
    # .copy() is used to ensure the resulting arrays are contiguous in memory,
    # which optimizes downstream BLAS operations.
    eigenvalues_desc: np.ndarray = eigenvalues_asc[::-1].copy()
    eigenvectors_desc: np.ndarray = eigenvectors_asc[:, ::-1].copy()

    return eigenvalues_desc, eigenvectors_desc

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 2: Validate reconstruction residual and orthonormality
# -------------------------------------------------------------------------------------------------------------------------------

def validate_eigendecomposition(
    c_sym: np.ndarray,
    eigenvalues: np.ndarray,
    eigenvectors: np.ndarray
) -> Dict[str, Any]:
    """
    Validates the numerical accuracy of the spectral decomposition by checking
    the reconstruction residual and the orthonormality of the eigenvector matrix.

    Args:
        c_sym (np.ndarray): The original symmetric correlation matrix.
        eigenvalues (np.ndarray): The sorted eigenvalues.
        eigenvectors (np.ndarray): The sorted eigenvector matrix U.

    Returns:
        Dict[str, Any]: Diagnostic metrics for the decomposition.

    Raises:
        ValueError: If the reconstruction error or orthonormality error exceeds strict tolerances.
    """
    k_assets = c_sym.shape[0]

    # 1. Reconstruction residual check: || C U - U \Lambda ||_max < 10^-8
    # U * eigenvalues leverages numpy broadcasting to multiply each column j of U by \Lambda_j
    reconstruction_matrix = c_sym @ eigenvectors - eigenvectors * eigenvalues[np.newaxis, :]
    max_residual: float = float(np.max(np.abs(reconstruction_matrix)))

    tol_residual: float = 1e-8
    if max_residual >= tol_residual:
        raise ValueError(f"Eigendecomposition reconstruction failed. Max residual {max_residual} >= tolerance {tol_residual}")

    # 2. Orthonormality check: || U^T U - I ||_max < 10^-10
    identity_matrix = np.eye(k_assets, dtype=np.float64)
    ortho_matrix = eigenvectors.T @ eigenvectors
    max_ortho_error: float = float(np.max(np.abs(ortho_matrix - identity_matrix)))

    tol_ortho: float = 1e-10
    if max_ortho_error >= tol_ortho:
        raise ValueError(f"Eigenvector orthonormality failed. Max error {max_ortho_error} >= tolerance {tol_ortho}")

    diagnostics = {
        "max_reconstruction_residual": max_residual,
        "max_orthonormality_error": max_ortho_error,
        "tolerance_residual": tol_residual,
        "tolerance_ortho": tol_ortho,
        "decomposition_validated": True
    }

    return diagnostics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 3: Validate non-negativity and handle near-zero negatives
# -------------------------------------------------------------------------------------------------------------------------------

def validate_and_adjust_eigenvalues(
    eigenvalues: np.ndarray,
    k_assets: int,
    diagnostics: Dict[str, Any]
) -> np.ndarray:
    """
    Validates that all eigenvalues are non-negative. Floors any near-zero
    negative eigenvalues (numerical noise) to a small positive constant to
    prevent downstream division-by-zero during whitening.

    Also verifies the trace identity: sum(\Lambda_k) == tr(C) == K.

    Args:
        eigenvalues (np.ndarray): The sorted eigenvalues.
        k_assets (int): The dimension of the matrix (479).
        diagnostics (Dict[str, Any]): The diagnostics dictionary to update.

    Returns:
        np.ndarray: The validated and potentially adjusted eigenvalue array.

    Raises:
        ValueError: If any eigenvalue is significantly negative, indicating
                    the correlation matrix was not positive semi-definite.
    """
    adjusted_eigenvalues = eigenvalues.copy()
    min_eigenvalue: float = float(np.min(adjusted_eigenvalues))

    diagnostics["raw_min_eigenvalue"] = min_eigenvalue
    diagnostics["eigenvalues_adjusted"] = 0

    # Check for negative eigenvalues
    if min_eigenvalue < 0:
        # Define the threshold for numerical noise vs structural failure
        noise_threshold = -1e-6
        floor_value = 1e-10

        if min_eigenvalue <= noise_threshold:
            raise ValueError(f"Correlation matrix is not positive semi-definite. Minimum eigenvalue is {min_eigenvalue}, which exceeds the noise threshold of {noise_threshold}.")

        # Identify indices of near-zero negative eigenvalues
        negative_mask = adjusted_eigenvalues < 0
        num_adjusted = int(np.sum(negative_mask))

        # Apply the floor
        adjusted_eigenvalues[negative_mask] = floor_value

        diagnostics["eigenvalues_adjusted"] = num_adjusted
        diagnostics["adjustment_floor_applied"] = floor_value

    # Trace identity check: sum of eigenvalues must equal the trace of the correlation matrix.
    # Since the correlation matrix has 1.0 on the diagonal, the trace is exactly K.
    sum_eigenvalues: float = float(np.sum(adjusted_eigenvalues))
    trace_error: float = abs(sum_eigenvalues - k_assets)

    tol_trace: float = 1e-6
    if trace_error >= tol_trace:
        raise ValueError(f"Trace identity failed. Sum of eigenvalues ({sum_eigenvalues}) differs from K ({k_assets}) by {trace_error} >= {tol_trace}")

    diagnostics["trace_identity_error"] = trace_error

    return adjusted_eigenvalues

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def spectral_decomposition(
    c_sym: np.ndarray,
    manifest: MappingProxyType
) -> Tuple[np.ndarray, np.ndarray, Dict[str, Any]]:
    """
    Orchestrates the spectral decomposition of the correlation matrix.

    Computes the eigendecomposition, sorts modes in descending order, rigorously
    validates reconstruction and orthonormality, and ensures all eigenvalues
    are strictly positive for downstream whitening.

    Args:
        c_sym (np.ndarray): The validated, symmetric correlation matrix.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[np.ndarray, np.ndarray, Dict[str, Any]]:
            - 1D array of validated/adjusted eigenvalues (descending).
            - 2D array of eigenvectors U.
            - Dictionary containing decomposition diagnostics.

    Raises:
        Exception: Bubbles up any ValueError encountered during decomposition or validation.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]

    # Step 1: Compute and sort
    eigenvalues_raw, eigenvectors = compute_sorted_eigendecomposition(c_sym)

    # Step 2: Validate reconstruction and orthonormality
    diagnostics = validate_eigendecomposition(c_sym, eigenvalues_raw, eigenvectors)

    # Step 3: Validate non-negativity and adjust if necessary
    eigenvalues_final = validate_and_adjust_eigenvalues(eigenvalues_raw, k_assets, diagnostics)

    return eigenvalues_final, eigenvectors, diagnostics
````

````python
# Task 11: Construct the whitened mode matrix via PCA-whitening

# ==============================================================================
# Task 11: Construct the whitened mode matrix via PCA-whitening
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 1: Compute the unscaled rotated returns Y = U^T M
# -------------------------------------------------------------------------------------------------------------------------------

def compute_unscaled_rotated_returns(
    normalized_memmap_path: str,
    eigenvectors: np.ndarray,
    manifest: MappingProxyType
) -> Tuple[str, np.memmap]:
    """
    Computes the unscaled rotated returns Y = U^T M using a memory-efficient
    chunkwise matrix multiplication strategy.

    Args:
        normalized_memmap_path (str): Path to the 2D normalized matrix M (K x T).
        eigenvectors (np.ndarray): The sorted eigenvector matrix U (K x K).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, np.memmap]:
            - The absolute file path to the Y matrix memmap.
            - The writeable memmap object for Y.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    # Open M in read-only mode
    m_memmap = np.memmap(normalized_memmap_path, dtype=np.float64, mode='r', shape=shape_2d)

    # Pre-allocate Y memmap
    manifest_hash: str = manifest.get("manifest_hash", "unknown_hash")
    temp_dir: str = tempfile.gettempdir()
    y_path: str = os.path.join(temp_dir, f"rotated_unscaled_Y_{manifest_hash}.npy")
    y_memmap = np.memmap(y_path, dtype=np.float64, mode='w+', shape=shape_2d)

    # U^T has shape (K x K)
    u_t: np.ndarray = eigenvectors.T

    # Process in chunks of columns to bound memory
    chunk_size: int = 100_000
    num_chunks: int = int(np.ceil(t_total / chunk_size))

    for b in range(num_chunks):
        start_col = b * chunk_size
        end_col = min((b + 1) * chunk_size, t_total)

        # Extract chunk of M: shape (K x chunk_size)
        m_chunk: np.ndarray = m_memmap[:, start_col:end_col]

        # Compute Y_chunk = U^T * M_chunk
        y_chunk: np.ndarray = u_t @ m_chunk

        # Write to Y memmap
        y_memmap[:, start_col:end_col] = y_chunk

    y_memmap.flush()

    # Explicitly delete M memmap reference
    del m_memmap
    gc.collect()

    return y_path, y_memmap

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 2: Compute the whitened mode matrix R = \Lambda^{-1/2} Y
# -------------------------------------------------------------------------------------------------------------------------------

def compute_whitened_modes(
    y_memmap: np.memmap,
    eigenvalues: np.ndarray,
    manifest: MappingProxyType
) -> Tuple[str, np.memmap]:
    """
    Computes the whitened mode matrix R by scaling each row of Y by the inverse
    square root of its corresponding eigenvalue.

    Equation: R_k(t) = Y_k(t) / sqrt(\Lambda_k)

    Args:
        y_memmap (np.memmap): The unscaled rotated returns Y (K x T).
        eigenvalues (np.ndarray): The validated, strictly positive eigenvalues.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, np.memmap]:
            - The absolute file path to the R matrix memmap.
            - The writeable memmap object for R.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    # Pre-allocate R memmap
    manifest_hash: str = manifest.get("manifest_hash", "unknown_hash")
    temp_dir: str = tempfile.gettempdir()
    r_path: str = os.path.join(temp_dir, f"whitened_modes_R_{manifest_hash}.npy")
    r_memmap = np.memmap(r_path, dtype=np.float64, mode='w+', shape=shape_2d)

    # Compute scale factors: \Lambda^{-1/2}
    # Eigenvalues are guaranteed positive by Task 10
    scale_factors: np.ndarray = 1.0 / np.sqrt(eigenvalues)

    # Process chunk-by-chunk to bound memory
    chunk_size: int = 100_000
    num_chunks: int = int(np.ceil(t_total / chunk_size))

    for b in range(num_chunks):
        start_col = b * chunk_size
        end_col = min((b + 1) * chunk_size, t_total)

        # Extract chunk of Y
        y_chunk: np.ndarray = y_memmap[:, start_col:end_col]

        # Scale rows using numpy broadcasting
        r_chunk: np.ndarray = y_chunk * scale_factors[:, np.newaxis]

        # Write to R memmap
        r_memmap[:, start_col:end_col] = r_chunk

    r_memmap.flush()

    return r_path, r_memmap

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 3: Validate whitening and apply deterministic sign convention
# -------------------------------------------------------------------------------------------------------------------------------

def validate_and_apply_sign_convention(
    r_memmap: np.memmap,
    y_memmap: np.memmap,
    eigenvectors: np.ndarray,
    eigenvalues: np.ndarray,
    manifest: MappingProxyType
) -> Tuple[np.ndarray, Dict[str, Any], List[Dict[str, Any]]]:
    """
    Validates that the empirical covariance of R is the identity matrix.
    Applies a deterministic sign convention (sum of eigenvector entries > 0)
    to U, Y, and R for visual consistency.

    Args:
        r_memmap (np.memmap): The whitened modes R (opened in r+ mode).
        y_memmap (np.memmap): The unscaled modes Y (opened in r+ mode).
        eigenvectors (np.ndarray): The eigenvector matrix U.
        eigenvalues (np.ndarray): The eigenvalues.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[np.ndarray, Dict[str, Any], List[Dict[str, Any]]]:
            - The sign-adjusted eigenvector matrix U.
            - Validation report dictionary.
            - Metadata dictionary for the top 5 modes.

    Raises:
        ValueError: If the empirical covariance of R deviates from Identity.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]

    # 1. Validate Whitening: || (1/T) R R^T - I ||_max < 10^-6
    cov_r_accum = np.zeros((k_assets, k_assets), dtype=np.float64)
    chunk_size: int = 100_000
    num_chunks: int = int(np.ceil(t_total / chunk_size))

    for b in range(num_chunks):
        start_col = b * chunk_size
        end_col = min((b + 1) * chunk_size, t_total)
        r_chunk = r_memmap[:, start_col:end_col]
        cov_r_accum += r_chunk @ r_chunk.T

    cov_r = cov_r_accum / float(t_total)
    identity = np.eye(k_assets, dtype=np.float64)
    whitening_error = float(np.max(np.abs(cov_r - identity)))

    tol_whitening = 1e-6
    if whitening_error >= tol_whitening:
        raise ValueError(f"Whitening validation failed. Max covariance deviation {whitening_error} >= tolerance {tol_whitening}")

    # 2. Apply Deterministic Sign Convention
    u_adjusted = eigenvectors.copy()
    sign_flips_applied = 0

    for k in range(k_assets):
        # Sum of entries for eigenvector k (column k)
        sum_entries = np.sum(u_adjusted[:, k])

        if sum_entries < 0:
            # Flip sign of eigenvector
            u_adjusted[:, k] = -u_adjusted[:, k]
            # Flip sign of corresponding row in Y and R
            y_memmap[k, :] = -y_memmap[k, :]
            r_memmap[k, :] = -r_memmap[k, :]
            sign_flips_applied += 1

    y_memmap.flush()
    r_memmap.flush()

    # 3. Generate Mode Metadata for Top 5 Modes
    mode_metadata: List[Dict[str, Any]] = []
    for k in range(5):
        u_k = u_adjusted[:, k]
        metadata = {
            "mode_index_1_based": k + 1,
            "eigenvalue": float(eigenvalues[k]),
            "eigenvector_sum": float(np.sum(u_k)),
            "eigenvector_std": float(np.std(u_k)),
            "sign_flipped": np.sum(eigenvectors[:, k]) < 0
        }
        mode_metadata.append(metadata)

    report = {
        "whitening_validated": True,
        "max_covariance_deviation": whitening_error,
        "tolerance_whitening": tol_whitening,
        "sign_flips_applied": sign_flips_applied
    }

    return u_adjusted, report, mode_metadata

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def construct_whitened_modes(
    normalized_memmap_path: str,
    eigenvectors: np.ndarray,
    eigenvalues: np.ndarray,
    manifest: MappingProxyType
) -> Tuple[str, str, np.ndarray, List[Dict[str, Any]], Dict[str, Any]]:
    """
    Orchestrates the construction of the unscaled and whitened mode matrices.

    Computes Y = U^T M and R = \Lambda^{-1/2} Y using chunked operations to
    bound memory. Validates the whitening transformation and applies a
    deterministic sign convention for visual consistency.

    Args:
        normalized_memmap_path (str): Path to the 2D normalized matrix M.
        eigenvectors (np.ndarray): The eigenvector matrix U.
        eigenvalues (np.ndarray): The validated eigenvalues.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[str, str, np.ndarray, List[Dict[str, Any]], Dict[str, Any]]:
            - Path to the unscaled rotated returns Y memmap.
            - Path to the whitened modes R memmap.
            - The sign-adjusted eigenvector matrix U.
            - Metadata dictionary for the top 5 modes.
            - Validation report dictionary.

    Raises:
        Exception: Bubbles up any ValueError encountered during computation or validation.
    """
    # Step 1: Compute Y
    y_path, y_memmap = compute_unscaled_rotated_returns(normalized_memmap_path, eigenvectors, manifest)

    # Step 2: Compute R
    r_path, r_memmap = compute_whitened_modes(y_memmap, eigenvalues, manifest)

    # Step 3: Validate and apply sign convention
    # Note: y_memmap and r_memmap are passed as writeable objects and modified in-place
    u_adjusted, val_report, mode_metadata = validate_and_apply_sign_convention(
        r_memmap, y_memmap, eigenvectors, eigenvalues, manifest
    )

    # Ensure memmaps are closed
    del y_memmap
    del r_memmap
    gc.collect()

    return y_path, r_path, u_adjusted, mode_metadata, val_report
````

````python
# Task 12: Generate spectral interpretation and distributional diagnostics

# ==============================================================================
# Task 12: Generate spectral interpretation and distributional diagnostics
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 1: Produce eigenvalue density and verify market-mode structure
# -------------------------------------------------------------------------------------------------------------------------------

def generate_eigenvalue_diagnostics(
    eigenvalues: np.ndarray,
    eigenvectors: np.ndarray
) -> Dict[str, Any]:
    """
    Computes the eigenvalue density histogram and verifies the uniformity
    of the first eigenvector (the market mode).

    Args:
        eigenvalues (np.ndarray): The sorted eigenvalues (descending).
        eigenvectors (np.ndarray): The sign-adjusted eigenvector matrix U.

    Returns:
        Dict[str, Any]: Dictionary containing histogram data and uniformity metrics.
    """
    # 1. Eigenvalue Density Histogram
    max_eig = float(eigenvalues[0])
    # Use 100 bins from 0 to max_eig + 1 to capture both bulk and outliers
    bins = np.linspace(0, max_eig + 1.0, 101)
    counts, edges = np.histogram(eigenvalues, bins=bins, density=True)

    # 2. Market-Mode Uniformity Check (Mode 1 -> index 0)
    u1 = eigenvectors[:, 0]
    abs_u1 = np.abs(u1)

    # Coefficient of Variation of absolute loadings
    cv_1 = float(np.std(abs_u1) / np.mean(abs_u1))

    # Sign coherence (fraction of entries with the dominant sign)
    # Since we applied a positive-sum sign convention in Task 11, the dominant sign should be positive.
    positive_fraction = float(np.sum(u1 > 0) / len(u1))

    diagnostics = {
        "eigenvalue_histogram": {
            "counts": counts.tolist(),
            "bin_edges": edges.tolist()
        },
        "market_mode_uniformity": {
            "cv_absolute_loadings": cv_1,
            "positive_sign_fraction": positive_fraction,
            "is_market_like": cv_1 < 1.0 and positive_fraction > 0.8
        }
    }

    return diagnostics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 2: Extract eigenvector loadings and compute sector concentration
# -------------------------------------------------------------------------------------------------------------------------------

def compute_sector_concentration(
    eigenvectors: np.ndarray,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Extracts the loadings for the first three modes and computes their
    concentration across GICS sectors (if a sector map is provided).

    Equation: w_{k,s} = \sum_{l \in s} U_{lk}^2

    Args:
        eigenvectors (np.ndarray): The sign-adjusted eigenvector matrix U.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Dictionary containing raw loadings and sector concentrations.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    tickers: tuple = config["asset_universe_exact_appendix_order"]

    # Check if GICS sector map is provided in the manifest (auxiliary data)
    # If not, we gracefully degrade to returning just the raw loadings.
    sector_map: Optional[Dict[str, str]] = manifest.get("auxiliary_data", {}).get("gics_sector_map", None)

    results: Dict[str, Any] = {
        "mode_loadings": {},
        "sector_concentrations": {},
        "sector_map_available": sector_map is not None
    }

    # Process first 3 modes
    for k in range(3):
        mode_name = f"mode_{k+1}"
        u_k = eigenvectors[:, k]

        # Store raw loadings mapped to tickers
        results["mode_loadings"][mode_name] = {ticker: float(val) for ticker, val in zip(tickers, u_k)}

        if sector_map:
            # Compute squared loadings
            sq_loadings = u_k ** 2

            # Aggregate by sector
            sector_sums: Dict[str, float] = {}
            for i, ticker in enumerate(tickers):
                sector = sector_map.get(ticker, "Unknown")
                sector_sums[sector] = sector_sums.get(sector, 0.0) + sq_loadings[i]

            # Normalize (though sum should be 1.0 due to orthonormality)
            total_sum = sum(sector_sums.values())
            normalized_sums = {s: v / total_sum for s, v in sector_sums.items()}

            results["sector_concentrations"][mode_name] = normalized_sums

    return results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 3: Compute tail densities for Y, R, and the aggregated bulk
# -------------------------------------------------------------------------------------------------------------------------------

def compute_tail_densities(
    y_memmap: np.memmap,
    r_memmap: np.memmap,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Computes empirical probability density histograms for the tails of the
    unscaled modes (Y), whitened modes (R), and the aggregated bulk.

    Uses a memory-efficient chunked accumulation strategy for the aggregated density.

    Args:
        y_memmap (np.memmap): The unscaled rotated returns Y (read-only).
        r_memmap (np.memmap): The whitened modes R (read-only).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Dictionary containing histogram data and tail-weight metrics.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]

    # Define fixed bin edges for the tail region (absolute values)
    # We focus on the range [0, 50] with 500 bins (width 0.1)
    bin_edges = np.linspace(0, 50, 501)
    bin_width = bin_edges[1] - bin_edges[0]

    results: Dict[str, Any] = {
        "bin_edges": bin_edges.tolist(),
        "y_densities": {},
        "r_densities": {},
        "tail_weight_fractions_5_to_20": {}
    }

    # Helper function to compute density and tail fraction
    def _process_row(row_data: np.ndarray) -> Tuple[np.ndarray, float]:
        abs_data = np.abs(row_data)
        counts, _ = np.histogram(abs_data, bins=bin_edges)
        density = counts / (len(row_data) * bin_width)

        # Fraction of observations in [5, 20]
        mask = (abs_data >= 5.0) & (abs_data <= 20.0)
        tail_frac = float(np.sum(mask) / len(row_data))

        return density, tail_frac

    # 1. Process first 5 modes for Y and R
    for k in range(5):
        mode_name = f"mode_{k+1}"

        # Y density
        y_row = y_memmap[k, :]
        y_dens, _ = _process_row(y_row)
        results["y_densities"][mode_name] = y_dens.tolist()

        # R density
        r_row = r_memmap[k, :]
        r_dens, r_tail_frac = _process_row(r_row)
        results["r_densities"][mode_name] = r_dens.tolist()
        results["tail_weight_fractions_5_to_20"][mode_name] = r_tail_frac

    # 2. Compute Aggregated Density for R (pooling all K modes)
    aggregated_counts = np.zeros(len(bin_edges) - 1, dtype=np.float64)
    aggregated_tail_count = 0

    for k in range(k_assets):
        r_row = r_memmap[k, :]
        abs_data = np.abs(r_row)

        counts, _ = np.histogram(abs_data, bins=bin_edges)
        aggregated_counts += counts

        mask = (abs_data >= 5.0) & (abs_data <= 20.0)
        aggregated_tail_count += np.sum(mask)

    total_elements = k_assets * t_total
    aggregated_density = aggregated_counts / (total_elements * bin_width)
    aggregated_tail_frac = float(aggregated_tail_count / total_elements)

    results["r_densities"]["aggregated"] = aggregated_density.tolist()
    results["tail_weight_fractions_5_to_20"]["aggregated"] = aggregated_tail_frac

    return results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def generate_spectral_diagnostics(
    eigenvalues: np.ndarray,
    eigenvectors: np.ndarray,
    y_memmap_path: str,
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the generation of spectral interpretation and distributional diagnostics.

    Computes eigenvalue densities, verifies market/sector mode structures, and
    calculates empirical tail densities for the unscaled, whitened, and aggregated modes.

    Args:
        eigenvalues (np.ndarray): The validated eigenvalues.
        eigenvectors (np.ndarray): The sign-adjusted eigenvector matrix U.
        y_memmap_path (str): Path to the unscaled rotated returns Y memmap.
        r_memmap_path (str): Path to the whitened modes R memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: A comprehensive dictionary containing all spectral diagnostics.

    Raises:
        Exception: Bubbles up any errors encountered during computation.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    # Step 1: Eigenvalue diagnostics
    eig_diag = generate_eigenvalue_diagnostics(eigenvalues, eigenvectors)

    # Step 2: Sector concentration
    sector_diag = compute_sector_concentration(eigenvectors, manifest)

    # Step 3: Tail densities
    # Open memmaps in read-only mode
    y_memmap = np.memmap(y_memmap_path, dtype=np.float64, mode='r', shape=shape_2d)
    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=shape_2d)

    density_diag = compute_tail_densities(y_memmap, r_memmap, manifest)

    # Clean up memmap references
    del y_memmap
    del r_memmap
    gc.collect()

    # Combine all diagnostics into a single serializable artifact
    spectral_diagnostics = {
        "eigenvalue_analysis": eig_diag,
        "eigenvector_analysis": sector_diag,
        "distributional_analysis": density_diag
    }

    return spectral_diagnostics
````

````python
# Task 13: Extract static peaks-over-threshold exceedances

# ==============================================================================
# Task 13: Extract static peaks-over-threshold exceedances
# ==============================================================================

# Define a NamedTuple for clear dictionary keying
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# Define a Dataclass-like dictionary structure for exceedance results
# (Using TypedDict in modern Python, but standard Dict here for broad compatibility)
# ExceedanceObject = Dict[str, Any] containing:
# - u_alpha: float
# - x_exceedances: np.ndarray (float64)
# - t_positions: np.ndarray (int64)
# - n: int
# - is_sufficient: bool

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 1: Define the positive and negative tail analysis series
# -------------------------------------------------------------------------------------------------------------------------------

def extract_analysis_series(
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Dict[SeriesKey, np.ndarray]:
    """
    Extracts the first 5 modes from the whitened mode matrix R and defines
    the positive (X = R) and negative (X = -R) analysis series for each.

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap (K x T).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[SeriesKey, np.ndarray]: A dictionary mapping (mode, tail) to the
                                     contiguous 1D numpy array of length T.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    shape_2d = (k_assets, t_total)

    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["static_evt"]
    modes_to_analyze: tuple = evt_config.get("modes_for_evt_analysis", (1, 2, 3, 4, 5))

    # Open R in read-only mode
    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=shape_2d)

    analysis_series: Dict[SeriesKey, np.ndarray] = {}

    for mode_k in modes_to_analyze:
        # mode_k is 1-indexed. Array row is 0-indexed.
        row_idx = mode_k - 1

        # Extract row and force contiguous memory layout via .copy()
        # This reads ~43 MB into RAM per row.
        r_row: np.ndarray = r_memmap[row_idx, :].copy()

        # Positive tail series: X^+ = R
        key_pos = SeriesKey(mode=mode_k, tail="positive")
        analysis_series[key_pos] = r_row

        # Negative tail series: X^- = -R
        key_neg = SeriesKey(mode=mode_k, tail="negative")
        analysis_series[key_neg] = -r_row

    # Explicitly delete memmap reference
    del r_memmap
    gc.collect()

    return analysis_series

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 2: Compute the static threshold u_alpha for each quantile
# -------------------------------------------------------------------------------------------------------------------------------

def compute_static_thresholds(
    analysis_series: Dict[SeriesKey, np.ndarray],
    manifest: MappingProxyType
) -> Dict[SeriesKey, Dict[float, float]]:
    """
    Computes the static threshold u_alpha for each analysis series and each
    configured tail quantile alpha using type-7 linear interpolation.

    Equation: u_alpha = Q_{1-alpha}(X)

    Args:
        analysis_series (Dict[SeriesKey, np.ndarray]): The extracted 1D series.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[SeriesKey, Dict[float, float]]: Nested dictionary mapping
                                             (mode, tail) -> {alpha: threshold}.
    """
    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["static_evt"]
    alpha_grid: tuple = evt_config["tail_quantile_grid_alpha"]

    # Calculate the quantile levels (1 - alpha)
    quantile_levels = [1.0 - a for a in alpha_grid]

    thresholds: Dict[SeriesKey, Dict[float, float]] = {}

    for key, series in analysis_series.items():
        # Compute all quantiles for this series in a single highly-optimized pass
        # method='linear' corresponds to type-7 interpolation
        try:
            # numpy >= 1.22
            u_vals = np.quantile(series, q=quantile_levels, method='linear')
        except TypeError:
            # numpy < 1.22
            u_vals = np.quantile(series, q=quantile_levels, interpolation='linear')

        # Map back to alpha values
        thresholds[key] = {alpha: float(u) for alpha, u in zip(alpha_grid, u_vals)}

    return thresholds

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 3: Extract exceedance magnitudes, times, and counts
# -------------------------------------------------------------------------------------------------------------------------------

def extract_exceedances(
    analysis_series: Dict[SeriesKey, np.ndarray],
    thresholds: Dict[SeriesKey, Dict[float, float]],
    manifest: MappingProxyType
) -> Dict[SeriesKey, Dict[float, Dict[str, Any]]]:
    """
    Extracts the exceedance magnitudes and global time positions for each
    (mode, tail, alpha) triplet using strict inequality (X > u).

    No declustering is applied, strictly adhering to the manuscript.

    Args:
        analysis_series (Dict[SeriesKey, np.ndarray]): The extracted 1D series.
        thresholds (Dict[SeriesKey, Dict[float, float]]): The computed thresholds.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[SeriesKey, Dict[float, Dict[str, Any]]]: Nested dictionary containing
                                                      the exceedance objects.

    Raises:
        RuntimeError: If any extracted exceedance magnitude is not strictly positive.
    """
    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["static_evt"]
    # Default to 30 if not explicitly set in manifest
    min_exceedance_count: int = evt_config.get("min_exceedance_count", 30)

    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]] = {}

    for key, series in analysis_series.items():
        exceedance_objects[key] = {}

        for alpha, u_alpha in thresholds[key].items():
            # Strict inequality: X > u
            exceed_mask: np.ndarray = series > u_alpha

            # Extract global time positions (zero-based indices)
            # np.where returns a tuple, we take the first element
            t_positions: np.ndarray = np.where(exceed_mask)[0].astype(np.int64)

            # Extract magnitudes: x_i = X(t_i) - u
            x_exceedances: np.ndarray = series[t_positions] - u_alpha

            n_count: int = len(x_exceedances)

            # Strict validation
            if n_count > 0 and not (x_exceedances > 0).all():
                raise RuntimeError(f"Extracted exceedances for {key}, alpha={alpha} contain non-positive values.")

            is_sufficient: bool = n_count >= min_exceedance_count

            exceedance_objects[key][alpha] = {
                "u_alpha": u_alpha,
                "x_exceedances": x_exceedances,
                "t_positions": t_positions,
                "n": n_count,
                "is_sufficient": is_sufficient
            }

    return exceedance_objects

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def extract_static_pot_exceedances(
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Dict[SeriesKey, Dict[float, Dict[str, Any]]]:
    """
    Orchestrates the extraction of static Peaks-Over-Threshold (POT) exceedances.

    Defines positive and negative tail series for the top 5 modes, computes
    static quantile thresholds, and extracts exceedance magnitudes and times
    without declustering.

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[SeriesKey, Dict[float, Dict[str, Any]]]: A nested dictionary mapping
            (mode, tail) -> alpha -> exceedance_object.

    Raises:
        Exception: Bubbles up any RuntimeError encountered during extraction.
    """
    # Step 1: Extract analysis series
    analysis_series = extract_analysis_series(r_memmap_path, manifest)

    # Step 2: Compute thresholds
    thresholds = compute_static_thresholds(analysis_series, manifest)

    # Step 3: Extract exceedances
    exceedance_objects = extract_exceedances(analysis_series, thresholds, manifest)

    # Free the large analysis series arrays from memory
    del analysis_series
    gc.collect()

    return exceedance_objects
````

````python
# Task 14: Fit the generalized Pareto distribution by maximum likelihood

# ==============================================================================
# Task 14: Fit the generalized Pareto distribution by maximum likelihood
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 1: Define the GPD log-likelihood function
# -------------------------------------------------------------------------------------------------------------------------------

def gpd_negative_log_likelihood(params: np.ndarray, x: np.ndarray) -> float:
    """
    Computes the negative log-likelihood of the Generalized Pareto Distribution.

    Uses the reparameterization \sigma = exp(\eta) to enforce \sigma > 0.
    Enforces the support constraint 1 + \gamma * x / \sigma > 0 via a large penalty.
    Handles the \gamma \approx 0 limit smoothly to avoid numerical instability.

    Args:
        params (np.ndarray): 1D array containing [\gamma, \eta].
        x (np.ndarray): 1D array of strictly positive exceedance magnitudes.

    Returns:
        float: The negative log-likelihood value (scalar).
    """
    gamma: float = params[0]
    eta: float = params[1]

    # Reparameterize to ensure strictly positive scale
    # Use np.clip to prevent overflow in exp if optimizer explores extreme eta
    sigma: float = np.exp(np.clip(eta, -100, 100))

    n: int = len(x)

    # Check support constraint: 1 + gamma * x / sigma > 0
    # If gamma >= 0, this is always true since x > 0 and sigma > 0.
    # If gamma < 0, we must check the maximum x.
    if gamma < 0:
        max_x = np.max(x)
        if 1.0 + gamma * max_x / sigma <= 0:
            # Return a large penalty to push optimizer away from infeasible region
            return 1e20

    # Compute negative log-likelihood
    if abs(gamma) < 1e-8:
        # Exponential limit (\gamma -> 0)
        # l(\sigma) = -n * \ln(\sigma) - \sum(x_i) / \sigma
        nll = n * np.log(sigma) + np.sum(x) / sigma
    else:
        # General GPD formula
        # l(\gamma, \sigma) = -n * \ln(\sigma) - (1 + 1/\gamma) * \sum \ln(1 + \gamma * x_i / \sigma)
        # Use log1p for numerical stability when \gamma * x / \sigma is small
        z = gamma * x / sigma
        sum_log1p = np.sum(np.log1p(z))
        nll = n * np.log(sigma) + (1.0 + 1.0 / gamma) * sum_log1p

    # Return float64 scalar
    return float(nll)

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 2: Optimize the negative log-likelihood using L-BFGS-B
# -------------------------------------------------------------------------------------------------------------------------------

def optimize_gpd_mle(
    x_exceedances: np.ndarray,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Optimizes the GPD negative log-likelihood using the L-BFGS-B algorithm.

    Initializes \gamma = 0.1 and \sigma = mean(x).

    Args:
        x_exceedances (np.ndarray): 1D array of exceedance magnitudes.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Dictionary containing the fitted parameters and optimizer metadata.
    """
    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["static_evt"]

    # Extract optimizer settings (with fallbacks to defaults if unspecified)
    maxiter: int = evt_config.get("gpd_optimizer_maxiter", 10000)
    ftol: float = evt_config.get("gpd_optimizer_ftol", 1e-8)
    gtol: float = evt_config.get("gpd_optimizer_gtol", 1e-8)
    init_gamma: float = evt_config.get("gpd_initial_gamma", 0.1)

    # Initialize parameters
    # sigma_0 = sample mean excess
    mean_x = float(np.mean(x_exceedances))
    init_sigma = mean_x if mean_x > 0 else 1.0
    init_eta = np.log(init_sigma)

    x0 = np.array([init_gamma, init_eta], dtype=np.float64)

    # Define the objective function bound to the data
    def objective(params):
        return gpd_negative_log_likelihood(params, x_exceedances)

    # Run L-BFGS-B optimization
    result = scipy.optimize.minimize(
        fun=objective,
        x0=x0,
        method='L-BFGS-B',
        options={
            'maxiter': maxiter,
            'ftol': ftol,
            'gtol': gtol,
            'disp': False
        }
    )

    # Extract results
    gamma_hat = float(result.x[0])
    eta_hat = float(result.x[1])
    sigma_hat = float(np.exp(np.clip(eta_hat, -100, 100)))

    # Compute gradient norm
    grad_norm = float(np.linalg.norm(result.jac)) if hasattr(result, 'jac') else np.nan

    # Check if support constraint is nearly binding
    support_binding = False
    if gamma_hat < 0:
        min_support_val = 1.0 + gamma_hat * np.max(x_exceedances) / sigma_hat
        if min_support_val < 1e-6:
            support_binding = True

    fit_results = {
        "gamma_hat": gamma_hat,
        "sigma_hat": sigma_hat,
        "nll_achieved": float(result.fun),
        "converged": bool(result.success),
        "n_iterations": int(result.nit),
        "gradient_norm": grad_norm,
        "support_binding": support_binding,
        "optimizer_message": str(result.message)
    }

    return fit_results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 3: Record fitted parameters and construct results table
# -------------------------------------------------------------------------------------------------------------------------------

def construct_gpd_fit_table(
    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]],
    manifest: MappingProxyType
) -> pd.DataFrame:
    """
    Iterates over all valid (mode, tail, alpha) triplets, performs the MLE fit,
    evaluates reliability, and constructs a comprehensive results DataFrame.

    Args:
        exceedance_objects (Dict[SeriesKey, Dict[float, Dict[str, Any]]]): The extracted exceedances.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        pd.DataFrame: A flat table containing all fit results and metadata.
    """
    records: List[Dict[str, Any]] = []

    for key, alpha_dict in exceedance_objects.items():
        for alpha, exc_obj in alpha_dict.items():

            n_count = exc_obj["n"]
            is_sufficient = exc_obj["is_sufficient"]
            u_alpha = exc_obj["u_alpha"]

            # Base record
            record = {
                "mode": key.mode,
                "tail": key.tail,
                "alpha": alpha,
                "threshold_u": u_alpha,
                "n_exceedances": n_count,
                "is_sufficient": is_sufficient
            }

            if is_sufficient:
                x_data = exc_obj["x_exceedances"]

                # Perform optimization
                fit_res = optimize_gpd_mle(x_data, manifest)

                # Evaluate strict reliability
                # Reliable if: converged AND gradient is small AND sufficient data
                is_reliable = (
                    fit_res["converged"] and
                    (not np.isnan(fit_res["gradient_norm"])) and
                    (fit_res["gradient_norm"] < 1e-4)
                )

                record.update(fit_res)
                record["is_reliable"] = is_reliable

            else:
                # Insufficient data, populate with NaNs
                null_res = {
                    "gamma_hat": np.nan,
                    "sigma_hat": np.nan,
                    "nll_achieved": np.nan,
                    "converged": False,
                    "n_iterations": 0,
                    "gradient_norm": np.nan,
                    "support_binding": False,
                    "optimizer_message": "Insufficient data",
                    "is_reliable": False
                }
                record.update(null_res)

            records.append(record)

    # Convert to DataFrame
    df_results = pd.DataFrame(records)

    # Sort for consistent downstream processing
    df_results.sort_values(by=["mode", "tail", "alpha"], ascending=[True, True, False], inplace=True)
    df_results.reset_index(drop=True, inplace=True)

    return df_results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def fit_gpd_mle(
    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]],
    manifest: MappingProxyType
) -> pd.DataFrame:
    """
    Orchestrates the Maximum Likelihood Estimation of the Generalized Pareto Distribution.

    Iterates through all extracted exceedance sets, applies the L-BFGS-B optimizer
    to the reparameterized log-likelihood, enforces support constraints, and
    compiles a rigorous diagnostic table of the fits.

    Args:
        exceedance_objects (Dict[SeriesKey, Dict[float, Dict[str, Any]]]): The exceedance data.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        pd.DataFrame: A comprehensive table of GPD parameters and convergence metrics.

    Raises:
        Exception: Bubbles up any unexpected errors during the optimization loop.
    """
    # The construction function handles the iteration, optimization, and aggregation
    gpd_fit_table = construct_gpd_fit_table(exceedance_objects, manifest)

    # Global diagnostic: check if an unacceptable number of fits failed
    total_fits = len(gpd_fit_table)
    failed_fits = len(gpd_fit_table[~gpd_fit_table["is_reliable"]])
    failure_rate = failed_fits / total_fits if total_fits > 0 else 1.0

    if failure_rate > 0.20:
        # Log a severe warning, but do not halt, as extreme quantiles naturally fail
        logging.warning(f"High GPD fit failure rate: {failure_rate:.1%} ({failed_fits}/{total_fits} fits unreliable). Check data or optimizer settings.")

    return gpd_fit_table
````

````python
# Task 15: Compute GPD asymptotic standard errors and NRMSD goodness-of-fit

# ==============================================================================
# Task 15: Compute GPD asymptotic standard errors and NRMSD goodness-of-fit
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 1: Compute asymptotic standard errors from Fisher information
# -------------------------------------------------------------------------------------------------------------------------------

def compute_asymptotic_standard_errors(gpd_fit_table: pd.DataFrame) -> pd.DataFrame:
    """
    Computes asymptotic standard errors for \hat\gamma and \hat\sigma using the
    Fisher information matrix formulas (Eq. 17).

    Enforces the regularity condition \hat\gamma > -0.5. If violated, standard
    errors are undefined and set to NaN.

    Args:
        gpd_fit_table (pd.DataFrame): The results table from Task 14.

    Returns:
        pd.DataFrame: The table augmented with standard errors and 95% CIs.
    """
    df = gpd_fit_table.copy()

    # Initialize new columns with NaNs
    df["se_gamma"] = np.nan
    df["se_sigma"] = np.nan
    df["ci_lower_gamma"] = np.nan
    df["ci_upper_gamma"] = np.nan
    df["ci_lower_sigma"] = np.nan
    df["ci_upper_sigma"] = np.nan

    # Filter for reliable fits that meet the regularity condition
    valid_mask = df["is_reliable"] & (df["gamma_hat"] > -0.5)

    # Extract vectors for vectorized computation
    gamma = df.loc[valid_mask, "gamma_hat"].values
    sigma = df.loc[valid_mask, "sigma_hat"].values
    n = df.loc[valid_mask, "n_exceedances"].values

    # Compute standard errors
    # se(\hat\gamma) = \sqrt{(1 + \hat\gamma)^2 / n}
    se_gamma = np.sqrt((1.0 + gamma)**2 / n)

    # se(\hat\sigma) = \sqrt{2 \hat\sigma^2 (1 + \hat\gamma) / n}
    se_sigma = np.sqrt(2.0 * (sigma**2) * (1.0 + gamma) / n)

    # Assign back to DataFrame
    df.loc[valid_mask, "se_gamma"] = se_gamma
    df.loc[valid_mask, "se_sigma"] = se_sigma

    # Compute 95% Confidence Intervals (approximate, assuming asymptotic normality)
    z_95 = 1.95996
    df.loc[valid_mask, "ci_lower_gamma"] = gamma - z_95 * se_gamma
    df.loc[valid_mask, "ci_upper_gamma"] = gamma + z_95 * se_gamma
    df.loc[valid_mask, "ci_lower_sigma"] = sigma - z_95 * se_sigma
    df.loc[valid_mask, "ci_upper_sigma"] = sigma + z_95 * se_sigma

    # Log warning for fits that violate regularity condition
    invalid_regularity = df["is_reliable"] & (df["gamma_hat"] <= -0.5)
    if invalid_regularity.any():
        import logging
        logging.warning(f"{invalid_regularity.sum()} fits violated the MLE regularity condition (\hat\gamma > -0.5). Standard errors set to NaN.")

    return df

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 2: Construct Q-Q diagnostic data
# -------------------------------------------------------------------------------------------------------------------------------

def construct_qq_diagnostic_data(
    gpd_fit_table: pd.DataFrame,
    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]]
) -> Dict[Tuple[int, str, float], Dict[str, np.ndarray]]:
    """
    Constructs Quantile-Quantile (Q-Q) diagnostic data for each reliable fit.

    Computes empirical quantiles using the Weibull plotting position i/(n+1)
    and theoretical quantiles using the inverse GPD CDF.

    Args:
        gpd_fit_table (pd.DataFrame): The augmented results table.
        exceedance_objects (Dict): The raw exceedance data.

    Returns:
        Dict: Mapping (mode, tail, alpha) -> {"empirical": array, "theoretical": array}
    """
    qq_data: Dict[Tuple[int, str, float], Dict[str, np.ndarray]] = {}

    # Iterate only over reliable fits
    reliable_fits = gpd_fit_table[gpd_fit_table["is_reliable"]]

    for _, row in reliable_fits.iterrows():
        mode = int(row["mode"])
        tail = str(row["tail"])
        alpha = float(row["alpha"])
        gamma_hat = float(row["gamma_hat"])
        sigma_hat = float(row["sigma_hat"])

        key = SeriesKey(mode=mode, tail=tail)
        x_raw = exceedance_objects[key][alpha]["x_exceedances"]
        n = len(x_raw)

        # Sort empirical exceedances
        x_sorted = np.sort(x_raw)

        # Compute plotting positions: p_i = i / (n + 1)
        p_i = np.arange(1, n + 1, dtype=np.float64) / (n + 1.0)

        # Compute theoretical GPD quantiles
        if abs(gamma_hat) < 1e-8:
            # Exponential limit
            q_gpd = -sigma_hat * np.log(1.0 - p_i)
        else:
            # General GPD inverse CDF
            q_gpd = (sigma_hat / gamma_hat) * ((1.0 - p_i)**(-gamma_hat) - 1.0)

        qq_data[(mode, tail, alpha)] = {
            "empirical": x_sorted,
            "theoretical": q_gpd
        }

    return qq_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 3: Compute NRMSD and identify credible threshold regions
# -------------------------------------------------------------------------------------------------------------------------------

def compute_nrmsd_and_credible_regions(
    gpd_fit_table: pd.DataFrame,
    qq_data: Dict[Tuple[int, str, float], Dict[str, np.ndarray]]
) -> Tuple[pd.DataFrame, Dict[str, Any]]:
    """
    Computes the Normalized Root Mean Square Deviation (NRMSD) for each fit.
    Identifies the credible threshold region for each (mode, tail) pair based
    on \hat\gamma stability and NRMSD minimization.

    Equation: \rho = (1 / (max(x) - min(x))) * sqrt((1/n) * sum((q_i - x_i)^2))

    Args:
        gpd_fit_table (pd.DataFrame): The table with standard errors.
        qq_data (Dict): The Q-Q diagnostic arrays.

    Returns:
        Tuple[pd.DataFrame, Dict[str, Any]]:
            - The final augmented DataFrame including 'rho'.
            - Metadata dictionary defining the credible regions.
    """
    df = gpd_fit_table.copy()
    df["rho"] = np.nan

    # 1. Compute NRMSD (\rho)
    for idx, row in df.iterrows():
        if not row["is_reliable"]:
            continue

        mode = int(row["mode"])
        tail = str(row["tail"])
        alpha = float(row["alpha"])

        qq = qq_data.get((mode, tail, alpha))
        if qq is None:
            continue

        x_emp = qq["empirical"]
        q_theo = qq["theoretical"]

        range_x = x_emp[-1] - x_emp[0]
        if range_x <= 0:
            continue # Avoid division by zero

        rmsd = np.sqrt(np.mean((q_theo - x_emp)**2))
        rho = rmsd / range_x

        df.at[idx, "rho"] = rho

    # 2. Identify Credible Threshold Regions
    credible_regions: Dict[str, Any] = {}

    # Group by mode and tail
    grouped = df[df["is_reliable"]].groupby(["mode", "tail"])

    for (mode, tail), group in grouped:
        # Sort by alpha descending (which means threshold u is increasing)
        group_sorted = group.sort_values("alpha", ascending=False).reset_index(drop=True)

        if len(group_sorted) < 3:
            credible_regions[f"mode_{mode}_{tail}"] = {"status": "insufficient_data"}
            continue

        best_window_idx = -1
        min_rho_in_stable_window = float('inf')
        best_alpha = None
        best_gamma = None

        # Slide a window of size 3 across the alphas
        window_size = 3
        for i in range(len(group_sorted) - window_size + 1):
            window = group_sorted.iloc[i:i+window_size]

            gamma_range = window["gamma_hat"].max() - window["gamma_hat"].min()
            mean_se = window["se_gamma"].mean()

            # Stability criterion: range of gamma < average standard error
            if gamma_range < mean_se:
                # Find the alpha in this window that minimizes rho
                min_rho_idx = window["rho"].idxmin()
                min_rho_val = window.loc[min_rho_idx, "rho"]

                if min_rho_val < min_rho_in_stable_window:
                    min_rho_in_stable_window = min_rho_val
                    best_window_idx = i
                    best_alpha = window.loc[min_rho_idx, "alpha"]
                    best_gamma = window.loc[min_rho_idx, "gamma_hat"]

        if best_window_idx != -1:
            window = group_sorted.iloc[best_window_idx:best_window_idx+window_size]
            credible_regions[f"mode_{mode}_{tail}"] = {
                "status": "identified",
                "alpha_range": [float(window["alpha"].max()), float(window["alpha"].min())],
                "optimal_alpha": float(best_alpha),
                "optimal_gamma": float(best_gamma),
                "min_rho": float(min_rho_in_stable_window)
            }
        else:
            credible_regions[f"mode_{mode}_{tail}"] = {"status": "no_stable_region_found"}

    return df, credible_regions

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def evaluate_gpd_fit_quality(
    gpd_fit_table: pd.DataFrame,
    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]],
    manifest: MappingProxyType
) -> Tuple[pd.DataFrame, Dict[Tuple[int, str, float], Dict[str, np.ndarray]], Dict[str, Any]]:
    """
    Orchestrates the evaluation of GPD fit quality.

    Computes asymptotic standard errors, generates Q-Q diagnostic arrays,
    calculates the NRMSD goodness-of-fit metric, and programmatically identifies
    credible threshold regions based on parameter stability.

    Args:
        gpd_fit_table (pd.DataFrame): The raw fit results from Task 14.
        exceedance_objects (Dict): The raw exceedance data from Task 13.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, Dict, Dict[str, Any]]:
            - The fully augmented results DataFrame.
            - The Q-Q diagnostic data dictionary.
            - The credible regions metadata dictionary.
    """
    # Step 1: Compute standard errors
    df_with_se = compute_asymptotic_standard_errors(gpd_fit_table)

    # Step 2: Construct Q-Q data
    qq_data = construct_qq_diagnostic_data(df_with_se, exceedance_objects)

    # Step 3: Compute NRMSD and identify credible regions
    final_fit_table, credible_regions = compute_nrmsd_and_credible_regions(df_with_se, qq_data)

    return final_fit_table, qq_data, credible_regions
````

````python
# Task 16: Estimate the extremal index via the Ferro–Segers method

# ==============================================================================
# Task 16: Estimate the extremal index via the Ferro–Segers method
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 1: Compute inter-exceedance times
# -------------------------------------------------------------------------------------------------------------------------------

def compute_inter_exceedance_times(t_positions: np.ndarray) -> Tuple[np.ndarray, bool, str]:
    """
    Computes the inter-exceedance times (\tau_i) from the sorted global time positions.

    Args:
        t_positions (np.ndarray): 1D array of zero-based global time indices.

    Returns:
        Tuple[np.ndarray, bool, str]:
            - 1D float64 array of inter-exceedance times.
            - Boolean flag indicating if estimation is possible (n >= 2).
            - Status message.
    """
    n = len(t_positions)

    if n < 2:
        return np.array([], dtype=np.float64), False, "Fewer than 2 exceedances; cannot compute tau."

    # Strict validation: positions must be strictly monotonically increasing
    if not np.all(np.diff(t_positions) > 0):
        raise ValueError("Exceedance time positions are not strictly monotonically increasing.")

    # Compute tau_i = t_{i+1} - t_i
    # Cast to float64 immediately to prevent integer overflow during moment calculations
    tau = np.diff(t_positions).astype(np.float64)

    status = "Reliable" if n >= 30 else "Unreliable (n < 30)"
    return tau, True, status

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 2: Apply the Ferro-Segers estimator
# -------------------------------------------------------------------------------------------------------------------------------

def apply_ferro_segers_estimator(tau: np.ndarray) -> Dict[str, Any]:
    """
    Applies the Ferro-Segers estimator to compute the extremal index \hat\Theta.

    Equation (15):
    If max(\tau) > 2: min( 2*<(\tau-1)>^2 / <(\tau-1)(\tau-2)>, 1 )
    If max(\tau) <= 2: min( 2*<\tau>^2 / <\tau^2>, 1 )

    Args:
        tau (np.ndarray): 1D float64 array of inter-exceedance times.

    Returns:
        Dict[str, Any]: Dictionary containing \hat\Theta and associated metadata.
    """
    if len(tau) == 0:
        return {
            "theta_hat": np.nan,
            "theta_branch": "none",
            "tau_max": np.nan,
            "tau_mean": np.nan,
            "n_tau": 0,
            "mean_cluster_size": np.nan
        }

    tau_max = float(np.max(tau))
    tau_mean = float(np.mean(tau))
    n_tau = len(tau)

    if tau_max > 2.0:
        branch = "max_tau_gt_2"
        tau_m1 = tau - 1.0
        tau_m2 = tau - 2.0

        numerator = 2.0 * (np.mean(tau_m1) ** 2)
        denominator = np.mean(tau_m1 * tau_m2)

        if denominator == 0.0:
            # Extreme clustering case where all tau are 1 or 2 such that mean product is 0
            theta_hat = 0.0
        elif denominator < 0.0:
            # Estimator breakdown
            theta_hat = np.nan
        else:
            ratio = numerator / denominator
            theta_hat = min(ratio, 1.0)

    else:
        branch = "max_tau_leq_2"
        numerator = 2.0 * (np.mean(tau) ** 2)
        denominator = np.mean(tau ** 2)

        # Denominator cannot be zero since tau >= 1
        ratio = numerator / denominator
        theta_hat = min(ratio, 1.0)

    # Compute mean cluster size (1 / \hat\Theta)
    if np.isnan(theta_hat):
        mcs = np.nan
    elif theta_hat == 0.0:
        mcs = np.inf
    else:
        mcs = 1.0 / theta_hat

    return {
        "theta_hat": float(theta_hat),
        "theta_branch": branch,
        "tau_max": tau_max,
        "tau_mean": tau_mean,
        "n_tau": n_tau,
        "mean_cluster_size": float(mcs)
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 3: Store results with interpretation metadata
# -------------------------------------------------------------------------------------------------------------------------------

def augment_table_with_extremal_index(
    gpd_fit_table: pd.DataFrame,
    theta_results: Dict[Tuple[int, str, float], Dict[str, Any]]
) -> pd.DataFrame:
    """
    Merges the computed extremal index results into the main GPD fit DataFrame.

    Args:
        gpd_fit_table (pd.DataFrame): The results table from Task 15.
        theta_results (Dict): Mapping of (mode, tail, alpha) to Ferro-Segers outputs.

    Returns:
        pd.DataFrame: The fully augmented table including \hat\Theta.
    """
    df = gpd_fit_table.copy()

    # Initialize new columns
    new_cols = ["theta_hat", "theta_branch", "tau_max", "tau_mean", "n_tau", "is_estimable", "mean_cluster_size"]
    for col in new_cols:
        if col in ["theta_branch"]:
            df[col] = "none"
        elif col in ["is_estimable"]:
            df[col] = False
        else:
            df[col] = np.nan

    # Populate columns
    for idx, row in df.iterrows():
        mode = int(row["mode"])
        tail = str(row["tail"])
        alpha = float(row["alpha"])

        key = (mode, tail, alpha)
        if key in theta_results:
            res = theta_results[key]
            df.at[idx, "theta_hat"] = res["theta_hat"]
            df.at[idx, "theta_branch"] = res["theta_branch"]
            df.at[idx, "tau_max"] = res["tau_max"]
            df.at[idx, "tau_mean"] = res["tau_mean"]
            df.at[idx, "n_tau"] = res["n_tau"]
            df.at[idx, "is_estimable"] = res["is_estimable"]
            df.at[idx, "mean_cluster_size"] = res["mean_cluster_size"]

    return df

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def estimate_extremal_index(
    exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]],
    gpd_fit_table: pd.DataFrame,
    manifest: MappingProxyType
) -> pd.DataFrame:
    """
    Orchestrates the estimation of the extremal index \hat\Theta.

    Extracts inter-exceedance times, applies the Ferro-Segers estimator,
    handles edge cases (extreme clustering or insufficient data), and
    merges the results into the main analytical table.

    Args:
        exceedance_objects (Dict): The raw exceedance data from Task 13.
        gpd_fit_table (pd.DataFrame): The augmented results table from Task 15.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        pd.DataFrame: The final static EVT table containing GPD fits, NRMSD, and \hat\Theta.
    """
    theta_results: Dict[Tuple[int, str, float], Dict[str, Any]] = {}

    for key, alpha_dict in exceedance_objects.items():
        for alpha, exc_obj in alpha_dict.items():
            t_positions = exc_obj["t_positions"]

            # Step 1: Compute inter-exceedance times
            tau, is_estimable, status = compute_inter_exceedance_times(t_positions)

            # Step 2: Apply estimator
            res = apply_ferro_segers_estimator(tau)
            res["is_estimable"] = is_estimable
            res["status_message"] = status

            theta_results[(key.mode, key.tail, alpha)] = res

    # Step 3: Augment table
    final_table = augment_table_with_extremal_index(gpd_fit_table, theta_results)

    return final_table
````

````python
# Task 17: Compute absolute-return autocorrelation for the first five modes

# ==============================================================================
# Task 17: Compute absolute-return autocorrelation for the first five modes
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 1: Define absolute-return series and compute sample mean
# -------------------------------------------------------------------------------------------------------------------------------

def extract_absolute_returns(
    r_memmap_path: str,
    mode_k: int,
    t_total: int
) -> Tuple[np.ndarray, float]:
    """
    Extracts the absolute-return series for a specific mode and computes its full-sample mean.

    Equation: X_t = |R_k(t)|, \bar{X} = (1/T) \sum X_t

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap (K x T).
        mode_k (int): The 1-indexed mode number (1 to 5).
        t_total (int): The total number of return intervals T.

    Returns:
        Tuple[np.ndarray, float]:
            - 1D float64 array of absolute returns (length T).
            - The scalar sample mean \bar{X}.
    """
    # Open R in read-only mode. We assume K=479 based on prior validation.
    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=(479, t_total))

    # Extract row (0-indexed) and force contiguous memory layout
    r_row: np.ndarray = r_memmap[mode_k - 1, :].copy()

    # Compute absolute returns
    x_abs: np.ndarray = np.abs(r_row)

    # Compute full-sample mean using float64 for precision
    x_bar: float = float(np.mean(x_abs, dtype=np.float64))

    # Explicitly delete memmap reference
    del r_memmap

    return x_abs, x_bar

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 2: Compute empirical ACF via FFT
# -------------------------------------------------------------------------------------------------------------------------------

def compute_acf_fft(x_abs: np.ndarray, x_bar: float, h_max: int) -> np.ndarray:
    """
    Computes the empirical autocorrelation function (ACF) up to lag h_max
    using a highly optimized FFT-based approach.

    Equation: \widehat{\rho}_h = \frac{\sum (X_t - \bar{X})(X_{t+h} - \bar{X})}{\sum (X_t - \bar{X})^2}

    Args:
        x_abs (np.ndarray): 1D array of absolute returns (length T).
        x_bar (float): The sample mean of x_abs.
        h_max (int): The maximum lag to compute.

    Returns:
        np.ndarray: 1D float64 array of ACF values for lags 1 through h_max.

    Raises:
        RuntimeError: If the lag-0 autocovariance normalization fails.
    """
    t_total = len(x_abs)

    # 1. Center the series
    z = x_abs - x_bar

    # 2. Compute the denominator (full-sample variance)
    denom = np.sum(z**2, dtype=np.float64)

    if denom == 0:
        # Flat series, ACF is undefined. Return zeros.
        return np.zeros(h_max, dtype=np.float64)

    # 3. Zero-pad to avoid circular correlation artifacts
    # The minimum safe length for linear correlation is 2*T - 1.
    # We use scipy.fft.next_fast_len to find the optimal highly-composite length >= 2*T - 1.
    fast_len = scipy.fft.next_fast_len(2 * t_total - 1)

    # 4. Compute FFT
    # rfft is optimized for real-valued inputs
    f_z = scipy.fft.rfft(z, n=fast_len)

    # 5. Compute power spectrum
    power_spectrum = f_z * np.conj(f_z)

    # 6. Compute inverse FFT to get autocovariance
    # irfft returns real values. We only need the first h_max + 1 elements.
    acov_full = scipy.fft.irfft(power_spectrum, n=fast_len)

    # Extract lags 0 through h_max
    acov = acov_full[:h_max + 1]

    # Defensive check: acov[0] should equal denom within floating point tolerance
    if abs(acov[0] - denom) / denom > 1e-5:
        raise RuntimeError("FFT-based autocovariance failed normalization check at lag 0.")

    # 7. Normalize to get ACF and drop lag 0
    acf = acov[1:] / denom

    return acf

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 3: Store ACF vectors and verify qualitative behavior
# -------------------------------------------------------------------------------------------------------------------------------

def compile_acf_diagnostics(
    acf_dict: Dict[int, np.ndarray],
    h_max: int
) -> Dict[str, Any]:
    """
    Compiles summary diagnostics for the computed ACF vectors and verifies
    the expected qualitative findings from the manuscript.

    Args:
        acf_dict (Dict[int, np.ndarray]): Mapping of mode_k -> ACF array.
        h_max (int): The maximum lag computed.

    Returns:
        Dict[str, Any]: A structured dictionary containing the ACF arrays and diagnostics.
    """
    results: Dict[str, Any] = {
        "acf_vectors": {},
        "diagnostics": {},
        "qualitative_validation": {}
    }

    for mode_k, acf in acf_dict.items():
        # Store the full vector (converted to list for JSON serialization downstream)
        results["acf_vectors"][f"mode_{mode_k}"] = acf.tolist()

        # Note: array index 0 is lag 1, index 99 is lag 100, etc.
        diag = {
            "acf_at_lag_1": float(acf[0]) if h_max >= 1 else None,
            "acf_at_lag_100": float(acf[99]) if h_max >= 100 else None,
            "acf_at_lag_1000": float(acf[999]) if h_max >= 1000 else None,
            "acf_at_lag_10000": float(acf[9999]) if h_max >= 10000 else None
        }

        # Estimate half-life: first lag where ACF < ACF[0] / 2
        half_threshold = acf[0] / 2.0
        below_threshold_indices = np.where(acf < half_threshold)[0]

        if len(below_threshold_indices) > 0:
            # Add 1 because index 0 is lag 1
            diag["half_life_estimate_seconds"] = int(below_threshold_indices[0] + 1)
        else:
            diag["half_life_estimate_seconds"] = f"> {h_max}"

        results["diagnostics"][f"mode_{mode_k}"] = diag

    # Qualitative Validation
    if 1 in acf_dict and 2 in acf_dict:
        acf_m1 = acf_dict[1]
        acf_m2 = acf_dict[2]

        # Check if Mode 2 (Energy) has stronger persistence than Mode 1 (Market) at long lags
        if h_max >= 10000:
            m2_stronger_10k = acf_m2[9999] > acf_m1[9999]
            results["qualitative_validation"]["mode_2_persistence_gt_mode_1_at_10k"] = bool(m2_stronger_10k)

        # Check if all modes show positive autocorrelation at lag 10000
        # 95% CI for white noise is ~ 1.96 / sqrt(T) ~ 0.00084
        all_positive_long_range = all(acf[9999] > 0.001 for acf in acf_dict.values()) if h_max >= 10000 else None
        results["qualitative_validation"]["all_modes_show_long_range_dependence"] = bool(all_positive_long_range)

    return results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def compute_absolute_return_acf(
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the computation of the absolute-return autocorrelation function.

    Extracts the first 5 modes, computes the ACF up to 10,000 lags using a fast
    FFT-based autocovariance algorithm, and compiles diagnostic metrics to verify
    long-range volatility persistence.

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: A comprehensive dictionary containing ACF vectors and diagnostics.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    t_total: int = config["T_total_return_intervals_expected"]

    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    modes_to_analyze: tuple = diag_config.get("modes_for_evt_analysis", (1, 2, 3, 4, 5))
    h_max: int = diag_config.get("acf_max_lag_seconds", 10000)

    acf_dict: Dict[int, np.ndarray] = {}

    for mode_k in modes_to_analyze:
        # Step 1: Extract series and mean
        x_abs, x_bar = extract_absolute_returns(r_memmap_path, mode_k, t_total)

        # Step 2: Compute ACF via FFT
        acf = compute_acf_fft(x_abs, x_bar, h_max)
        acf_dict[mode_k] = acf

        # Free memory before next iteration
        del x_abs
        gc.collect()

    # Step 3: Compile diagnostics
    acf_data = compile_acf_diagnostics(acf_dict, h_max)

    return acf_data
````

````python
# Task 18: Infer GEV block-maxima distributions from static POT fits

# ==============================================================================
# Task 18: Infer GEV block-maxima distributions from static POT fits
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 1: Compute empirical exceedance probability and GEV parameters
# -------------------------------------------------------------------------------------------------------------------------------

def infer_gev_parameters(
    gpd_fit_table: pd.DataFrame,
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Tuple[pd.DataFrame, Dict[SeriesKey, Dict[str, Any]]]:
    """
    Computes the empirical exceedance probability and infers GEV parameters (a, b)
    from the GPD fits for both the alpha-grid and a direct threshold of u=15.

    Args:
        gpd_fit_table (pd.DataFrame): The augmented GPD results table.
        r_memmap_path (str): Path to the whitened modes memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, Dict[SeriesKey, Dict[str, Any]]]:
            - The GPD table augmented with GEV parameters.
            - A dictionary containing the direct-threshold (u=15) GEV inference.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    t_total: int = config["T_total_return_intervals_expected"]
    m_block_size: int = config["return_intervals_per_day_expected"]
    k_assets: int = config["K_assets"]

    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["block_maxima_inference"]
    u_benchmark: float = evt_config.get("inference_threshold_u_static_example", 15.0)

    df = gpd_fit_table.copy()
    df["zeta_u"] = df["n_exceedances"] / float(t_total)
    df["gev_a"] = np.nan
    df["gev_b"] = np.nan

    # 1. Infer GEV parameters for the alpha-grid
    for idx, row in df.iterrows():
        if not row["is_reliable"]:
            continue

        gamma = float(row["gamma_hat"])
        sigma = float(row["sigma_hat"])
        u = float(row["threshold_u"])
        zeta = float(row["zeta_u"])

        zm = zeta * m_block_size

        if abs(gamma) < 1e-8:
            a = sigma
            b = u + sigma * np.log(zm)
        else:
            a = sigma * (zm ** gamma)
            b = u + (sigma / gamma) * ((zm ** gamma) - 1.0)

        df.at[idx, "gev_a"] = a
        df.at[idx, "gev_b"] = b

    # 2. Direct-threshold inference for u = 15
    # We must extract exceedances and fit the GPD specifically for u=15
    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=(k_assets, t_total))
    direct_u_results: Dict[SeriesKey, Dict[str, Any]] = {}

    modes_to_analyze = manifest["algorithmic_parameters"]["diagnostics_and_plotting"].get("modes_for_evt_analysis", [1,2,3,4,5])

    for mode_k in modes_to_analyze:
        r_row = r_memmap[mode_k - 1, :].copy()

        for tail, series in [("positive", r_row), ("negative", -r_row)]:
            key = SeriesKey(mode=mode_k, tail=tail)

            # Extract exceedances
            exceed_mask = series > u_benchmark
            x_exceedances = series[exceed_mask] - u_benchmark
            n_count = len(x_exceedances)

            res = {
                "threshold_u": u_benchmark,
                "n_exceedances": n_count,
                "zeta_u": n_count / float(t_total),
                "is_reliable": False
            }

            if n_count >= 30:
                # Fit GPD using the function from Task 14
                # Assuming optimize_gpd_mle is available in the global namespace
                fit_res = optimize_gpd_mle(x_exceedances, manifest)
                res.update(fit_res)

                is_reliable = fit_res["converged"] and (not np.isnan(fit_res["gradient_norm"])) and (fit_res["gradient_norm"] < 1e-4)
                res["is_reliable"] = is_reliable

                if is_reliable:
                    gamma = fit_res["gamma_hat"]
                    sigma = fit_res["sigma_hat"]
                    zm = res["zeta_u"] * m_block_size

                    if abs(gamma) < 1e-8:
                        res["gev_a"] = sigma
                        res["gev_b"] = u_benchmark + sigma * np.log(zm)
                    else:
                        res["gev_a"] = sigma * (zm ** gamma)
                        res["gev_b"] = u_benchmark + (sigma / gamma) * ((zm ** gamma) - 1.0)

            direct_u_results[key] = res

    del r_memmap
    gc.collect()

    return df, direct_u_results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 2: Construct the implied GEV density on a fine grid
# -------------------------------------------------------------------------------------------------------------------------------

def evaluate_gev_density(x: np.ndarray, gamma: float, b: float, a: float) -> np.ndarray:
    """
    Evaluates the analytical GEV probability density function.

    Args:
        x (np.ndarray): Grid of evaluation points.
        gamma (float): Shape parameter.
        b (float): Location parameter.
        a (float): Scale parameter (must be > 0).

    Returns:
        np.ndarray: Density values corresponding to x.
    """
    density = np.zeros_like(x, dtype=np.float64)

    if abs(gamma) < 1e-8:
        # Gumbel limit
        w = (x - b) / a
        # Use np.exp carefully to avoid overflow warnings, though underflow to 0 is fine
        density = (1.0 / a) * np.exp(-w) * np.exp(-np.exp(-w))
    else:
        # General Fréchet/Weibull
        z = 1.0 + gamma * (x - b) / a

        # Support constraint: z > 0
        valid_mask = z > 0
        z_valid = z[valid_mask]

        # g(x) = (1/a) * z^{-(1 + 1/\gamma)} * exp(-z^{-1/\gamma})
        exponent = -1.0 / gamma
        term1 = np.power(z_valid, exponent - 1.0)
        term2 = np.exp(-np.power(z_valid, exponent))

        density[valid_mask] = (1.0 / a) * term1 * term2

    return density

def construct_gev_density_curves(
    direct_u_results: Dict[SeriesKey, Dict[str, Any]]
) -> Dict[SeriesKey, Dict[str, np.ndarray]]:
    """
    Constructs the implied GEV density curves for the direct-threshold (u=15) fits.

    Args:
        direct_u_results (Dict): The GEV parameters for u=15.

    Returns:
        Dict: Mapping (mode, tail) -> {"x_grid": array, "density": array}
    """
    density_curves: Dict[SeriesKey, Dict[str, np.ndarray]] = {}

    for key, res in direct_u_results.items():
        if not res.get("is_reliable", False) or "gev_a" not in res:
            continue

        gamma = res["gamma_hat"]
        a = res["gev_a"]
        b = res["gev_b"]

        # Define a reasonable grid based on location and scale
        # For heavy tails, extend far to the right
        x_min = b - 5 * a
        x_max = b + 20 * a if gamma > 0 else b - a/gamma + 1e-5

        x_grid = np.linspace(x_min, x_max, 2000)
        density = evaluate_gev_density(x_grid, gamma, b, a)

        # Verify numerical integration is close to 1.0
        integral = np.trapz(density, x_grid)
        if integral < 0.9 or integral > 1.1:
            import logging
            logging.warning(f"GEV density integral for {key} is {integral:.3f}, grid may not cover support adequately.")

        density_curves[key] = {
            "x_grid": x_grid,
            "density": density
        }

    return density_curves

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 3: Compute empirical daily maxima and compare
# -------------------------------------------------------------------------------------------------------------------------------

def compute_empirical_maxima_and_compare(
    r_memmap_path: str,
    day_boundary_map: Tuple[Tuple[pd.Timestamp, int, int], ...],
    direct_u_results: Dict[SeriesKey, Dict[str, Any]],
    density_curves: Dict[SeriesKey, Dict[str, np.ndarray]],
    manifest: MappingProxyType
) -> Dict[SeriesKey, Dict[str, Any]]:
    """
    Extracts empirical daily maxima for each mode and tail, computes their
    histogram, and performs a Kolmogorov-Smirnov test against the inferred GEV.

    Args:
        r_memmap_path (str): Path to the whitened modes memmap.
        day_boundary_map (Tuple): Map of day boundaries.
        direct_u_results (Dict): GEV parameters.
        density_curves (Dict): GEV density grids.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict: Comparison diagnostics and empirical data for plotting.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    n_days: int = len(day_boundary_map)

    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=(k_assets, t_total))

    comparison_results: Dict[SeriesKey, Dict[str, Any]] = {}

    # Iterate through keys and results
    for key, res in direct_u_results.items():
        if not res.get("is_reliable", False):
            continue

        mode_k = key.mode
        tail = key.tail

        r_row = r_memmap[mode_k - 1, :].copy()
        series = r_row if tail == "positive" else -r_row

        # Extract daily maxima
        daily_maxima = np.zeros(n_days, dtype=np.float64)
        for d, (_, start_col, end_col) in enumerate(day_boundary_map):
            daily_maxima[d] = np.max(series[start_col:end_col])

        # Compute empirical histogram using the same grid range as the density curve
        x_grid = density_curves[key]["x_grid"]
        bins = np.linspace(x_grid[0], x_grid[-1], 30)
        counts, edges = np.histogram(daily_maxima, bins=bins, density=True)

        # Perform KS Test
        # We need the GEV CDF for the KS test
        gamma = res["gamma_hat"]
        a = res["gev_a"]
        b = res["gev_b"]

        def gev_cdf(x):
            if abs(gamma) < 1e-8:
                return np.exp(-np.exp(-(x - b) / a))
            else:
                z = 1.0 + gamma * (x - b) / a
                z = np.maximum(z, 1e-10) # Enforce support
                return np.exp(-np.power(z, -1.0 / gamma))

        ks_stat, p_value = scipy.stats.kstest(daily_maxima, gev_cdf)

        comparison_results[key] = {
            "empirical_daily_maxima": daily_maxima.tolist(),
            "histogram_counts": counts.tolist(),
            "histogram_edges": edges.tolist(),
            "ks_statistic": float(ks_stat),
            "ks_p_value": float(p_value),
            "match_confirmed": p_value > 0.05 # Fail to reject null hypothesis that distributions are the same
        }

    del r_memmap
    gc.collect()

    return comparison_results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def infer_gev_from_pot(
    gpd_fit_table: pd.DataFrame,
    r_memmap_path: str,
    day_boundary_map: Tuple[Tuple[pd.Timestamp, int, int], ...],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the inference of GEV block-maxima distributions from POT fits.

    Computes GEV parameters for the alpha-grid, performs a dedicated POT fit
    at u=15, constructs analytical GEV density curves, extracts empirical daily
    maxima, and rigorously compares the inferred vs. empirical distributions.

    Args:
        gpd_fit_table (pd.DataFrame): The augmented GPD results table.
        r_memmap_path (str): Path to the whitened modes memmap.
        day_boundary_map (Tuple): Map of day boundaries.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: A comprehensive dictionary containing the augmented table,
                        GEV parameters, density curves, and comparison metrics.
    """
    # Step 1: Infer parameters
    augmented_table, direct_u_results = infer_gev_parameters(
        gpd_fit_table, r_memmap_path, manifest
    )

    # Step 2: Construct density curves for u=15
    density_curves = construct_gev_density_curves(direct_u_results)

    # Step 3: Compute empirical maxima and compare
    comparison_results = compute_empirical_maxima_and_compare(
        r_memmap_path, day_boundary_map, direct_u_results, density_curves, manifest
    )

    # Assemble final artifact
    gev_inference_artifact = {
        "augmented_gpd_table": augmented_table,
        "benchmark_u15_parameters": direct_u_results,
        "benchmark_u15_density_curves": density_curves,
        "benchmark_u15_comparisons": comparison_results
    }

    return gev_inference_artifact
````

````python
# Task 19: Compute intraday volatility profiles and deseasonalized residual modes

# ================================================================================
# Task 19: Compute intraday volatility profiles and deseasonalized residual modes
# ================================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 1: Compute the average absolute-return profile V_k
# -------------------------------------------------------------------------------------------------------------------------------

def compute_intraday_volatility_profiles(
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Tuple[Dict[int, np.ndarray], Dict[int, np.ndarray]]:
    """
    Computes the intraday volatility profile V_k for the first 5 modes by
    averaging absolute returns across all retained trading days.

    Equation: V_k(t_intra) = (1/N_days) * sum_d |R_k(d * T_day + t_intra)|

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[Dict[int, np.ndarray], Dict[int, np.ndarray]]:
            - Dictionary mapping mode_k -> 1D array of raw mode returns R_k.
            - Dictionary mapping mode_k -> 1D array of volatility profile V_k.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    k_assets: int = config["K_assets"]
    t_total: int = config["T_total_return_intervals_expected"]
    n_days: int = config["N_days_expected_after_filters"]
    t_day: int = config["return_intervals_per_day_expected"]

    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    modes_to_analyze: tuple = diag_config.get("modes_for_evt_analysis", (1, 2, 3, 4, 5))

    # Floor parameter to prevent division by zero during deseasonalization
    dyn_config: MappingProxyType = manifest["algorithmic_parameters"]["dynamic_analysis"]
    profile_floor: float = dyn_config.get("volatility_profile_floor", 1e-10)

    r_memmap = np.memmap(r_memmap_path, dtype=np.float64, mode='r', shape=(k_assets, t_total))

    raw_modes: Dict[int, np.ndarray] = {}
    v_profiles: Dict[int, np.ndarray] = {}

    for mode_k in modes_to_analyze:
        # Extract 1D row and copy to RAM
        r_row = r_memmap[mode_k - 1, :].copy()
        raw_modes[mode_k] = r_row

        # Reshape to (N_days, T_day)
        r_2d = r_row.reshape((n_days, t_day))

        # Compute mean of absolute values across days (axis=0)
        v_k = np.mean(np.abs(r_2d), axis=0, dtype=np.float64)

        # Apply floor to prevent extreme amplification
        min_v = np.min(v_k)
        if min_v < profile_floor:
            import logging
            logging.warning(f"Mode {mode_k} profile has minimum {min_v:.2e}, applying floor {profile_floor:.2e}")
            v_k = np.maximum(v_k, profile_floor)

        v_profiles[mode_k] = v_k

    del r_memmap
    gc.collect()

    return raw_modes, v_profiles

# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 2: Compute the deseasonalized residual mode series
# -------------------------------------------------------------------------------------------------------------------------------

def compute_residual_modes(
    raw_modes: Dict[int, np.ndarray],
    v_profiles: Dict[int, np.ndarray],
    manifest: MappingProxyType
) -> Tuple[Dict[int, np.ndarray], Dict[str, Any]]:
    """
    Computes the deseasonalized residual mode series by dividing the raw returns
    by the intraday volatility profile.

    Equation: \tilde{R}_k(t) = R_k(t) / V_k(t \bmod T_day)

    Args:
        raw_modes (Dict): Mapping of mode_k -> 1D array R_k.
        v_profiles (Dict): Mapping of mode_k -> 1D array V_k.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[Dict[int, np.ndarray], Dict[str, Any]]:
            - Dictionary mapping mode_k -> 1D array of residual returns \tilde{R}_k.
            - Validation diagnostics comparing profile flatness.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    n_days: int = config["N_days_expected_after_filters"]
    t_day: int = config["return_intervals_per_day_expected"]

    residual_modes: Dict[int, np.ndarray] = {}
    diagnostics: Dict[str, Any] = {}

    for mode_k, r_row in raw_modes.items():
        v_k = v_profiles[mode_k]

        # Reshape R_k to 2D for efficient broadcasting
        r_2d = r_row.reshape((n_days, t_day))

        # Divide by V_k (broadcasting across the 248 days)
        r_tilde_2d = r_2d / v_k[np.newaxis, :]

        # Flatten back to 1D
        r_tilde = r_tilde_2d.reshape(-1)
        residual_modes[mode_k] = r_tilde

        # Validation: Check if the residual profile is flat
        v_tilde_k = np.mean(np.abs(r_tilde_2d), axis=0, dtype=np.float64)

        cv_original = float(np.std(v_k) / np.mean(v_k))
        cv_residual = float(np.std(v_tilde_k) / np.mean(v_tilde_k))

        diagnostics[f"mode_{mode_k}"] = {
            "cv_original_profile": cv_original,
            "cv_residual_profile": cv_residual,
            "deseasonalization_effective": cv_residual < (cv_original * 0.1) # Expect 90% reduction in variance
        }

    return residual_modes, diagnostics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 3: Reproduce deseasonalization diagnostics (ACF and Theta)
# -------------------------------------------------------------------------------------------------------------------------------

def compute_residual_diagnostics(
    residual_modes: Dict[int, np.ndarray],
    manifest: MappingProxyType
) -> Tuple[Dict[int, np.ndarray], Dict[SeriesKey, Dict[float, Dict[str, Any]]]]:
    """
    Computes the ACF of absolute residuals and the fixed-threshold extremal index
    for the residual series to verify that clustering persists despite deseasonalization.

    Args:
        residual_modes (Dict): Mapping of mode_k -> 1D array \tilde{R}_k.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[Dict, Dict]:
            - Dictionary mapping mode_k -> residual ACF array.
            - Nested dictionary containing residual \hat\Theta results.
    """
    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    h_max: int = diag_config.get("acf_max_lag_seconds", 10000)

    residual_acf: Dict[int, np.ndarray] = {}

    # 1. Compute Residual ACF
    for mode_k, r_tilde in residual_modes.items():
        x_abs = np.abs(r_tilde)
        x_bar = float(np.mean(x_abs, dtype=np.float64))

        # Assuming compute_acf_fft is available from Task 17
        acf = compute_acf_fft(x_abs, x_bar, h_max)
        residual_acf[mode_k] = acf

    # 2. Compute Residual Fixed-Threshold Extremal Index
    # We must format the residual series into the structure expected by Task 13
    analysis_series: Dict[SeriesKey, np.ndarray] = {}
    for mode_k, r_tilde in residual_modes.items():
        analysis_series[SeriesKey(mode=mode_k, tail="positive")] = r_tilde
        analysis_series[SeriesKey(mode=mode_k, tail="negative")] = -r_tilde

    # Assuming compute_static_thresholds and extract_exceedances are available from Task 13
    thresholds = compute_static_thresholds(analysis_series, manifest)
    exceedance_objects = extract_exceedances(analysis_series, thresholds, manifest)

    residual_theta_results: Dict[SeriesKey, Dict[float, Dict[str, Any]]] = {}

    for key, alpha_dict in exceedance_objects.items():
        residual_theta_results[key] = {}
        for alpha, exc_obj in alpha_dict.items():
            t_positions = exc_obj["t_positions"]

            # Assuming compute_inter_exceedance_times and apply_ferro_segers_estimator are available from Task 16
            tau, is_estimable, _ = compute_inter_exceedance_times(t_positions)
            res = apply_ferro_segers_estimator(tau)
            res["is_estimable"] = is_estimable

            residual_theta_results[key][alpha] = res

    return residual_acf, residual_theta_results

# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def build_residual_modes(
    r_memmap_path: str,
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the computation of intraday volatility profiles and deseasonalized residuals.

    Extracts raw modes, computes the U-shaped volatility profile, divides to create
    stationary residuals, and computes diagnostic metrics (ACF and Theta) to prove
    that volatility clustering is not merely an artifact of intraday seasonality.

    Args:
        r_memmap_path (str): Path to the 2D whitened modes memmap.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: A comprehensive dictionary containing the profiles, residuals,
                        and diagnostic metrics.
    """
    # Step 1: Compute profiles
    raw_modes, v_profiles = compute_intraday_volatility_profiles(r_memmap_path, manifest)

    # Step 2: Compute residuals
    residual_modes, deseasonalization_diagnostics = compute_residual_modes(raw_modes, v_profiles, manifest)

    # Free raw modes to save memory
    del raw_modes
    gc.collect()

    # Step 3: Compute residual diagnostics
    residual_acf, residual_theta = compute_residual_diagnostics(residual_modes, manifest)

    # Assemble artifact
    residual_artifact = {
        "v_profiles": v_profiles,
        "residual_modes": residual_modes, # Kept in RAM (~215 MB for 5 modes)
        "deseasonalization_diagnostics": deseasonalization_diagnostics,
        "residual_acf": residual_acf,
        "residual_theta": residual_theta
    }

    return residual_artifact
````

````python
# Task 20: Run the rolling local-threshold extreme-value analysis

# ==============================================================================
# Task 20: Run the rolling local-threshold extreme-value analysis
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 1: Compute rolling backward-looking local thresholds
# -------------------------------------------------------------------------------------------------------------------------------

def compute_rolling_thresholds(
    residual_modes: Dict[int, np.ndarray],
    manifest: MappingProxyType
) -> Dict[SeriesKey, Dict[float, np.ndarray]]:
    """
    Computes the rolling backward-looking local threshold u(t) for each mode
    and local alpha. Strictly excludes the current observation to prevent look-ahead bias.

    Args:
        residual_modes (Dict[int, np.ndarray]): Deseasonalized residual series \tilde{R}_k.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[SeriesKey, Dict[float, np.ndarray]]: Nested dictionary mapping
            (mode, tail) -> {alpha: u(t) array of length T}.
            The first W elements of u(t) will be NaN.
    """
    dyn_config: MappingProxyType = manifest["algorithmic_parameters"]["dynamic_analysis"]
    w_size: int = dyn_config.get("rolling_window_size_W_seconds", 10000)
    alpha_grid: tuple = dyn_config["local_tail_quantile_grid_alpha"]

    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    modes_to_analyze: tuple = diag_config.get("modes_for_evt_analysis", (1, 2, 3, 4, 5))

    rolling_thresholds: Dict[SeriesKey, Dict[float, np.ndarray]] = {}

    for mode_k in modes_to_analyze:
        r_tilde = residual_modes[mode_k]

        # Define positive and negative tail series
        series_dict = {
            "positive": r_tilde,
            "negative": -r_tilde
        }

        for tail, series in series_dict.items():
            key = SeriesKey(mode=mode_k, tail=tail)
            rolling_thresholds[key] = {}

            # Convert to pandas Series for optimized rolling computation
            pd_series = pd.Series(series)

            # Shift by 1 to exclude current observation, then apply rolling window
            # min_periods=w_size ensures the first W-1 values are NaN
            rolling_obj = pd_series.shift(1).rolling(window=w_size, min_periods=w_size)

            for alpha in alpha_grid:
                # Compute quantile. method='linear' is type-7 interpolation.
                # This operation is computationally intensive (~1 min per alpha per mode)
                try:
                    u_t_series = rolling_obj.quantile(1.0 - alpha, interpolation='linear')
                except TypeError:
                    # For newer pandas versions
                    u_t_series = rolling_obj.quantile(1.0 - alpha, method='linear')

                rolling_thresholds[key][alpha] = u_t_series.to_numpy(dtype=np.float64)

    return rolling_thresholds

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 2: Extract local exceedances, fit pooled GPD, and compute diagnostics
# -------------------------------------------------------------------------------------------------------------------------------

def fit_dynamic_gpd_and_diagnostics(
    residual_modes: Dict[int, np.ndarray],
    rolling_thresholds: Dict[SeriesKey, Dict[float, np.ndarray]],
    manifest: MappingProxyType
) -> Tuple[pd.DataFrame, Dict[SeriesKey, Dict[float, Dict[str, Any]]]]:
    """
    Extracts local exceedances, pools them, fits a global GPD, and computes
    standard errors, NRMSD, and the dynamic extremal index \hat\Theta.

    Args:
        residual_modes (Dict): Deseasonalized residual series.
        rolling_thresholds (Dict): Time-varying thresholds u(t).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[pd.DataFrame, Dict]:
            - A comprehensive results DataFrame for the dynamic analysis.
            - A dictionary containing the raw exceedance objects (for downstream use).
    """
    dyn_config: MappingProxyType = manifest["algorithmic_parameters"]["dynamic_analysis"]
    w_size: int = dyn_config.get("rolling_window_size_W_seconds", 10000)

    evt_config: MappingProxyType = manifest["algorithmic_parameters"]["static_evt"]
    min_exceedance_count: int = evt_config.get("min_exceedance_count", 30)

    records: List[Dict[str, Any]] = []
    dynamic_exceedance_objects: Dict[SeriesKey, Dict[float, Dict[str, Any]]] = {}

    for key, alpha_dict in rolling_thresholds.items():
        mode_k = key.mode
        tail = key.tail
        series = residual_modes[mode_k] if tail == "positive" else -residual_modes[mode_k]

        dynamic_exceedance_objects[key] = {}

        for alpha, u_t in alpha_dict.items():
            # 1. Extract exceedances (ignoring the first W elements where u_t is NaN)
            valid_mask = ~np.isnan(u_t)
            exceed_mask = np.zeros_like(series, dtype=bool)
            exceed_mask[valid_mask] = series[valid_mask] > u_t[valid_mask]

            t_positions = np.where(exceed_mask)[0].astype(np.int64)
            x_exceedances = series[t_positions] - u_t[t_positions]
            n_count = len(x_exceedances)

            is_sufficient = n_count >= min_exceedance_count

            # Store exceedance object
            exc_obj = {
                "u_alpha": np.nan, # Threshold is time-varying, not scalar
                "x_exceedances": x_exceedances,
                "t_positions": t_positions,
                "n": n_count,
                "is_sufficient": is_sufficient
            }
            dynamic_exceedance_objects[key][alpha] = exc_obj

            # Base record
            record = {
                "analysis_type": "dynamic",
                "mode": mode_k,
                "tail": tail,
                "alpha": alpha,
                "window_size": w_size,
                "n_exceedances": n_count,
                "is_sufficient": is_sufficient
            }

            if is_sufficient:
                # 2. Fit GPD
                fit_res = optimize_gpd_mle(x_exceedances, manifest)
                is_reliable = fit_res["converged"] and (not np.isnan(fit_res["gradient_norm"])) and (fit_res["gradient_norm"] < 1e-4)
                record.update(fit_res)
                record["is_reliable"] = is_reliable

                # 3. Compute Extremal Index
                tau, is_estimable, _ = compute_inter_exceedance_times(t_positions)
                theta_res = apply_ferro_segers_estimator(tau)
                record.update(theta_res)
                record["is_estimable"] = is_estimable

            else:
                null_res = {
                    "gamma_hat": np.nan, "sigma_hat": np.nan, "nll_achieved": np.nan,
                    "converged": False, "n_iterations": 0, "gradient_norm": np.nan,
                    "support_binding": False, "is_reliable": False,
                    "theta_hat": np.nan, "theta_branch": "none", "tau_max": np.nan,
                    "tau_mean": np.nan, "n_tau": 0, "is_estimable": False, "mean_cluster_size": np.nan
                }
                record.update(null_res)

            records.append(record)

    df_results = pd.DataFrame(records)

    # 4. Compute Standard Errors and NRMSD
    # We reuse the Task 15 functions. They expect specific column names which we have provided.
    df_with_se = compute_asymptotic_standard_errors(df_results)
    qq_data = construct_qq_diagnostic_data(df_with_se, dynamic_exceedance_objects)
    final_df, _ = compute_nrmsd_and_credible_regions(df_with_se, qq_data)

    return final_df, dynamic_exceedance_objects

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 3: Compute time-varying GEV parameters and density surface
# -------------------------------------------------------------------------------------------------------------------------------

def compute_time_varying_gev(
    dynamic_fit_table: pd.DataFrame,
    rolling_thresholds: Dict[SeriesKey, Dict[float, np.ndarray]],
    manifest: MappingProxyType
) -> Tuple[Dict[Tuple[int, str, float], Dict[str, Any]], Dict[str, Any]]:
    """
    Computes the constant scale 'a' and time-varying location 'b(t)' for the GEV.
    Constructs the nonstationary density surface for the specific manuscript example.

    Args:
        dynamic_fit_table (pd.DataFrame): The dynamic GPD results table.
        rolling_thresholds (Dict): Time-varying thresholds u(t).
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Tuple[Dict, Dict]:
            - Mapping of (mode, tail, alpha) -> {"a": float, "b_t": array}.
            - Density surface data for the specific example.
    """
    config: MappingProxyType = manifest["algorithmic_parameters"]["data_preprocessing"]
    m_block_size: int = config["return_intervals_per_day_expected"]
    t_day: int = config["return_intervals_per_day_expected"]

    dyn_config: MappingProxyType = manifest["algorithmic_parameters"]["dynamic_analysis"]
    w_size: int = dyn_config.get("rolling_window_size_W_seconds", 10000)

    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    ex_mode = diag_config.get("example_nonstationary_pdf_mode", 2)
    ex_days = diag_config.get("nonstationary_pdf_plot_days", 40)
    ex_alpha = dyn_config.get("local_quantile_q_example", 0.999)
    ex_tail_alpha = round(1.0 - ex_alpha, 5) # e.g., 0.001

    time_varying_params: Dict[Tuple[int, str, float], Dict[str, Any]] = {}
    density_surface_data: Dict[str, Any] = {}

    for idx, row in dynamic_fit_table.iterrows():
        if not row["is_reliable"]:
            continue

        mode = int(row["mode"])
        tail = str(row["tail"])
        alpha = float(row["alpha"])
        gamma = float(row["gamma_hat"])
        sigma = float(row["sigma_hat"])

        key = SeriesKey(mode=mode, tail=tail)
        u_t = rolling_thresholds[key][alpha]

        # Approximation: \hat\zeta_{u(t)} \approx \alpha
        zm = alpha * m_block_size

        if abs(gamma) < 1e-8:
            a = sigma
            offset = sigma * np.log(zm)
        else:
            a = sigma * (zm ** gamma)
            offset = (sigma / gamma) * ((zm ** gamma) - 1.0)

        b_t = u_t + offset

        time_varying_params[(mode, tail, alpha)] = {
            "gev_a": a,
            "gev_b_t": b_t
        }

        # Construct density surface for the specific example
        if mode == ex_mode and tail == "positive" and abs(alpha - ex_tail_alpha) < 1e-6:
            # Extract first 40 days
            t_end = ex_days * t_day
            b_t_slice = b_t[w_size:t_end] # Skip the initial NaN window

            # Subsample time axis to save memory (e.g., every 100 seconds)
            subsample_step = 100
            b_t_sub = b_t_slice[::subsample_step]
            t_indices = np.arange(w_size, t_end, subsample_step)

            # Define x-grid
            x_min = np.nanmin(b_t_sub) - 3 * a
            x_max = np.nanmax(b_t_sub) + 15 * a
            x_grid = np.linspace(x_min, x_max, 200)

            # Evaluate density surface
            # b_t_sub is shape (N,), x_grid is shape (200,)
            # We want a surface of shape (N, 200)
            b_t_col = b_t_sub[:, np.newaxis]
            x_grid_row = x_grid[np.newaxis, :]

            if abs(gamma) < 1e-8:
                w = (x_grid_row - b_t_col) / a
                surface = (1.0 / a) * np.exp(-w) * np.exp(-np.exp(-w))
            else:
                z = 1.0 + gamma * (x_grid_row - b_t_col) / a
                surface = np.zeros_like(z)
                valid = z > 0
                exponent = -1.0 / gamma
                surface[valid] = (1.0 / a) * np.power(z[valid], exponent - 1.0) * np.exp(-np.power(z[valid], exponent))

            density_surface_data = {
                "mode": mode,
                "alpha": alpha,
                "time_indices": t_indices.tolist(),
                "x_grid": x_grid.tolist(),
                "density_surface": surface.tolist() # List of lists for JSON serialization
            }

    return time_varying_params, density_surface_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def fit_dynamic_evt_models(
    residual_modes: Dict[int, np.ndarray],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the rolling local-threshold extreme-value analysis.

    Computes time-varying thresholds to account for nonstationarity, extracts
    local exceedances, fits a global GPD to the pooled exceedances, computes
    the dynamic extremal index, and infers time-varying GEV parameters.

    Args:
        residual_modes (Dict[int, np.ndarray]): Deseasonalized residual series.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: A comprehensive dictionary containing the dynamic EVT
                        results table, time-varying parameters, and density surface.
    """
    # Step 1: Compute rolling thresholds
    rolling_thresholds = compute_rolling_thresholds(residual_modes, manifest)

    # Step 2: Extract exceedances, fit GPD, compute Theta and NRMSD
    dynamic_fit_table, dynamic_exceedance_objects = fit_dynamic_gpd_and_diagnostics(
        residual_modes, rolling_thresholds, manifest
    )

    # Step 3: Compute time-varying GEV parameters and density surface
    time_varying_params, density_surface_data = compute_time_varying_gev(
        dynamic_fit_table, rolling_thresholds, manifest
    )

    # Assemble artifact
    dynamic_evt_artifact = {
        "dynamic_fit_table": dynamic_fit_table,
        "time_varying_gev_parameters": time_varying_params,
        "example_density_surface": density_surface_data
    }

    return dynamic_evt_artifact
````

````python
# Task 21: Reproduce nonstationary diagnostics and assemble the full output package

# ==============================================================================
# Task 21: Reproduce nonstationary diagnostics and assemble the full output package
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 1: Reproduce the rolling-threshold visualization data for mode 2
# -------------------------------------------------------------------------------------------------------------------------------

def prepare_rolling_threshold_figure_data(
    residual_modes: Dict[int, np.ndarray],
    dynamic_evt_artifact: Dict[str, Any],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Prepares the subsampled time-series data required to reproduce the manuscript's
    figure showing the residual series, the static threshold, and the rolling threshold.

    Focuses on Mode 2, positive tail, alpha = 0.001, W = 10000.

    Args:
        residual_modes (Dict): Deseasonalized residual series.
        dynamic_evt_artifact (Dict): Outputs from Task 20.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: Subsampled arrays ready for plotting.
    """
    dyn_config: MappingProxyType = manifest["algorithmic_parameters"]["dynamic_analysis"]
    w_size: int = dyn_config.get("rolling_window_size_W_seconds", 10000)

    diag_config: MappingProxyType = manifest["algorithmic_parameters"]["diagnostics_and_plotting"]
    target_mode: int = diag_config.get("example_rolling_threshold_mode", 2)
    target_alpha: float = round(1.0 - dyn_config.get("local_quantile_q_example", 0.999), 5) # 0.001

    # 1. Extract full residual series
    r_tilde = residual_modes[target_mode]
    t_total = len(r_tilde)

    # 2. Compute static threshold (global quantile)
    # method='linear' matches type-7 interpolation
    try:
        u_static = float(np.quantile(r_tilde, 1.0 - target_alpha, method='linear'))
    except TypeError:
        u_static = float(np.quantile(r_tilde, 1.0 - target_alpha, interpolation='linear'))

    # 3. Extract rolling threshold
    # We must reconstruct it or extract it from the dynamic fit table if stored.
    # Since Task 20 didn't pass the raw u(t) arrays forward in the final artifact
    # (to save memory), we recompute just this one specific u(t) curve.
    pd_series = pd.Series(r_tilde)
    rolling_obj = pd_series.shift(1).rolling(window=w_size, min_periods=w_size)
    try:
        u_rolling = rolling_obj.quantile(1.0 - target_alpha, method='linear').to_numpy()
    except TypeError:
        u_rolling = rolling_obj.quantile(1.0 - target_alpha, interpolation='linear').to_numpy()

    # 4. Subsample for plotting (every 100th point)
    step = 100
    t_indices = np.arange(0, t_total, step)
    r_tilde_sub = r_tilde[::step]
    u_rolling_sub = u_rolling[::step]

    figure_data = {
        "mode": target_mode,
        "alpha": target_alpha,
        "window_size": w_size,
        "static_threshold": u_static,
        "time_indices": t_indices.tolist(),
        "residual_series_subsampled": r_tilde_sub.tolist(),
        "rolling_threshold_subsampled": u_rolling_sub.tolist()
    }

    return figure_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 2: Reproduce the rolling-threshold result curves and density surface
# -------------------------------------------------------------------------------------------------------------------------------

def prepare_dynamic_result_curves(
    dynamic_evt_artifact: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Extracts and formats the curve data (\hat\gamma, \rho, \hat\Theta vs \alpha)
    for the positive tail of the first 5 modes under rolling thresholds.

    Args:
        dynamic_evt_artifact (Dict): Outputs from Task 20.

    Returns:
        Dict[str, Any]: Structured dictionary of curve data and the density surface.
    """
    df_dyn = dynamic_evt_artifact["dynamic_fit_table"]

    # Filter for positive tail and reliable fits
    df_pos = df_dyn[(df_dyn["tail"] == "positive") & (df_dyn["is_reliable"] == True)].copy()

    curves_data: Dict[str, Dict[str, List[float]]] = {}

    for mode_k in range(1, 6):
        df_mode = df_pos[df_pos["mode"] == mode_k].sort_values("alpha", ascending=False)

        mode_key = f"mode_{mode_k}"
        curves_data[mode_key] = {
            "alpha": df_mode["alpha"].tolist(),
            "gamma_hat": df_mode["gamma_hat"].tolist(),
            "se_gamma": df_mode["se_gamma"].tolist(),
            "rho": df_mode["rho"].tolist(),
            "theta_hat": df_mode["theta_hat"].tolist()
        }

    # Extract density surface data directly from Task 20 artifact
    density_surface = dynamic_evt_artifact.get("example_density_surface", {})

    figure_data = {
        "positive_tail_curves": curves_data,
        "nonstationary_density_surface": density_surface
    }

    return figure_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 3: Assemble the full reproduction dossier and validate findings
# -------------------------------------------------------------------------------------------------------------------------------

def validate_manuscript_findings(
    spectral_diag: Dict[str, Any],
    static_evt_table: pd.DataFrame,
    dynamic_evt_table: pd.DataFrame,
    gev_inference: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Programmatically evaluates the manuscript's key findings using the formal
    decision rules defined in the disambiguation.

    This function synthesizes quantitative metrics from various pipeline stages
    to produce a structured confirmation assessment ("Confirmed", "Partially confirmed",
    or "Not confirmed") for each of the seven core claims made in the manuscript.

    Args:
        spectral_diag (Dict[str, Any]): Diagnostic metrics from the spectral decomposition (Task 12).
        static_evt_table (pd.DataFrame): Results from the static EVT analysis (Task 15/16).
        dynamic_evt_table (pd.DataFrame): Results from the dynamic rolling-window analysis (Task 20).
        gev_inference (Dict[str, Any]): Results from the GEV inference and comparison (Task 18).

    Returns:
        Dict[str, Any]: A dictionary mapping each finding key to its status and supporting evidence.
    """
    findings: Dict[str, Any] = {}

    # 1. Mode 1 is market-like
    # Check uniformity of eigenvector loadings (CV < 1.0) and sign coherence (> 0.8)
    m1_diag = spectral_diag["eigenvalue_analysis"]["market_mode_uniformity"]
    is_market = m1_diag["cv_absolute_loadings"] < 1.0 and m1_diag["positive_sign_fraction"] > 0.8
    findings["mode_1_is_market_like"] = {
        "status": "Confirmed" if is_market else "Not confirmed",
        "evidence": m1_diag
    }

    # 2. Modes 2, 3 are sector-like
    # Check if sector concentration metrics exceed 20% for Energy (Mode 2) and Utilities (Mode 3)
    sector_diag = spectral_diag["eigenvector_analysis"]
    if sector_diag["sector_map_available"]:
        # Simplified check: is the max concentration > 20%?
        m2_conc = max(sector_diag["sector_concentrations"]["mode_2"].values())
        m3_conc = max(sector_diag["sector_concentrations"]["mode_3"].values())
        is_sector = m2_conc > 0.20 and m3_conc > 0.20
        findings["modes_2_3_are_sector_like"] = {
            "status": "Confirmed" if is_sector else "Partially confirmed",
            "evidence": {"mode_2_max_conc": m2_conc, "mode_3_max_conc": m3_conc}
        }
    else:
        findings["modes_2_3_are_sector_like"] = {"status": "Unknown (No sector map)"}

    # 3. \hat\gamma > 0 for dominant modes (Fréchet domain)
    # Check static table for alpha=0.005 (a typical credible threshold)
    static_005 = static_evt_table[(static_evt_table["alpha"] == 0.005) & (static_evt_table["is_reliable"])]
    if not static_005.empty:
        # Confirm all reliable fits have positive shape parameter
        all_positive = (static_005["gamma_hat"] > 0).all()
        findings["gamma_gt_0_frechet_domain"] = {
            "status": "Confirmed" if all_positive else "Partially confirmed",
            "evidence": {"min_gamma": float(static_005["gamma_hat"].min())}
        }

    # 4. \hat\Theta < 1 under fixed thresholds (Clustering present)
    if not static_005.empty:
        # Confirm extremal index is significantly less than 1 (e.g., < 0.95)
        all_clustered = (static_005["theta_hat"] < 0.95).all()
        findings["theta_lt_1_static"] = {
            "status": "Confirmed" if all_clustered else "Not confirmed",
            "evidence": {"max_theta": float(static_005["theta_hat"].max())}
        }

    # 5. \hat\Theta -> 1 under rolling thresholds at high quantiles (Clustering removed)
    # Check dynamic table at extreme quantile alpha=0.0001
    dyn_1e4 = dynamic_evt_table[(dynamic_evt_table["alpha"] == 0.0001) & (dynamic_evt_table["is_reliable"])]
    if not dyn_1e4.empty:
        # Confirm majority of modes show theta > 0.9
        mostly_unclustered = (dyn_1e4["theta_hat"] > 0.9).mean() > 0.8
        findings["theta_to_1_dynamic"] = {
            "status": "Confirmed" if mostly_unclustered else "Partially confirmed",
            "evidence": {"mean_theta_at_1e4": float(dyn_1e4["theta_hat"].mean())}
        }

    # 6. POT-implied GEV matches empirical daily maxima
    # Check Kolmogorov-Smirnov test results from Task 18
    ks_results = gev_inference["benchmark_u15_comparisons"]
    if ks_results:
        # Confirm p-value > 0.05 (fail to reject null hypothesis of same distribution) for all modes
        all_match = all(res["match_confirmed"] for res in ks_results.values())
        findings["gev_matches_empirical"] = {
            "status": "Confirmed" if all_match else "Partially confirmed",
            "evidence": {str(k): res["ks_p_value"] for k, res in ks_results.items()}
        }

    # 7. Mode ordering changes between static and dynamic analysis
    if not static_005.empty and not dyn_1e4.empty:
        # Get positive tail rankings for static vs dynamic
        stat_pos = static_005[static_005["tail"] == "positive"].sort_values("mode")
        dyn_pos = dyn_1e4[dyn_1e4["tail"] == "positive"].sort_values("mode")

        if len(stat_pos) == len(dyn_pos) and len(stat_pos) > 2:
            # Compute Spearman rank correlation
            corr, _ = scipy.stats.spearmanr(stat_pos["gamma_hat"], dyn_pos["gamma_hat"])

            # Determine status based on correlation strength
            if corr < 0.8:
                status = "Confirmed" # Significant reordering
            elif corr < 0.95:
                status = "Partially confirmed" # Some reordering
            else:
                status = "Not confirmed" # Rankings are essentially identical

            findings["mode_ordering_changes"] = {
                "status": status,
                "evidence": {"spearman_correlation": float(corr)}
            }

    return findings

def assemble_dossier(
    all_artifacts: Dict[str, Any],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Assembles the final reproduction dossier, organizing all artifacts by phase
    and computing the validation memo.

    This function acts as the final aggregator, structuring the disparate outputs
    from all 21 tasks into a coherent, hierarchical dictionary suitable for
    serialization and distribution.

    Args:
        all_artifacts (Dict[str, Any]): Dictionary containing outputs from all pipeline tasks.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: The complete, structured reproduction dossier.
    """
    # Extract artifacts from the accumulator dictionary
    schema_report = all_artifacts["schema_report"]
    cleaning_audit = all_artifacts["cleaning_audit"]
    date_list = all_artifacts["date_list"]
    midpoint_val = all_artifacts["midpoint_val"]
    return_val = all_artifacts["return_val"]
    c_val = all_artifacts["c_val"]
    spectral_diag = all_artifacts["spectral_diag"]
    static_evt_table = all_artifacts["static_evt_table"]
    acf_data = all_artifacts["acf_data"]
    gev_inference = all_artifacts["gev_inference"]
    residual_artifact = all_artifacts["residual_artifact"]
    dynamic_evt_artifact = all_artifacts["dynamic_evt_artifact"]
    figure_data_1 = all_artifacts["figure_data_1"]
    figure_data_2 = all_artifacts["figure_data_2"]

    # Compute validation memo using the extracted artifacts
    validation_memo = validate_manuscript_findings(
        spectral_diag, static_evt_table, dynamic_evt_artifact["dynamic_fit_table"], gev_inference
    )

    # Construct the hierarchical dossier
    dossier = {
        "Phase_0_Configuration": {
            "manifest_hash": manifest.get("manifest_hash"),
            "schema_validation": schema_report
        },
        "Phase_1_Data_Engineering": {
            "cleaning_audit": cleaning_audit,
            "retained_dates_count": len(date_list),
            "midpoint_panel_summary": midpoint_val,
            "return_matrix_summary": return_val
        },
        "Phase_2_Spectral_Analysis": {
            "correlation_matrix_validation": c_val,
            "spectral_diagnostics": spectral_diag
        },
        "Phase_3_Static_EVT": {
            # Convert DataFrame to list of dicts for JSON serialization
            "static_evt_results": static_evt_table.to_dict(orient="records")
        },
        "Phase_4_ACF_and_Block_Maxima": {
            "acf_diagnostics": acf_data["diagnostics"],
            "gev_inference_summary": {
                "u15_comparisons": gev_inference["benchmark_u15_comparisons"]
            }
        },
        "Phase_5_Dynamic_Analysis": {
            "deseasonalization_diagnostics": residual_artifact["deseasonalization_diagnostics"],
            "dynamic_evt_results": dynamic_evt_artifact["dynamic_fit_table"].to_dict(orient="records")
        },
        "Phase_6_Figures_Data": {
            "rolling_threshold_example": figure_data_1,
            "dynamic_result_curves": figure_data_2
        },
        "Phase_7_Validation_Memo": validation_memo
    }

    return dossier

# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def assemble_reproduction_outputs(
    all_artifacts: Dict[str, Any],
    manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the reproduction of nonstationary diagnostics and the assembly
    of the final output package.

    Args:
        all_artifacts (Dict[str, Any]): A dictionary containing all outputs from Tasks 1-20.
        manifest (MappingProxyType): The frozen run manifest.

    Returns:
        Dict[str, Any]: The complete, serializable reproduction dossier.
    """
    # Step 1: Prepare rolling threshold figure data
    residual_modes = all_artifacts["residual_artifact"]["residual_modes"]
    dynamic_evt_artifact = all_artifacts["dynamic_evt_artifact"]

    figure_data_1 = prepare_rolling_threshold_figure_data(
        residual_modes, dynamic_evt_artifact, manifest
    )
    all_artifacts["figure_data_1"] = figure_data_1

    # Step 2: Prepare dynamic result curves
    figure_data_2 = prepare_dynamic_result_curves(dynamic_evt_artifact)
    all_artifacts["figure_data_2"] = figure_data_2

    # Step 3: Assemble dossier and validate findings
    dossier = assemble_dossier(all_artifacts, manifest)

    return dossier
````

````python
# Task 22: Create the end-to-end orchestrator pipeline

# ==============================================================================
# Task 22: Create the end-to-end orchestrator pipeline
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Helper: Deterministic File Hashing
# -------------------------------------------------------------------------------------------------------------------------------

def compute_file_hash(filepath: str, chunk_size: int = 8192) -> str:
    """
    Computes the SHA-256 cryptographic hash of a file on disk by reading it in chunks.

    This function ensures memory efficiency for large files (e.g., memory-mapped arrays)
    by processing the file stream incrementally. It provides a deterministic fingerprint
    of the file's binary content, which is essential for provenance tracking and
    verifying the integrity of cached artifacts in the research pipeline.

    Args:
        filepath (str): The absolute or relative path to the file to be hashed.
        chunk_size (int, optional): The size of each read block in bytes. Defaults to 8192 (8 KB).

    Returns:
        str: The hexadecimal representation of the SHA-256 hash.
             Returns "FILE_NOT_FOUND" if the path does not exist.
    """
    # Validate file existence to prevent runtime crashes
    if not os.path.exists(filepath):
        return "FILE_NOT_FOUND"

    # Initialize the SHA-256 hasher
    hasher = hashlib.sha256()

    # Open file in binary read mode
    with open(filepath, 'rb') as f:
        # Read the file in fixed-size chunks to bound memory usage
        for chunk in iter(lambda: f.read(chunk_size), b''):
            hasher.update(chunk)

    # Return the final digest as a hex string
    return hasher.hexdigest()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 22, Steps 1, 2, and 3: The Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def run_full_pipeline(
    df_raw_taq: pd.DataFrame,
    study_configuration: Dict[str, Any]
) -> Tuple[Dict[str, Any], Dict[str, Any]]:
    """
    End-to-end orchestrator for the Extreme Value Analysis Framework.

    Executes all 21 pipeline tasks in strict sequence. Implements rigorous memory
    management by explicitly deleting intermediate artifacts. Enforces a strict
    error policy: any exception halts execution immediately, logs the trace, and
    returns a partial dossier of all successfully completed artifacts.

    Args:
        df_raw_taq (pd.DataFrame): The raw, unprocessed TAQ DataFrame.
        study_configuration (Dict[str, Any]): The master configuration dictionary.

    Returns:
        Tuple[Dict[str, Any], Dict[str, Any]]:
            - The complete (or partial) reproduction dossier.
            - The pipeline execution log containing timing, memory, and validation data.
    """
    process = psutil.Process()
    execution_log: Dict[str, Any] = {
        "task_metrics": [],
        "global_peak_memory_bytes": 0,
        "pipeline_status": "INCOMPLETE",
        "failure_traceback": None,
        "failed_task": None
    }

    all_artifacts: Dict[str, Any] = {}
    manifest: MappingProxyType = None # type: ignore

    def execute_task(task_name: str, func: Callable, *args, **kwargs) -> Any:
        """Helper to execute a task with timing, memory tracking, and error handling."""
        logging.info(f"Starting {task_name}...")
        mem_before = process.memory_info().rss
        t_start = time.perf_counter()
        c_start = time.process_time()

        try:
            result = func(*args, **kwargs)
        except Exception as e:
            logging.error(f"Exception in {task_name}: {str(e)}")
            raise e # Re-raise to be caught by the main orchestrator block

        t_end = time.perf_counter()
        c_end = time.process_time()
        mem_after = process.memory_info().rss

        # Update global peak memory (approximate, as peak might occur inside the function)
        # For true peak, OS-level tracking (rusage) is better, but psutil is cross-platform
        current_peak = process.memory_info().peak_wset if hasattr(process.memory_info(), 'peak_wset') else max(mem_before, mem_after)
        if current_peak > execution_log["global_peak_memory_bytes"]:
            execution_log["global_peak_memory_bytes"] = current_peak

        metrics = {
            "task_name": task_name,
            "wall_clock_duration_sec": t_end - t_start,
            "cpu_duration_sec": c_end - c_start,
            "memory_rss_before_bytes": mem_before,
            "memory_rss_after_bytes": mem_after,
            "status": "PASS"
        }
        execution_log["task_metrics"].append(metrics)
        logging.info(f"Completed {task_name} in {metrics['wall_clock_duration_sec']:.2f}s")
        return result

    try:
        # ==============================================================================
        # Phase 0: Configuration and Validation
        # ==============================================================================
        manifest, config_report = execute_task(
            "Task 1: validate_study_configuration",
            validate_study_configuration, study_configuration
        )
        all_artifacts["config_report"] = config_report

        schema_report = execute_task(
            "Task 2: validate_raw_dataframe",
            validate_raw_dataframe, df_raw_taq, manifest
        )
        all_artifacts["schema_report"] = schema_report

        # ==============================================================================
        # Phase 1: Data Engineering
        # ==============================================================================
        df_cleaned, cleaning_audit = execute_task(
            "Task 3: cleanse_raw_quotes",
            cleanse_raw_quotes, df_raw_taq, manifest
        )
        all_artifacts["cleaning_audit"] = cleaning_audit
        del df_raw_taq # Free raw data
        gc.collect()

        df_session, df_pre_session, date_list = execute_task(
            "Task 4: filter_sample_dates_and_universe",
            filter_sample_dates_and_universe, df_cleaned, manifest
        )
        all_artifacts["date_list"] = date_list
        del df_cleaned
        gc.collect()

        df_sparse_midpoints, df_pre_session_midpoints = execute_task(
            "Task 5: assign_second_buckets_and_select",
            assign_second_buckets_and_select, df_session, df_pre_session, manifest
        )
        del df_session
        del df_pre_session
        gc.collect()

        midpoint_panel_path, midpoint_val = execute_task(
            "Task 6: build_complete_midpoint_panel",
            build_complete_midpoint_panel, df_sparse_midpoints, df_pre_session_midpoints, date_list, manifest
        )
        all_artifacts["midpoint_val"] = midpoint_val
        all_artifacts["midpoint_panel_path"] = midpoint_panel_path
        del df_sparse_midpoints
        del df_pre_session_midpoints
        gc.collect()

        returns_path, day_boundary_map, return_val = execute_task(
            "Task 7: compute_log_returns",
            compute_log_returns, midpoint_panel_path, date_list, manifest
        )
        all_artifacts["return_val"] = return_val
        all_artifacts["returns_path"] = returns_path
        all_artifacts["day_boundary_map"] = day_boundary_map

        # Optionally delete midpoint panel from disk if space is tight, but we keep it for the dossier reference

        normalized_path, means, stds, norm_val = execute_task(
            "Task 8: standardize_returns",
            standardize_returns, returns_path, manifest
        )
        all_artifacts["norm_val"] = norm_val
        all_artifacts["normalized_path"] = normalized_path

        # ==============================================================================
        # Phase 2: Spectral Analysis
        # ==============================================================================
        c_sym, c_val = execute_task(
            "Task 9: compute_correlation_matrix",
            compute_correlation_matrix, normalized_path, manifest
        )
        all_artifacts["c_val"] = c_val
        all_artifacts["c_sym"] = c_sym

        eigenvalues, eigenvectors, decomp_diag = execute_task(
            "Task 10: spectral_decomposition",
            spectral_decomposition, c_sym, manifest
        )
        all_artifacts["decomp_diag"] = decomp_diag

        y_path, r_path, u_adjusted, mode_metadata, whitening_val = execute_task(
            "Task 11: construct_whitened_modes",
            construct_whitened_modes, normalized_path, eigenvectors, eigenvalues, manifest
        )
        all_artifacts["whitening_val"] = whitening_val
        all_artifacts["r_path"] = r_path

        spectral_diag = execute_task(
            "Task 12: generate_spectral_diagnostics",
            generate_spectral_diagnostics, eigenvalues, u_adjusted, y_path, r_path, manifest
        )
        all_artifacts["spectral_diag"] = spectral_diag

        # ==============================================================================
        # Phase 3: Static EVT
        # ==============================================================================
        exceedance_objects = execute_task(
            "Task 13: extract_static_pot_exceedances",
            extract_static_pot_exceedances, r_path, manifest
        )

        gpd_fit_table = execute_task(
            "Task 14: fit_gpd_mle",
            fit_gpd_mle, exceedance_objects, manifest
        )

        gpd_fit_table, qq_data, credible_regions = execute_task(
            "Task 15: evaluate_gpd_fit_quality",
            evaluate_gpd_fit_quality, gpd_fit_table, exceedance_objects, manifest
        )

        static_evt_table = execute_task(
            "Task 16: estimate_extremal_index",
            estimate_extremal_index, exceedance_objects, gpd_fit_table, manifest
        )
        all_artifacts["static_evt_table"] = static_evt_table

        # ==============================================================================
        # Phase 4: ACF and Block Maxima
        # ==============================================================================
        acf_data = execute_task(
            "Task 17: compute_absolute_return_acf",
            compute_absolute_return_acf, r_path, manifest
        )
        all_artifacts["acf_data"] = acf_data

        gev_inference = execute_task(
            "Task 18: infer_gev_from_pot",
            infer_gev_from_pot, static_evt_table, r_path, day_boundary_map, manifest
        )
        all_artifacts["gev_inference"] = gev_inference

        # ==============================================================================
        # Phase 5: Dynamic Analysis
        # ==============================================================================
        residual_artifact = execute_task(
            "Task 19: build_residual_modes",
            build_residual_modes, r_path, manifest
        )
        all_artifacts["residual_artifact"] = residual_artifact

        dynamic_evt_artifact = execute_task(
            "Task 20: fit_dynamic_evt_models",
            fit_dynamic_evt_models, residual_artifact["residual_modes"], manifest
        )
        all_artifacts["dynamic_evt_artifact"] = dynamic_evt_artifact

        # ==============================================================================
        # Phase 6 & 7: Assembly
        # ==============================================================================
        dossier = execute_task(
            "Task 21: assemble_reproduction_outputs",
            assemble_reproduction_outputs, all_artifacts, manifest
        )

        execution_log["pipeline_status"] = "SUCCESS"

    except Exception as e:
        # Strict Error Policy: Halt, log trace, assemble partial dossier
        execution_log["pipeline_status"] = "FAILED"
        execution_log["failure_traceback"] = traceback.format_exc()
        # The last appended task metric is the one that failed (or the one we were about to start)
        execution_log["failed_task"] = execution_log["task_metrics"][-1]["task_name"] if execution_log["task_metrics"] else "Initialization"

        logging.critical(f"Pipeline failed at {execution_log['failed_task']}. Assembling partial dossier.")

        # Attempt to assemble a partial dossier with whatever artifacts exist
        # We use a safe assembly function that uses .get() to avoid KeyErrors
        def safe_assemble(artifacts, man):
            partial = {"WARNING": "THIS IS A PARTIAL DOSSIER DUE TO PIPELINE FAILURE."}
            if "schema_report" in artifacts: partial["Phase_0"] = artifacts["schema_report"]
            if "cleaning_audit" in artifacts: partial["Phase_1"] = artifacts["cleaning_audit"]
            if "static_evt_table" in artifacts: partial["Phase_3"] = artifacts["static_evt_table"].to_dict(orient="records")
            return partial

        dossier = safe_assemble(all_artifacts, manifest)

    # ==============================================================================
    # Finalize Execution Log
    # ==============================================================================

    # Embed manifest
    if manifest is not None:
        # Convert MappingProxyType back to dict for JSON serialization
        execution_log["frozen_manifest"] = json.loads(json.dumps(dict(manifest), default=str))

    # Compute cryptographic hash of the dossier
    # We serialize the dossier deterministically
    try:
        dossier_json = json.dumps(dossier, sort_keys=True, default=str)
        dossier_hash = hashlib.sha256(dossier_json.encode('utf-8')).hexdigest()
        execution_log["dossier_sha256_hash"] = dossier_hash
    except Exception as hash_e:
        execution_log["dossier_sha256_hash"] = f"HASH_FAILED: {str(hash_e)}"

    # Aggregate validation invariants
    validation_summary = []
    for key in ["config_report", "schema_report", "midpoint_val", "return_val", "c_val", "whitening_val"]:
        if key in all_artifacts:
            report = all_artifacts[key]
            for check_name, check_data in report.items():
                if isinstance(check_data, dict) and "status" in check_data:
                    validation_summary.append({
                        "phase": key,
                        "check_name": check_name,
                        "status": check_data["status"]
                    })
    execution_log["validation_invariant_summary"] = validation_summary

    return dossier, execution_log
````

````python
# Task 23: Conduct robustness and sensitivity analysis

# ==============================================================================
# Task 23: Conduct robustness and sensitivity analysis
# ==============================================================================

# Re-import SeriesKey for type hinting consistency
class SeriesKey(NamedTuple):
    mode: int
    tail: str

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 1: Threshold sensitivity analysis for the static EVT branch
# -------------------------------------------------------------------------------------------------------------------------------

def run_threshold_sensitivity(
    r_memmap_path: str,
    baseline_manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Re-runs the static EVT branch using a fine alpha-grid to test the stability
    of the shape parameter \hat\gamma across thresholds.

    Args:
        r_memmap_path (str): Path to the cached whitened modes memmap.
        baseline_manifest (MappingProxyType): The baseline run manifest.

    Returns:
        Dict[str, Any]: Threshold sensitivity report including stability ranges.
    """
    # Create a mutable copy of the manifest to inject the perturbed grid
    perturbed_config = json.loads(json.dumps(dict(baseline_manifest), default=str))

    # Define a fine grid: 30 points logarithmically spaced from 10^-1 to 10^-4
    fine_grid = np.logspace(-1, -4, 30).tolist()
    perturbed_config["algorithmic_parameters"]["static_evt"]["tail_quantile_grid_alpha"] = fine_grid

    # We don't need to re-freeze for this isolated test, but we pass it as a dict
    # (Assuming the downstream functions can accept a dict if they only read from it)
    # To be strictly safe, we wrap it in MappingProxyType
    def _freeze(obj):
        if isinstance(obj, dict): return MappingProxyType({k: _freeze(v) for k, v in obj.items()})
        if isinstance(obj, list): return tuple(_freeze(v) for v in obj)
        return obj
    perturbed_manifest = _freeze(perturbed_config)

    # Execute static EVT pipeline
    exceedance_objects = extract_static_pot_exceedances(r_memmap_path, perturbed_manifest)
    gpd_fit_table = fit_gpd_mle(exceedance_objects, perturbed_manifest)
    gpd_fit_table, qq_data, credible_regions = evaluate_gpd_fit_quality(gpd_fit_table, exceedance_objects, perturbed_manifest)
    # We skip extremal index here as we only care about \hat\gamma stability

    stability_report = {}

    # Analyze stability
    grouped = gpd_fit_table[gpd_fit_table["is_reliable"]].groupby(["mode", "tail"])
    for (mode, tail), group in grouped:
        group_sorted = group.sort_values("alpha", ascending=False).reset_index(drop=True)

        max_stable_len = 0
        best_range = None
        best_gamma = None

        # Find maximal contiguous range where max(gamma) - min(gamma) < mean(se_gamma)
        n_pts = len(group_sorted)
        for i in range(n_pts):
            for j in range(i + 3, n_pts + 1): # Minimum window size of 3
                window = group_sorted.iloc[i:j]
                gamma_range = window["gamma_hat"].max() - window["gamma_hat"].min()
                mean_se = window["se_gamma"].mean()

                if gamma_range < mean_se:
                    if len(window) > max_stable_len:
                        max_stable_len = len(window)
                        best_range = [float(window["alpha"].max()), float(window["alpha"].min())]
                        # Gamma at minimum rho within this stable window
                        min_rho_idx = window["rho"].idxmin()
                        best_gamma = float(window.loc[min_rho_idx, "gamma_hat"])

        stability_report[f"mode_{mode}_{tail}"] = {
            "max_stable_window_size": max_stable_len,
            "stable_alpha_range": best_range,
            "representative_gamma": best_gamma,
            "is_stable": max_stable_len >= 5 # Arbitrary threshold for "stable" on a 30-pt grid
        }

    return {
        "fine_grid_alphas": fine_grid,
        "stability_analysis": stability_report
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 2: Rolling-window sensitivity analysis
# -------------------------------------------------------------------------------------------------------------------------------

def _run_single_window(
    w: int,
    residual_modes: Dict[int, np.ndarray],
    baseline_config: Dict[str, Any]
) -> pd.DataFrame:
    """
    Worker function for parallel execution of the dynamic EVT analysis with a
    specific rolling window size.

    This helper function isolates the execution of a single window-size scenario,
    allowing for concurrent processing in the sensitivity analysis. It creates a
    perturbed configuration, freezes it into an immutable manifest, and invokes
    the dynamic EVT fitting pipeline.

    Args:
        w (int): The rolling window size in seconds (e.g., 5000, 10000, 20000).
        residual_modes (Dict[int, np.ndarray]): The cached deseasonalized residual series.
        baseline_config (Dict[str, Any]): The mutable baseline configuration dictionary.

    Returns:
        pd.DataFrame: The resulting dynamic fit table for the specified window size,
                      augmented with a 'W' column.
    """
    # Create a deep copy of the baseline configuration to avoid side effects
    perturbed_config = copy.deepcopy(baseline_config)

    # Inject the perturbed window parameters
    perturbed_config["algorithmic_parameters"]["dynamic_analysis"]["rolling_window_size_W_seconds"] = w
    perturbed_config["algorithmic_parameters"]["dynamic_analysis"]["rolling_window_min_periods"] = w

    # Helper to recursively freeze the configuration into a MappingProxyType
    def _freeze(obj: Any) -> Any:
        if isinstance(obj, dict):
            return MappingProxyType({k: _freeze(v) for k, v in obj.items()})
        if isinstance(obj, list):
            return tuple(_freeze(v) for v in obj)
        return obj

    # Freeze the perturbed configuration
    perturbed_manifest = _freeze(perturbed_config)

    # Execute the dynamic EVT pipeline for this specific window size
    # fit_dynamic_evt_models is assumed to be available in the global namespace
    dynamic_evt_artifact = fit_dynamic_evt_models(residual_modes, perturbed_manifest)

    # Extract the results table and tag it with the window size
    df = dynamic_evt_artifact["dynamic_fit_table"]
    df["W"] = w

    return df

def run_rolling_window_sensitivity(
    residual_modes: Dict[int, np.ndarray],
    baseline_manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Re-runs the dynamic EVT branch for alternative window lengths to test robustness.

    Args:
        residual_modes (Dict): Cached deseasonalized residual series.
        baseline_manifest (MappingProxyType): The baseline run manifest.

    Returns:
        Dict[str, Any]: Rolling window sensitivity report.
    """
    window_lengths = [5000, 10000, 15000, 20000]
    baseline_config = json.loads(json.dumps(dict(baseline_manifest), default=str))

    all_results = []

    # Parallelize across window lengths
    with concurrent.futures.ProcessPoolExecutor(max_workers=min(4, len(window_lengths))) as executor:
        futures = {
            executor.submit(_run_single_window, w, residual_modes, baseline_config): w
            for w in window_lengths
        }
        for future in concurrent.futures.as_completed(futures):
            try:
                df = future.result()
                all_results.append(df)
            except Exception as e:
                w = futures[future]
                import logging
                logging.error(f"Window size {w} failed: {e}")

    if not all_results:
        return {"status": "FAILED"}

    combined_df = pd.concat(all_results, ignore_index=True)

    # Robustness checks
    robustness_checks = {}

    # 1. Robustness of \hat\gamma when \alpha * W >= 1
    # Check variance of \hat\gamma across W for alpha=1e-3 (where alpha*W >= 5 for all W)
    df_1e3 = combined_df[(combined_df["alpha"] == 0.001) & (combined_df["is_reliable"])]
    gamma_std_across_w = df_1e3.groupby(["mode", "tail"])["gamma_hat"].std().mean()
    robustness_checks["gamma_robustness_alpha_1e3"] = {
        "mean_std_across_W": float(gamma_std_across_w),
        "is_robust": gamma_std_across_w < 0.05
    }

    # 2. \hat\Theta -> 1 at high quantiles across all W
    df_1e4 = combined_df[(combined_df["alpha"] == 0.0001) & (combined_df["is_reliable"])]
    theta_mean_across_w = df_1e4.groupby("W")["theta_hat"].mean().to_dict()
    robustness_checks["theta_to_1_across_W"] = {
        "mean_theta_by_W": {str(k): float(v) for k, v in theta_mean_across_w.items()},
        "is_robust": all(v > 0.85 for v in theta_mean_across_w.values())
    }

    return {
        "window_lengths_tested": window_lengths,
        "robustness_checks": robustness_checks,
        "combined_results_summary": combined_df.groupby(["W", "alpha"])["gamma_hat"].mean().to_dict()
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 3: Static-versus-dynamic comparison and final robustness report
# -------------------------------------------------------------------------------------------------------------------------------

def compile_robustness_report(
    static_evt_table: pd.DataFrame,
    dynamic_evt_table: pd.DataFrame,
    threshold_sensitivity: Dict[str, Any],
    window_sensitivity: Dict[str, Any],
    gev_inference: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Compiles the final robustness report, comparing static vs dynamic results
    and formally evaluating the manuscript's key findings.

    This function synthesizes the results from the baseline static analysis,
    the baseline dynamic analysis, and the sensitivity analyses to produce
    a definitive confirmation assessment for each of the manuscript's core claims.

    Args:
        static_evt_table (pd.DataFrame): Results from the static EVT analysis.
        dynamic_evt_table (pd.DataFrame): Results from the dynamic EVT analysis.
        threshold_sensitivity (Dict[str, Any]): Results from the threshold sensitivity analysis.
        window_sensitivity (Dict[str, Any]): Results from the rolling-window sensitivity analysis.
        gev_inference (Dict[str, Any]): Results from the GEV inference and comparison.

    Returns:
        Dict[str, Any]: A structured dictionary containing the full robustness report.
    """
    report: Dict[str, Any] = {}

    # 1. Head-to-head comparison table at reference alpha = 0.005
    # This alpha is chosen as a representative credible threshold
    ref_alpha = 0.005

    # Filter tables for the reference alpha and set index for easy lookup
    stat_ref = static_evt_table[static_evt_table["alpha"] == ref_alpha].set_index(["mode", "tail"])
    dyn_ref = dynamic_evt_table[dynamic_evt_table["alpha"] == ref_alpha].set_index(["mode", "tail"])

    comparison_table: Dict[str, Dict[str, Dict[str, float]]] = {}

    for idx in stat_ref.index:
        if idx in dyn_ref.index:
            # Extract key metrics for side-by-side comparison
            comparison_table[f"mode_{idx[0]}_{idx[1]}"] = {
                "static": {
                    "gamma": float(stat_ref.loc[idx, "gamma_hat"]),
                    "theta": float(stat_ref.loc[idx, "theta_hat"]),
                    "rho": float(stat_ref.loc[idx, "rho"])
                },
                "dynamic": {
                    "gamma": float(dyn_ref.loc[idx, "gamma_hat"]),
                    "theta": float(dyn_ref.loc[idx, "theta_hat"]),
                    "rho": float(dyn_ref.loc[idx, "rho"])
                }
            }
    report["head_to_head_comparison_alpha_005"] = comparison_table

    # 2. Mode ordering analysis
    # Compare the ranking of modes by \hat\gamma (tail heaviness) between static and dynamic
    # Focus on the positive tail as it represents market crashes/spikes
    stat_pos = stat_ref.xs("positive", level="tail").sort_values("gamma_hat", ascending=False)
    dyn_pos = dyn_ref.xs("positive", level="tail").sort_values("gamma_hat", ascending=False)

    stat_order = stat_pos.index.tolist()
    dyn_order = dyn_pos.index.tolist()

    # Compute Spearman rank correlation to quantify the reordering
    # We align the dynamic series to the static index to ensure we compare the same modes
    corr, _ = scipy.stats.spearmanr(stat_pos["gamma_hat"], dyn_pos.loc[stat_pos.index, "gamma_hat"])

    report["mode_ordering"] = {
        "static_ranking": stat_order,
        "dynamic_ranking": dyn_order,
        "spearman_correlation": float(corr),
        "ordering_changed": stat_order != dyn_order
    }

    # 3. Formal Confirmation Assessment
    # Apply the strict decision rules defined in the disambiguation
    assessments: Dict[str, Dict[str, str]] = {}

    # Finding 1: Fréchet domain (\hat\gamma > 0) for dominant modes
    # Check if all reliable static fits at alpha=0.005 have positive gamma
    all_pos = (stat_ref["gamma_hat"] > 0).all()
    assessments["frechet_domain_dominant_modes"] = {
        "determination": "Confirmed" if all_pos else "Not confirmed",
        "evidence": f"Min gamma at alpha=0.005 is {stat_ref['gamma_hat'].min():.3f}"
    }

    # Finding 2: \hat\Theta < 1 under fixed thresholds (Clustering present)
    # Check if all reliable static fits show significant clustering
    all_clustered = (stat_ref["theta_hat"] < 0.95).all()
    assessments["theta_lt_1_static"] = {
        "determination": "Confirmed" if all_clustered else "Not confirmed",
        "evidence": f"Max theta at alpha=0.005 is {stat_ref['theta_hat'].max():.3f}"
    }

    # Finding 3: \hat\Theta -> 1 under rolling thresholds at high quantiles (Clustering removed)
    # Check dynamic results at the most extreme quantile alpha=0.0001
    dyn_1e4 = dynamic_evt_table[dynamic_evt_table["alpha"] == 0.0001]
    # Criterion: Mean theta across modes > 0.9
    mostly_unclustered = (dyn_1e4["theta_hat"] > 0.9).mean() > 0.8
    assessments["theta_to_1_dynamic"] = {
        "determination": "Confirmed" if mostly_unclustered else "Partially confirmed",
        "evidence": f"Mean theta at alpha=1e-4 is {dyn_1e4['theta_hat'].mean():.3f}"
    }

    # Finding 4: POT-implied GEV matches empirical daily maxima
    # Check KS test p-values from Task 18
    ks_results = gev_inference["benchmark_u15_comparisons"]
    # Criterion: All p-values > 0.05 (fail to reject null hypothesis)
    all_match = all(res["ks_p_value"] > 0.05 for res in ks_results.values())
    assessments["gev_matches_empirical"] = {
        "determination": "Confirmed" if all_match else "Partially confirmed",
        "evidence": "All KS p-values > 0.05" if all_match else "Some KS p-values <= 0.05"
    }

    # Finding 5: Mode ordering changes between static and dynamic
    # Criterion based on Spearman correlation strength
    if corr < 0.8:
        det = "Confirmed" # Significant reordering
    elif corr < 0.95:
        det = "Partially confirmed" # Some reordering
    else:
        det = "Not confirmed" # Rankings are essentially identical

    assessments["mode_ordering_changes"] = {
        "determination": det,
        "evidence": f"Spearman correlation is {corr:.3f}"
    }

    report["formal_assessments"] = assessments
    report["threshold_sensitivity_summary"] = threshold_sensitivity
    report["window_sensitivity_summary"] = window_sensitivity

    return report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def run_robustness_analysis(
    baseline_artifacts: Dict[str, Any],
    baseline_manifest: MappingProxyType
) -> Dict[str, Any]:
    """
    Orchestrates the robustness and sensitivity analysis.

    Executes threshold sensitivity on the static branch, rolling-window sensitivity
    on the dynamic branch, and compiles a formal robustness report evaluating
    the manuscript's key claims.

    Args:
        baseline_artifacts (Dict[str, Any]): Cached artifacts from the baseline run.
        baseline_manifest (MappingProxyType): The baseline run manifest.

    Returns:
        Dict[str, Any]: The final robustness report.
    """
    r_memmap_path = baseline_artifacts["r_path"]
    residual_modes = baseline_artifacts["residual_artifact"]["residual_modes"]
    static_evt_table = baseline_artifacts["static_evt_table"]
    dynamic_evt_table = baseline_artifacts["dynamic_evt_artifact"]["dynamic_fit_table"]
    gev_inference = baseline_artifacts["gev_inference"]

    # Step 1: Threshold sensitivity
    logging.info("Running threshold sensitivity analysis...")
    threshold_sens = run_threshold_sensitivity(r_memmap_path, baseline_manifest)

    # Step 2: Window sensitivity
    logging.info("Running rolling-window sensitivity analysis...")
    window_sens = run_rolling_window_sensitivity(residual_modes, baseline_manifest)

    # Step 3: Compile report
    logging.info("Compiling final robustness report...")
    robustness_report = compile_robustness_report(
        static_evt_table, dynamic_evt_table, threshold_sens, window_sens, gev_inference
    )

    return robustness_report
````

````python
# Top-Level Orchestrator

# ==============================================================================
# Top-Level Orchestrator: Execute the Complete Research Pipeline
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Top-Level Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def execute_research_pipeline(
    df_raw_taq: pd.DataFrame,
    study_configuration: Dict[str, Any]
) -> Dict[str, Any]:
    """
    The top-level orchestrator for the Extreme Value Analysis Framework.

    Executes, in a sequential ashion, the baseline pipeline (Tasks 1-22),
    deterministically reconstructs the required memory state, and executes the
    robustness and sensitivity analysis (Task 23). Returns a unified,
    implementation-grade master dictionary containing all research artifacts.

    Args:
        df_raw_taq (pd.DataFrame): The raw, unprocessed TAQ DataFrame.
        study_configuration (Dict[str, Any]): The master configuration dictionary.

    Returns:
        Dict[str, Any]: A master dictionary containing:
            - 'baseline_dossier': The complete reproduction dossier.
            - 'execution_log': Timing, memory, and validation metrics.
            - 'robustness_report': Sensitivity analysis and formal findings assessment.
    """
    master_output: Dict[str, Any] = {}

    # ==============================================================================
    # Step 1: Execute the Baseline Pipeline (Task 22)
    # ==============================================================================
    logging.info("Initiating Baseline Research Pipeline (Tasks 1-22)...")

    dossier, execution_log = run_full_pipeline(df_raw_taq, study_configuration)

    master_output["baseline_dossier"] = dossier
    master_output["execution_log"] = execution_log

    # Strict Error Policy: Halt if the baseline pipeline failed
    if execution_log.get("pipeline_status") != "SUCCESS":
        logging.critical("Baseline pipeline failed. Aborting robustness analysis and returning partial results.")
        master_output["robustness_report"] = {"status": "ABORTED_DUE_TO_BASELINE_FAILURE"}
        return master_output

    logging.info("Baseline Pipeline completed successfully. Preparing for Robustness Analysis...")

    # ==============================================================================
    # Step 2: Deterministically Reconstruct Baseline Artifacts for Task 23
    # ==============================================================================
    # Task 22 aggressively garbage-collects to save RAM. We must reconstruct the
    # specific in-memory artifacts required by Task 23 using the deterministic
    # manifest hash and the serialized dossier.

    try:
        # 2a. Reconstruct and freeze the manifest
        manifest_dict = execution_log["frozen_manifest"]

        def _recursive_freeze(obj: Any) -> Any:
            if isinstance(obj, dict):
                return MappingProxyType({k: _recursive_freeze(v) for k, v in obj.items()})
            if isinstance(obj, list):
                return tuple(_recursive_freeze(v) for v in obj)
            return obj

        manifest: MappingProxyType = _recursive_freeze(manifest_dict)
        manifest_hash = manifest["manifest_hash"]

        # 2b. Reconstruct the whitened modes memmap path (r_path)
        temp_dir = tempfile.gettempdir()
        r_path = os.path.join(temp_dir, f"whitened_modes_R_{manifest_hash}.npy")

        if not os.path.exists(r_path):
            raise FileNotFoundError(f"Critical Error: Whitened modes memmap not found at {r_path}. Cannot proceed with sensitivity analysis.")

        # 2c. Recompute the residual modes (loads them into RAM)
        logging.info("Reconstructing deseasonalized residual modes into RAM...")
        residual_artifact = build_residual_modes(r_path, manifest)

        # 2d. Deserialize tabular data from the dossier back into pandas DataFrames
        static_evt_table = pd.DataFrame(dossier["Phase_3_Static_EVT"]["static_evt_results"])
        dynamic_evt_table = pd.DataFrame(dossier["Phase_5_Dynamic_Analysis"]["dynamic_evt_results"])

        # 2e. Extract GEV inference comparisons
        gev_inference = {
            "benchmark_u15_comparisons": dossier["Phase_4_ACF_and_Block_Maxima"]["gev_inference_summary"]["u15_comparisons"]
        }

        # 2f. Package the baseline artifacts exactly as expected by Task 23
        baseline_artifacts = {
            "r_path": r_path,
            "residual_artifact": {"residual_modes": residual_artifact["residual_modes"]},
            "static_evt_table": static_evt_table,
            "dynamic_evt_artifact": {"dynamic_fit_table": dynamic_evt_table},
            "gev_inference": gev_inference
        }

    except Exception as e:
        logging.error(f"Failed to reconstruct baseline artifacts for Task 23: {str(e)}")
        master_output["robustness_report"] = {"status": "FAILED_ARTIFACT_RECONSTRUCTION", "error": str(e)}
        return master_output

    # ==============================================================================
    # Step 3: Execute Robustness and Sensitivity Analysis (Task 23)
    # ==============================================================================
    logging.info("Initiating Robustness and Sensitivity Analysis (Task 23)...")

    try:
        robustness_report = run_robustness_analysis(baseline_artifacts, manifest)
        master_output["robustness_report"] = robustness_report
        logging.info("Robustness Analysis completed successfully.")
    except Exception as e:
        logging.error(f"Exception during Robustness Analysis: {str(e)}")
        master_output["robustness_report"] = {"status": "FAILED_EXECUTION", "error": str(e)}

    # Return the unified, implementation-grade master dictionary
    return master_output
````
